## 1. Principio de funcionamiento de una máquina de refrigeración

Los ciclos de refrigeración son procesos termodinámicos que permiten la transferencia de calor desde una región de baja temperatura a una de alta temperatura. Estos ciclos son fundamentales en aplicaciones como aires acondicionados, refrigeradores y sistemas de refrigeración industrial. En esta lección, abordaremos los principios fundamentales de las máquinas de refrigeración.

## 2. Química y Termodinámica del Cambio de Fase en Ciclos de Refrigeración por Compresión

El cambio de fase es un fenómeno fundamental en los ciclos de refrigeración, y su comprensión es esencial para el diseño y operación eficiente de los sistemas de refrigeración. En los ciclos de refrigeración por compresión, el refrigerante experimenta cambios de fase repetidos entre líquido y vapor.

### 2.1. Química del Cambio de Fase

El cambio de fase implica una transición de una fase a otra, como de líquido a gas (ebullición) o de gas a líquido (condensación). A nivel molecular, esto se traduce en cambios en la energía cinética y potencial de las moléculas del refrigerante:

- **Ebullición:** Cuando un líquido se calienta, sus moléculas ganan energía cinética. Al alcanzar el punto de ebullición, la energía cinética es suficiente para superar las fuerzas intermoleculares que mantienen las moléculas juntas en la fase líquida, permitiendo que se evaporen y formen vapor.

- **Condensación:** Al enfriarse, las moléculas de vapor pierden energía cinética. Cuando la energía cinética es insuficiente para superar las fuerzas intermoleculares, las moléculas se agrupan y condensan, formando líquido.

Es importante destacar que el cambio de fase ocurre a una temperatura constante, conocida como temperatura de saturación, para una presión dada. Durante este proceso, el refrigerante absorbe o libera calor latente sin cambiar su temperatura.

### 2.2. Termodinámica del Cambio de Fase

Desde una perspectiva termodinámica, el cambio de fase implica cambios en varias propiedades termodinámicas, como la entalpía, la entropía y el volumen específico.

- **Entalpía (h):** La entalpía es una medida de la energía total de un sistema. Durante un cambio de fase, la entalpía del refrigerante cambia debido al calor latente asociado con la transición. Por ejemplo, durante la ebullición, el refrigerante absorbe calor latente de vaporización, lo que resulta en un aumento de su entalpía. Durante la condensación, el refrigerante libera calor latente de condensación, lo que disminuye su entalpía.

- **Entropía (s):** La entropía es una medida del desorden o aleatoriedad molecular de un sistema. Durante el cambio de fase, la entropía también experimenta cambios. En la ebullición, el paso de líquido a gas implica un aumento en el desorden molecular, lo que se traduce en un aumento de la entropía. Por otro lado, durante la condensación, el paso de gas a líquido implica una disminución en el desorden molecular, lo que resulta en una disminución de la entropía.

- **Volumen específico (v):** El volumen específico es el volumen por unidad de masa. Durante el cambio de fase, el volumen específico del refrigerante cambia significativamente. En la ebullición, el refrigerante pasa de un estado líquido denso a un estado gaseoso menos denso, lo que resulta en un aumento en el volumen específico. Durante la condensación, el proceso es inverso, y el volumen específico disminuye.

## 3. Diagrama Presión–Entalpía (P–h)

Para comprender a fondo la termodinámica del cambio de fase en los ciclos de refrigeración, es esencial considerar las propiedades termodinámicas en diferentes puntos del ciclo y cómo estas propiedades varían con las condiciones operativas.


El diagrama P-h es una herramienta esencial para visualizar y analizar los ciclos de refrigeración. En este diagrama, la entalpía se traza en el eje horizontal y la presión en el eje vertical. Las curvas de saturación líquida y saturación de vapor definen la región de mezcla bifásica.

Las principales características del diagrama P-h incluyen:

- **Línea de Saturación Líquida:** Representa las condiciones bajo las cuales el refrigerante está en equilibrio como líquido saturado. A la izquierda de esta línea, el refrigerante está en estado subenfriado.

- **Línea de Saturación de Vapor:** Representa las condiciones bajo las cuales el refrigerante está en equilibrio como vapor saturado. A la derecha de esta línea, el refrigerante está en estado sobrecalentado.

- **Región Bifásica:** El área entre las líneas de saturación líquida y de vapor. En esta región, el refrigerante existe en equilibrio como una mezcla de líquido y vapor.

El diagrama P-h también muestra las líneas de isoterma, que representan condiciones de temperatura constante. Estas líneas son útiles para analizar los cambios de fase y las propiedades termodinámicas en diferentes puntos del ciclo de refrigeración.


![image](https://amechieneer.com/wp-content/uploads/2021/04/image-1-908x600.png)


El ciclo de Rankine es un modelo idealizado que se utiliza para describir el funcionamiento de máquinas térmicas, como las turbinas de vapor y los ciclos de refrigeración. Aunque en la práctica existen pérdidas y no todos los procesos son isentrópicos, este modelo proporciona una base para entender el funcionamiento teórico de estos sistemas.

El ciclo de Rankine idealizado consta de los siguientes procesos:

1. **Compresión isentrópica:** El refrigerante, en estado gaseoso a baja presión, es comprimido isentrópicamente en el compresor, aumentando su presión y temperatura.
2. **Condensación isobárica:** En el condensador, el refrigerante libera calor al entorno y se condensa, cambiando de estado gaseoso a líquido a presión constante.
3. **Expansión isentrópica:** A través de la válvula de expansión, el refrigerante se expande isentrópicamente, reduciendo su presión y temperatura.
4. **Evaporación isobárica:** En el evaporador, el refrigerante absorbe calor del entorno y se evapora, cambiando de estado líquido a gaseoso a presión constante.



## 4. Componentes básicos de un ciclo de refrigeración por compresión

Un ciclo de refrigeración típico consta de los siguientes componentes principales:

1. **Compresor:** Aumenta la presión y la temperatura del refrigerante.
2. **Condensador:** Libera calor del refrigerante, lo que provoca que se condense y cambie de fase de gas a líquido.
3. **Válvula de expansión:** Reduce la presión del refrigerante, lo que provoca una disminución en su temperatura.
4. **Evaporador:** Absorbe calor del entorno, lo que provoca que el refrigerante se evapore y cambie de fase de líquido a gas.

El refrigerante circula a través de estos componentes en un ciclo cerrado, experimentando cambios de fase y de temperatura.

El ciclo de refrigeración por compresión de vapor es el ciclo más comúnmente utilizado en la mayoría de las aplicaciones de refrigeración. Consiste en cuatro componentes principales: el compresor, el condensador, la válvula de expansión y el evaporador.

### Proceso del ciclo:
1. **Compresión:** El refrigerante en estado de vapor a baja presión es comprimido por el compresor, aumentando su temperatura y presión.
2. **Condensación:** El vapor de refrigerante a alta presión pasa por el condensador, donde libera calor y se condensa a líquido.
3. **Expansión:** El líquido refrigerante pasa por la válvula de expansión, donde su presión y temperatura disminuyen drásticamente.
4. **Evaporación:** El refrigerante a baja presión y temperatura absorbe calor del espacio a refrigerar en el evaporador y se evapora, volviendo al compresor para comenzar el ciclo nuevamente.

### 4.1. Ecuación de Balance de Energía

La energía total en el sistema se conserva, por lo que el calor absorbido en el evaporador más el trabajo del compresor es igual al calor liberado en el condensador:

$$ Q_{in} + W = Q_{out} $$

### 4.2. Coeficiente de Rendimiento (COP)

El COP es una medida de la eficiencia de un ciclo de refrigeración y se define como la relación entre el calor absorbido en el evaporador y el trabajo realizado por el compresor:

$$ COP = \frac{Q_{evap}}{W_{comp}} $$

Donde:
- $Q_{evap}$ es la cantidad de calor absorbido en el evaporador.
- $W_{comp}$ es el trabajo realizado por el compresor.

### 4.3. Ecuación de la Eficiencia del Compresor ($\eta_c$)

La eficiencia del compresor se define como la relación entre el trabajo real y el trabajo ideal necesarios para comprimir el refrigerante:

$$ \eta_c = \frac{W_{real}}{W_{ideal}} $$

## 5. Ciclo de Refrigeración por Absorción

![image](https://www.systema.it/assets/images/immagini-cooling/Schema_principio_assorbimento_EN_2018_oriz.jpg)

El ciclo de refrigeración por absorción es una alternativa al ciclo de compresión de vapor, especialmente en aplicaciones donde la electricidad es costosa o no está disponible. En lugar de un compresor, utiliza un absorbedor y un generador para circular el refrigerante.

El coeficiente de rendimiento (COP) para un refrigerador por absorción se define de manera similar al ciclo de compresión de vapor, pero el trabajo del compresor es reemplazado por el calor suministrado al generador.

$$ COP = \frac{Q_{evap}}{Q_{gen}} $$

Donde:
- $Q_{evap}$ es la cantidad de calor absorbido en el evaporador.
- $Q_{gen}$ es la cantidad de calor suministrado al generador.

A diferencia de los ciclos de refrigeración por compresión, donde un solo fluido (refrigerante) experimenta cambios de fase, los ciclos de refrigeración por absorción involucran dos fluidos principales: un refrigerante y un absorbente. Estos dos fluidos interactúan entre sí, lo que resulta en fenómenos de absorción y desorción.

### 5.1. Química del Cambio de Fase

En el ciclo de absorción, el refrigerante se disuelve en el absorbente en el absorbedor, formando una solución diluida. Esta solución luego se lleva al generador, donde se aplica calor para separar el refrigerante del absorbente.

- **Absorción:** Durante este proceso, el refrigerante en forma de vapor es absorbido por el absorbente líquido. A nivel molecular, las moléculas de refrigerante se disuelven en el absorbente debido a las interacciones intermoleculares favorables entre el refrigerante y el absorbente. Esta absorción es un proceso exotérmico, lo que significa que se libera calor.

- **Desorción:** En el generador, se aplica calor a la solución diluida, lo que provoca que el refrigerante se separe del absorbente y se evapore. A nivel molecular, el calor suministrado proporciona la energía necesaria para superar las interacciones intermoleculares entre el refrigerante y el absorbente, permitiendo que el refrigerante se evapore. Este proceso es endotérmico, lo que significa que requiere calor.

### 5.2. Termodinámica del Cambio de Fase en Absorción

Desde una perspectiva termodinámica, el ciclo de absorción es más complejo que el ciclo de compresión debido a la interacción entre el refrigerante y el absorbente.

- **Entalpía de Absorción:** Es la cantidad de calor liberado o absorbido durante el proceso de absorción a presión constante. Es una propiedad importante para determinar la cantidad de calor que debe ser eliminado en el absorbedor.

- **Entalpía de Desorción:** Es la cantidad de calor requerido para separar el refrigerante del absorbente a presión constante en el generador.

Estas entalpías son específicas para cada par refrigerante-absorbente y son esenciales para el diseño y operación de sistemas de refrigeración por absorción.

## 6. Ciclos de refrigeración reales vs ideales

Mientras que el ciclo de Rankine describe un modelo idealizado de un ciclo de refrigeración, en la práctica, existen varias diferencias entre un ciclo real y uno ideal. Estas diferencias se deben principalmente a las pérdidas inherentes en los componentes reales y a las no idealidades en el comportamiento del refrigerante.

Algunas de las principales diferencias son:

- **Pérdidas en el compresor:** En un compresor real, no toda la energía suministrada se utiliza para comprimir el refrigerante. Parte de esta energía se pierde debido a la fricción y otras ineficiencias.
- **Subenfriamiento y sobrecalentamiento:** En ciclos reales, el refrigerante puede no condensarse o evaporarse completamente en el condensador o evaporador, respectivamente. Esto puede llevar a subenfriamiento o sobrecalentamiento, lo que afecta la eficiencia del ciclo.
- **Pérdidas de presión:** En los componentes reales, especialmente en los intercambiadores de calor y las tuberías, existen pérdidas de presión que no se consideran en el modelo idealizado.

A pesar de estas diferencias, el ciclo de Rankine proporciona una base útil para el análisis y diseño de sistemas de refrigeración. Sin embargo, es esencial tener en cuenta estas no idealidades al trabajar con sistemas reales.

## 7. Compresor

El compresor es el corazón del ciclo de refrigeración y tiene la función principal de comprimir el refrigerante, lo que aumenta su temperatura y presión. Al hacerlo, el refrigerante se convierte en un gas a alta presión y temperatura. Esta compresión es esencial para mover el refrigerante a través del sistema y permitir la transferencia de calor.

![image](https://www.mundohvacr.com/wp-content/uploads/2010/02/83-refrigeracion3.jpg)

### 7.1 Tipos de Compresores

Existen varios tipos de compresores utilizados en sistemas de refrigeración:

1. **Compresores de Pistón:** Son los más comunes y utilizan un pistón que se mueve hacia adelante y hacia atrás dentro de un cilindro para comprimir el refrigerante.

![image](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTUiAFSDq1JgyyvajCPGnkGmj4lSpyo51KmePSV3ve12DD4YbxGDwxEJdAECEUsVsDX7ZI&usqp=CAU)

2. **Compresores de Tornillo:** Utilizan dos tornillos helicoidales que, al girar, comprimen el refrigerante entre ellos.

![image](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQHTEg78t64UUHYeqGfbJNG1psr6NsA-IMfothtLIk5cDdYl4gdP5ndBwE-qO7gNBWK2TQ&usqp=CAU)

3. **Compresores Centrífugos:** Utilizan un rotor que gira a alta velocidad para comprimir el refrigerante mediante la acción centrífuga.

![image](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTxnN-Zmcxxqccqd10OWy1qRqbQ4FH2CAE-yjlfhmyHWzudg3mvvqaIQeazbXwtZAQBDt8&usqp=CAU)

4. **Compresores Scroll:** Utilizan dos espirales interconectadas que comprimen el refrigerante al moverse una contra la otra.

![image](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcT04bJl4TWISL28rw89zS20vU5BqdxR507tNloyA2i6zreIIXtHiXc0hxF6u1D0VdfOAiw&usqp=CAU)

![image](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSV9R5xMEPVcHOpSrBI9gz_tyM129zpy2dTxBiwXp5z23sDcGMK4-EkYX0B67hQ9IOiRiY&usqp=CAU)

Cada tipo de compresor tiene sus propias ventajas y desventajas, y la elección de uno sobre otro depende de la aplicación específica y de las necesidades del sistema de refrigeración.

Desde una perspectiva termodinámica, el compresor realiza trabajo sobre el refrigerante para aumentar su energía interna y, por lo tanto, su temperatura y presión. La relación entre la presión y la temperatura durante la compresión es descrita por la ecuación de estado del refrigerante específico utilizado.

![image](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRu3waw7iPotzjeF4VPzIyt_VaUJIg4tk9VaQ&usqp=CAU)

La eficiencia del compresor se puede describir mediante la relación de compresión, que es la relación entre la presión de descarga y la presión de succión. Matemáticamente, se expresa como:

$$ \text{Relación de Compresión} = \frac{P_{descarga}}{P_{succión}} $$

Donde:

- $P_{descarga}$ es la presión del refrigerante a la salida del compresor.
- $P_{succión}$ es la presión del refrigerante a la entrada del compresor.

Es esencial que el compresor opere dentro de un rango de presiones adecuado para evitar daños y garantizar una operación eficiente del sistema de refrigeración.

## 8. Condensador

El condensador es un componente esencial en el ciclo de refrigeración, donde el refrigerante libera calor al ambiente y cambia de estado gaseoso a líquido. Durante este proceso, el refrigerante pasa por un cambio de fase, lo que implica una transferencia de calor latente.

### 8.1. Balance de Masa y Energía

Para un condensador operando en estado estacionario, el balance de masa se puede expresar como:

$$ \dot{m}_{entrada} = \dot{m}_{salida} $$

Donde $\dot{m}$ es la tasa de flujo másico del refrigerante.

El balance de energía, considerando la transferencia de calor y el trabajo realizado, se puede expresar como:

$$ Q_{condensador} = \dot{m} (h_{entrada} - h_{salida}) $$

Donde:

- $Q_{condensador}$ es la tasa de transferencia de calor en el condensador.
- $h_{entrada}$ y $h_{salida}$ son las entalpías del refrigerante a la entrada y salida del condensador, respectivamente.

### 8.2. Transferencia de Calor y Cambio de Fase

La transferencia de calor en el condensador ocurre principalmente debido al cambio de fase del refrigerante de gas a líquido. Durante este proceso, el refrigerante libera calor latente de condensación. La eficiencia de la transferencia de calor se ve afectada por varios factores, incluidos los números adimensionales de transferencia de calor.

El **número de Nusselt (Nu)** es un número adimensional que relaciona la transferencia de calor convectivo en una superficie con la transferencia de calor conductiva a través del fluido. Se define como:

$$ Nu = \frac{hL}{k} $$

Donde:

- $h$ es el coeficiente de transferencia de calor convectivo.
- $L$ es la longitud característica.
- $k$ es la conductividad térmica del fluido.

El **número de Prandtl (Pr)** es otro número adimensional que relaciona la viscosidad del fluido con su capacidad para conducir calor. Se define como:

$$ Pr = \frac{\mu c_p}{k} $$

Donde:

- $\mu$ es la viscosidad dinámica del fluido.
- $c_p$ es el calor específico a presión constante.

Además, el **número de Reynolds (Re)**, que relaciona las fuerzas inerciales con las fuerzas viscosas en un fluido, también es crucial. Se define como:

$$ Re = \frac{\rho vL}{\mu} $$

Donde:

- $\rho$ es la densidad del fluido.
- $v$ es la velocidad del fluido.
- $L$ es la longitud característica.
- $\mu$ es la viscosidad dinámica del fluido.

Estos números adimensionales, junto con otros parámetros, juegan un papel crucial en la determinación de la eficiencia de la transferencia de calor en el evaporador y en la predicción del tipo de flujo (laminar o turbulento) en el dispositivo.

## 9. Válvula de Expansión

La válvula de expansión es un componente crítico en el ciclo de refrigeración que regula el flujo de refrigerante entre el condensador y el evaporador. Su función principal es reducir la presión del refrigerante, lo que resulta en una disminución en su temperatura. Durante este proceso, el refrigerante no realiza trabajo ni intercambia calor con su entorno.

### 9.1. Balance de Masa y Energía

Para una válvula de expansión en estado estacionario, el balance de masa se puede expresar como:

$$ \dot{m}_{entrada} = \dot{m}_{salida} $$

Dado que no hay trabajo ni transferencia de calor en la válvula, el balance de energía se reduce a:

$$ h_{entrada} = h_{salida} $$

Donde $h$ es la entalpía del refrigerante. Esto indica que la entalpía se conserva a través de la válvula de expansión.

### 9.2. Transferencia de Calor y Cambio de Fase

Aunque la válvula de expansión en sí no está diseñada para intercambiar calor, el proceso de expansión puede resultar en un cambio parcial de fase del refrigerante, especialmente si el refrigerante entra a la válvula en un estado cercano al punto de saturación. Este cambio de fase es isentálpico, lo que significa que no hay cambio en la entropía del refrigerante.

La relación entre la entalpía y la entropía durante este proceso se puede describir mediante las ecuaciones de estado del refrigerante y las tablas termodinámicas correspondientes.

Es importante destacar que, aunque puede haber un cambio de fase en la válvula de expansión, la mayoría de la evaporación del refrigerante ocurre en el evaporador, que es el siguiente componente en el ciclo de refrigeración.

## 10. Evaporador

El evaporador es otro componente crucial en el ciclo de refrigeración. En el evaporador, el refrigerante absorbe calor del espacio o sustancia que se desea enfriar y cambia de estado líquido a gaseoso. Este proceso implica una transferencia de calor latente.

### 10.1. Balance de Masa y Energía

Para un evaporador operando en estado estacionario, el balance de masa se puede expresar como:

$$ \dot{m}_{entrada} = \dot{m}_{salida} $$

El balance de energía, considerando la transferencia de calor y el trabajo realizado, se puede expresar como:

$$ Q_{evaporador} = \dot{m} (h_{salida} - h_{entrada}) $$

Donde:

- $Q_{evaporador}$ es la tasa de transferencia de calor en el evaporador.
- $h_{entrada}$ y $h_{salida}$ son las entalpías del refrigerante a la entrada y salida del evaporador, respectivamente.

### 10.2. Transferencia de Calor y Cambio de Fase

La transferencia de calor en el evaporador ocurre principalmente debido al cambio de fase del refrigerante de líquido a gas. Durante este proceso, el refrigerante absorbe calor latente de evaporación del espacio o sustancia que se desea enfriar.



## 11. Ciclo de Refrigeración por Absorción

![image](https://upload.wikimedia.org/wikipedia/commons/thumb/f/f9/Absorptive_refrigeration.svg/220px-Absorptive_refrigeration.svg.png)

A diferencia de los sistemas de refrigeración por compresión mecánica, donde un compresor es utilizado para aumentar la presión del refrigerante, los sistemas de refrigeración por absorción utilizan un proceso de absorción y desorción para lograr el mismo objetivo. Estos sistemas son especialmente útiles en aplicaciones donde la energía térmica (como el calor residual o solar) está disponible.

![image](https://upload.wikimedia.org/wikipedia/commons/thumb/4/4a/Absorption_refrigerator_working.svg/400px-Absorption_refrigerator_working.svg.png)


El ciclo de refrigeración por absorción consta de los siguientes componentes principales:

1. **Absorbedor:** Aquí, el refrigerante evaporado es absorbido por una solución, reduciendo la presión y permitiendo que más refrigerante se evapore y enfríe el espacio deseado.
2. **Generador (o Desorbedor):** La solución rica en refrigerante es calentada, lo que provoca que el refrigerante se desorba y se evapore.
3. **Condensador:** Similar al ciclo de compresión, el refrigerante en estado gaseoso se condensa liberando calor.
4. **Válvula de Expansión:** Reduce la presión del refrigerante líquido antes de entrar al evaporador.
5. **Evaporador:** El refrigerante líquido absorbe calor y se evapora, enfriando el espacio deseado.

Adicionalmente, una bomba de solución se utiliza para transportar la solución rica en refrigerante del absorbedor al generador.



## 12. Absorbedor

El absorbedor juega un papel crucial en el ciclo de refrigeración por absorción. En este componente, el refrigerante vaporizado es absorbido por una solución absorbente, creando una solución rica en refrigerante.

### 12.1. Balance de Masa

Para un absorbedor en estado estacionario, el balance de masa para el refrigerante y la solución absorbente se puede expresar como:

$$ \dot{m}_{ref,entrada} + \dot{m}_{sol,entrada} = \dot{m}_{sol, salida} $$

### 12.2. Balance de Energía

El balance de energía en el absorbedor, considerando la transferencia de calor y el trabajo realizado, se puede expresar como:

$$ Q_{absorbedor} = \dot{m}_{sol} (h_{salida} - h_{entrada}) $$

Donde $Q_{absorbedor}$ es la tasa de transferencia de calor en el absorbedor y $h$ es la entalpía de la solución. El balance de energía, considerando la transferencia de calor y el trabajo realizado, se puede expresar como:

$$ Q_{absorbedor} = \dot{m}_{ref,entrada} h_{ref,entrada} + \dot{m}_{sol,entrada} h_{sol,entrada} - \dot{m}_{sol,salida} h_{sol,salida} $$

Donde $Q_{absorbedor}$ es la tasa de transferencia de calor en el absorbedor y $h$ representa las entalpías correspondientes.

El absorbedor juega un papel crucial en el ciclo de refrigeración por absorción. En este componente, el refrigerante vaporizado se combina con la solución absorbente, resultando en una solución diluida rica en refrigerante.

#### Fases de las Sustancias

Dentro del absorbedor, coexisten dos fases:

1. **Fase de vapor:** Compuesta principalmente por el refrigerante en estado gaseoso.
2. **Fase líquida:** Una solución de refrigerante y absorbente.

El proceso de absorción implica la transferencia de masa del refrigerante desde la fase de vapor a la fase líquida.

### 12.3. Transferencia de Masa y Calor

El proceso de absorción es simultáneamente un fenómeno de transferencia de masa y calor. Mientras el refrigerante se desplaza desde la fase de vapor a la fase líquida, se libera calor debido a la naturaleza exotérmica del proceso.

El coeficiente de transferencia de masa, $k_m$, y el coeficiente de transferencia de calor, $h$, son parámetros esenciales para describir la eficiencia del proceso de absorción. Estos coeficientes dependen de varios factores, incluidos el tipo de refrigerante y absorbente, las propiedades termodinámicas y físicas de la solución, y las condiciones operativas del absorbedor.

El **número de Sherwood (Sh)** es un número adimensional que relaciona la transferencia de masa convectiva en una superficie con la difusión molecular a través del fluido. Se define como:

$$ Sh = \frac{k_m L}{D} $$

Donde:

- $L$ es la longitud característica.
- $D$ es el coeficiente de difusión del refrigerante en la solución absorbente.

Estos números adimensionales, junto con otros parámetros, juegan un papel crucial en la determinación de la eficiencia de la transferencia de masa y calor en el absorbedor.

## 11. Generador (Desorbedor)

El generador, también conocido como desorbedor, es donde la solución rica en refrigerante se calienta, causando que el refrigerante se desorba y se evapore.

### 11.1. Balance de Masa

Para un generador en estado estacionario, el balance de masa para el refrigerante y la solución absorbente se puede expresar como:

$$ \dot{m}_{sol,entrada} = \dot{m}_{ref,salida} + \dot{m}_{sol,salida} $$

### 11.2. Balance de Energía

El balance de energía en el generador, considerando la transferencia de calor y el trabajo realizado, se puede expresar como:

$$ Q_{generador} + W_{bomba} = \dot{m}_{sol} (h_{salida} - h_{entrada}) $$

Donde $Q_{generador}$ es la tasa de transferencia de calor en el generador, $W_{bomba}$ es el trabajo realizado por la bomba de solución y $h$ es la entalpía de la solución. El balance de energía, teniendo en cuenta la transferencia de calor y el trabajo realizado, se puede expresar como:

$$ Q_{generador} = \dot{m}_{sol,entrada} h_{sol,entrada} - (\dot{m}_{ref,salida} h_{ref,salida} + \dot{m}_{sol,salida} h_{sol,salida}) $$

Donde $Q_{generador}$ es la tasa de transferencia de calor en el generador.

Al igual que en el absorbedor, el proceso de desorción en el generador implica tanto la transferencia de masa como de calor. El refrigerante se desplaza desde la fase líquida a la fase de vapor, absorbiendo calor en el proceso.

El coeficiente de transferencia de masa, $k_m$, y el coeficiente de transferencia de calor, $h$, son esenciales para describir la eficiencia del proceso de desorción. Estos coeficientes dependen de varios factores, incluidos el tipo de refrigerante y absorbente, las propiedades termodinámicas y físicas de la solución, y las condiciones operativas del generador.

El **número de Sherwood (Sh)**, previamente definido, es igualmente relevante en el generador para describir la eficiencia de la transferencia de masa. Además, el **número de Reynolds (Re)** y el **número de Prandtl (Pr)**, que relacionan las propiedades del fluido con la transferencia de calor y masa, también son cruciales en el generador.

## 12. Válvula de Expansión (en sistemas de absorción)

Al igual que en los sistemas de compresión, la válvula de expansión en un sistema de absorción reduce la presión del refrigerante antes de que entre al evaporador. Sin embargo, debido a las características únicas del ciclo de absorción, la operación y las consideraciones termodinámicas pueden variar.

### 12.1. Balance de Masa y Energía

Para una válvula de expansión en un sistema de absorción operando en estado estacionario, el balance de masa se puede expresar como:

$$ \dot{m}_{entrada} = \dot{m}_{salida} $$

Dado que no hay trabajo ni transferencia de calor en la válvula, el balance de energía se reduce a:

$$ h_{entrada} = h_{salida} $$

Esto indica que la entalpía se conserva a través de la válvula de expansión, similar a los sistemas de compresión.

## 13. Evaporador (en sistemas de absorción)

El evaporador en un sistema de absorción funciona de manera similar al de un sistema de compresión. Es aquí donde el refrigerante absorbe calor del espacio o sustancia que se desea enfriar y cambia de estado líquido a gaseoso.

### 13.1. Balance de Masa y Energía

Para un evaporador en un sistema de absorción operando en estado estacionario, el balance de masa se puede expresar como:

$$ \dot{m}_{entrada} = \dot{m}_{salida} $$

El balance de energía, considerando la transferencia de calor y el trabajo realizado, se puede expresar como:

$$ Q_{evaporador} = \dot{m} (h_{salida} - h_{entrada}) $$

Donde:

- $Q_{evaporador}$ es la tasa de transferencia de calor en el evaporador.
- $h_{entrada}$ y $h_{salida}$ son las entalpías del refrigerante a la entrada y salida del evaporador, respectivamente.

La transferencia de calor en el evaporador ocurre principalmente debido al cambio de fase del refrigerante de líquido a gas. Durante este proceso, el refrigerante absorbe calor latente de evaporación.

El **número de Nusselt (Nu)**, **número de Prandtl (Pr)** y **número de Reynolds (Re)**, previamente definidos, son igualmente relevantes en el evaporador para describir la eficiencia de la transferencia de calor y la naturaleza del flujo dentro del dispositivo.

- El evaporador actúa como **superficie fría** donde el aire cede calor y, si se alcanza el punto de rocío, se produce **condensación de humedad** (deshumidificación).  
- El diseño de sistemas de climatización requiere combinar:
  - **Ciclo de refrigeración del refrigerante**.

### 14.1 Carta psicrométrica

La carta psicrométrica relaciona:

- **Temperatura de bulbo seco** $T_{\text{BS}}$.  
- **Humedad específica o razón de humedad** $w$.  
- **Temperatura de bulbo húmedo** $T_{\text{BH}}$.  
- **Humedad relativa** $\phi$.  
- **Entalpía específica del aire húmedo** $h$.  
- **Volumen específico del aire húmedo** $v$.

Con esta carta se pueden analizar procesos típicos:

- **Enfriamiento y deshumidificación** (coils de expansión directa).  
- **Calentamiento sensible**.  
- **Humidificación** (adiabática / no adiabática).

## 15. Resumen conceptual

- Las máquinas de refrigeración **transportan calor** desde una fuente fría hacia una fuente caliente usando **trabajo o calor** como insumo principal.  
- El **cambio de fase** del refrigerante y sus propiedades termodinámicas ( \(h,s,v\) ) son el núcleo de la operación de los ciclos por **compresión de vapor** y por **absorción**.  
- El **diagrama P–h** es la herramienta estándar para el análisis de estos ciclos.  
- El **COP** es la métrica principal de desempeño, distinta para ciclos de compresión y de absorción.  
- Los **ciclos reales** incluyen pérdidas, subenfriamiento, sobrecalentamiento y caídas de presión, que deben ser considerados en diseño e implementación.  
- En acondicionamiento de aire, la **psicrometría** permite conectar el comportamiento del **aire húmedo** con el ciclo de refrigeración del refrigerante.
