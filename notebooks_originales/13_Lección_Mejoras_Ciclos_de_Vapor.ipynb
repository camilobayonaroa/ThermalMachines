##1. Planta térmica de turbina de vapor

Las plantas para la generación de energía basadas en el vapor de agua constan esencialmente de:

1) **Caldera y supercalentadores**  
   Generan **vapor saturado** y elevan su **temperatura** (supercalentamiento) antes de la turbina.

2) **Turbinas de vapor (múltiples cuerpos y extracciones)**  
   HP–MP–LP con **sangrías** para **regeneración** (precalentamiento del agua de alimentación).

3) **Condensador**  
   Licúa el vapor de escape a **baja presión** (vacío), cerrando el ciclo.

4) **Gobernador / control**  
   Regula **caudal** y **entropía** de entrada a turbina (válvulas, control de presión/temperatura).

5) **Calentadores de agua de alimentación**  
   **Cerrados** (superficie) u **abiertos** (contacto directo); recuperan calor y elevan $T$ del agua.

6) **Bombas**  
   Mueven el líquido y elevan su **presión** (p. ej., bomba de condensado y de alimentación).

7) **Generador y auxiliares**  
   Alternador, sistemas eléctricos, **torre de enfriamiento**, **tratamiento de agua**, etc.








##2. Ciclo Rankine standard

El ciclo utilizado para la transformación de energía térmica de vapores es el llamado ciclo de Rankine. Su desarrollo en los últimos cien años ha sido notable, pues a través de una serie de avances técnicos que han resultado en incrementos sucesivos de la temperatura máxima del ciclo y de la capacidad neta de las turbinas y generadores, se ha pasado de sistemas que operaban inicialmente a bajas presiones y bajas temperaturas hasta los sistemas más modernos de los ciclos supercríticos y combinados de hoy en día. La figura 1 muestra un esquema simplificado en el que se observan los cambios de estado del vapor al pasar sucesivamente por la turbina (3-4), el condensador (4-1), la bomba de líquido (1-2) y el generador de calor o caldera (2-3). El proceso ideal que ocurre en la turbina es el de expansión adiabática, es decir, sin evolución de calor, en una sucesión de estados reversibles. Este se asimila a un proceso a entropía constante en el que el vapor saturado a alta presión (punto 3) se expande reversiblemente en la turbina hasta una presión baja (punto 4). El vapor restante es condensado, extrayendo calor QB a temperatura constante TB en el condensador; el líquido se comprime por medio de una bomba hasta la presión de la caldera, en donde se genera vapor en el proceso 2-3 mediante la adición QA.


**Secuencia de estados** (ideal):
- **Bomba:** $1 \rightarrow 2$ (compresión de líquido subenfriado)  
- **Caldera:** $2 \rightarrow 3$ (aporte de calor hasta vapor saturado/sobrecalentado)  
- **Turbina:** $3 \rightarrow 4$ (expansión adiabática ideal, isentrópica)  
- **Condensador:** $4 \rightarrow 1$ (rechazo de calor a presión casi constante)

La eficiencia del ciclo teórico se aproxima a la del ciclo de Carnot, que opera entre las temperaturas promedio más altas y más bajas del mismo. Como el rendimiento del ciclo de Carnot viene dado por $1 - TB/TA$, donde $TB$ es la temperatura más baja y $TA$ es la temperatura más alta, es claro que se puede obtener un mejor rendimiento al disminuir la temperatura de extracción de calor $TB$ o al aumentar la temperatura de adición de calor $TA$. En las plantas termoeléctricas, esto se logra mediante la **operación del condensador a una presión por debajo de la atmosférica** o mediante **procesos de supercalentamiento del vapor antes de entrar a la turbina** **o recalentando el vapor una vez que ha sido parcialmente expandido en la turbina**. Estas medidas se complementan en el ciclo actual con esquemas de regeneración, cuyo objetivo es disminuir el calor aportado por el combustible. En la práctica, esto se realiza mediante extracciones de vapor en la turbina, con el fin de precalentar el agua que fluye hacia la caldera.


###2.1 Eficiencia térmica del ciclo

El rendimiento termodinámico del ciclo de Rankine, como ya se ha dicho, depende de las temperaturas promedio de adición de calor en la caldera y de eliminación de calor en el condensador. Por esta razón y en virtud de los consumos de combustibles de las plantas de fuerza que operan en el mundo y que generan la mayor parte de la energía que se consume, se ha tratado de mejorar hasta donde ha sido posible el rendimiento del ciclo. El rendimiento termodinámico interno del ciclo se define como la relación entre el trabajo neto realizado y el calor aportado al vapor. Esta relación se puede definir como:

$$
\eta = \frac{w}{q_H}
$$

Donde se ha considerado, en su determinación, el trabajo y el calor por unidad de masa. El trabajo neto es igual al trabajo de la turbina o el motor menos el trabajo realizado por las bombas de líquido y los equipos auxiliares, es decir:

$$
w = w_{turb} - \Sigma w_{bombas \ y \ auxiliares}
$$

El trabajo de las bombas se puede estimar fácilmente, puesto que es igual a:

$$
w_{bomba \ ideal} = \int v dp = \underline{v} \Delta P
$$

Donde $\underline{v}$ es el volumen específico promedio para el intervalo de presiones considerado. Este sería el trabajo ideal de compresión; sin embargo, las bombas son equipos relativamente ineficientes y el trabajo real es entonces mayor que el dado por esta expresión. La eficiencia de la bomba se define como:

$$
\eta_b = \frac{w_{bomba \ ideal}}{w_{bomba \ actual}}
$$

En este caso, cabe anotar que el rendimiento o eficiencia de las bombas es relativamente bajo, siendo del orden del 50 al 70%.

El calor $q_H$ aportado al vapor es el que gana el agua al pasar por la caldera, más posiblemente el que pueda ganar el vapor al pasar por el supercalentador y los recalentadores. En el caso de la caldera, la energía del combustible no se aprovecha en su totalidad, sino que, como se explica más adelante, una parte va al vapor y el resto se pierde en los gases en la chimenea, por la conducción y radiación térmica a los alrededores y otras pérdidas similares. El rendimiento de la caldera se determina mediante la relación:

$$
\eta_{cal} = \frac{q_{H}}{PCI}
$$

Donde $PCI$ es el poder calorífico inferior del combustible en kJ/kg.


La **eficiencia teórica** se aproxima a la de **Carnot** entre $T_B$ (baja) y $T_A$ (alta):
$$
\eta_{\text{Carnot}} = 1 - \frac{T_B}{T_A}
$$
Mejoras: **disminuir $T_B$** (más vacío en condensador) y **aumentar $T_A$** (mayor $P/T$ de caldera, **supercalentamiento** y **recalentamiento**). La **regeneración** reduce el calor suministrado en caldera al **precalentar** el agua de alimentación con **sangrías** de turbina.


##3. Turbina

En el caso de la turbina o turbinas, dado que es el equipo que realiza la mayor transformación de energía del sistema, su rendimiento interno es muy importante, ya que tiene un gran efecto sobre el rendimiento global. Si el intercambio de energía se considera adiabático, es decir, sin transferencia de calor, entonces, el trabajo ideal por unidad de masa se puede expresar como:

$$
w_i = h_3 - h_{4i}
$$

Donde $h_3$ es la eficiencia isentrópica de la turbina. Obsérvese que en el diagrama $T$ vs $s$, la posición del punto 4 queda a la derecha del punto 3; esto explica un aumento de la entropía sin que se haya realizado transferencia de calor.

En las plantas de turbina de vapor actuales no se permite el uso de vapor húmedo, puesto que las gotitas de agua que contienen erosionan rápidamente las aspas de la turbina. Por esto, el vapor saturado pasa por un intercambiador llamado supercalentador con el fin de que cuando ocurra la expansión, el estado final se encuentre supercalentado o con un contenido mínimo de humedad. Este caso se muestra en la siguiente figura, en la que el punto $4i$ sería el estado final del vapor si la expansión en la turbina ocurriera de forma ideal, mientras que el punto 4 representa un estado más cercano a la realidad de la expansión irreversible en la turbina. Este último se encuentra muy cerca del punto de saturación con un contenido de agua líquida muy bajo.

##4. Recalentamiento

Una forma posible de mejorar el rendimiento térmico es aumentar la temperatura media de adición de calor. Esto se logra mediante la expansión del vapor de forma parcial en la turbina de alta presión, para ser regresado a la caldera, donde aumenta su temperatura hasta niveles similares a los que tenía inicialmente, para luego ser expandido nuevamente en una turbina de baja presión. Esto podría realizarse muchas veces con el fin de que el ciclo tuviera el aspecto mostrado en la figura 4; sin embargo, las plantas actuales rara vez tienen más de un recalentamiento. El nivel de temperatura al final del recalentamiento (punto 5) puede ser similar al del punto 3. Una ventaja adicional del recalentamiento es que el vapor tiende a salir de la turbina bastante seco, ayudando a proteger el equipo contra la erosión.


Estrategia para **elevar $T$ promedio de adición de calor** y **reducir humedad** en etapas finales:

1. Expansión parcial en **turbina HP**: $3 \rightarrow 4$  
2. **Recalentamiento** en caldera: $4 \rightarrow 5$
3. Expansión en **turbina MP/LP**: $5 \rightarrow 6$

Suele emplearse **un** recalentamiento; $T_5 \approx T_3$.

```python
# ============================================================
# CICLO RANKINE CON RECALENTAMIENTO — CASO BASE IDEAL
# ============================================================

!pip -q install CoolProp

from CoolProp.CoolProp import PropsSI
import matplotlib.pyplot as plt
import numpy as np

# ------------------------------
# PARÁMETROS DEL CICLO
# ------------------------------

# Condiciones del ciclo
P_cond   = 1.0e4     # [Pa] presión del condensador (0.1 bar)
P_boiler = 15.0e6    # [Pa] presión de la caldera (150 bar)
P_reheat = 3.5e6     # [Pa] presión después de la turbina HP (35 bar)

T3  = 813.0          # [K]  salida de caldera (540°C)
T5  = 813.0          # [K]  salida del recalentador (540°C)

eta_pump = 0.85      # eficiencia de la bomba
eta_t_HP = 0.88      # eficiencia isentrópica turbina HP

m_dot = 100.0        # [kg/s] caudal másico de referencia


# ============================================================
# ESTADOS DEL CICLO (1 → 2 → 3 → 4 → 5 → 6 IDEAL)
# ============================================================

# Estado 1: salida del condensador (líquido saturado a P_cond)
h1 = PropsSI("H", "P", P_cond, "Q", 0, "Water")
s1 = PropsSI("S", "P", P_cond, "Q", 0, "Water")
T1 = PropsSI("T", "P", P_cond, "Q", 0, "Water")

# Bomba (1 → 2) — elevación a presión de caldera
h2s = PropsSI("H", "P", P_boiler, "S", s1, "Water")   # ideal
w_p_ideal = h2s - h1
w_p_real  = w_p_ideal / eta_pump
h2 = h1 + w_p_real
s2 = PropsSI("S", "P", P_boiler, "H", h2, "Water")
T2 = PropsSI("T", "P", P_boiler, "H", h2, "Water")

# Caldera (2 → 3)
h3 = PropsSI("H", "P", P_boiler, "T", T3, "Water")
s3 = PropsSI("S", "P", P_boiler, "T", T3, "Water")
T3 = T3  # ya definido

# Turbina HP (3 → 4) — expansión P_boiler → P_reheat
h4s = PropsSI("H", "P", P_reheat, "S", s3, "Water")
w_t_HP_ideal = h3 - h4s
w_t_HP_real  = eta_t_HP * w_t_HP_ideal
h4 = h3 - w_t_HP_real
s4 = PropsSI("S", "P", P_reheat, "H", h4, "Water")
T4 = PropsSI("T", "P", P_reheat, "H", h4, "Water")

# Recalentador (4 → 5)
h5 = PropsSI("H", "P", P_reheat, "T", T5, "Water")
s5 = PropsSI("S", "P", P_reheat, "T", T5, "Water")
T5 = T5  # ya definido

# Turbina LP — CASO BASE SIN CONDENSACIÓN (isentrópica)
h6s = PropsSI("H", "P", P_cond, "S", s5, "Water")
h6  = h6s
s6  = PropsSI("S", "P", P_cond, "H", h6, "Water")
T6  = PropsSI("T", "P", P_cond, "H", h6, "Water")

# Condensador (6 → 1)
q_cond = h6 - h1


# ============================================================
# BALANCES DE ENERGÍA DEL CICLO COMPLETO
# ============================================================

# Calores
q_cal   = h3 - h2
q_reh   = h5 - h4
q_in    = q_cal + q_reh

# Trabajos específicos
w_t_LP_ideal = h5 - h6s
w_net = w_t_HP_real + w_t_LP_ideal - w_p_real

# Potencias
Wdot_t_HP = m_dot * w_t_HP_real
Wdot_t_LP = m_dot * w_t_LP_ideal
Wdot_pump = m_dot * w_p_real
Wdot_net  = m_dot * w_net
Qdot_in   = m_dot * q_in

# Eficiencia térmica
eta_th = Wdot_net / Qdot_in


# ============================================================
# RESULTADOS EN TEXTO
# ============================================================

print("\n============= CICLO BASE SIN CONDENSACIÓN =============")
print(f"Estado 1: T1 = {T1:.2f} K,  P1 = {P_cond:.3e} Pa,  h1 = {h1:.2f} J/kg,  s1 = {s1:.2f} J/kg/K")
print(f"Estado 2: T2 = {T2:.2f} K,  P2 = {P_boiler:.3e} Pa, h2 = {h2:.2f} J/kg,  s2 = {s2:.2f} J/kg/K")
print(f"Estado 3: T3 = {T3:.2f} K,  P3 = {P_boiler:.3e} Pa, h3 = {h3:.2f} J/kg,  s3 = {s3:.2f} J/kg/K")
print(f"Estado 4: T4 = {T4:.2f} K,  P4 = {P_reheat:.3e} Pa, h4 = {h4:.2f} J/kg,  s4 = {s4:.2f} J/kg/K")
print(f"Estado 5: T5 = {T5:.2f} K,  P5 = {P_reheat:.3e} Pa, h5 = {h5:.2f} J/kg,  s5 = {s5:.2f} J/kg/K")
print(f"Estado 6: T6 = {T6:.2f} K,  P6 = {P_cond:.3e} Pa, h6 = {h6:.2f} J/kg,  s6 = {s6:.2f} J/kg/K")

print("\n----- TRABAJOS ESPECÍFICOS -----")
print(f"w_p,real       = {w_p_real:.2f} J/kg")
print(f"w_t,HP,real    = {w_t_HP_real:.2f} J/kg")
print(f"w_t,LP,ideal   = {w_t_LP_ideal:.2f} J/kg")
print(f"w_net          = {w_net:.2f} J/kg")

print("\n----- POTENCIAS -----")
print(f"W_t_HP         = {Wdot_t_HP/1e6:.3f} MW")
print(f"W_t_LP         = {Wdot_t_LP/1e6:.3f} MW")
print(f"W_pump         = {Wdot_pump/1e6:.3f} MW")
print(f"W_net          = {Wdot_net/1e6:.3f} MW")

print("\n----- EFICIENCIA -----")
print(f"Q_in           = {Qdot_in/1e6:.3f} MW")
print(f"η_th (ideal)   = {eta_th:.4f}")
print("\n========================================================\n")


# ============================================================
# DIAGRAMA T–s DEL CICLO
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
from CoolProp.CoolProp import PropsSI

# --- DOMO COMPLETO DE SATURACIÓN ---
# Rango de temperatura desde punto triple hasta punto crítico
T_min = PropsSI("T_TRIPLE", "Water")            # ~273.16 K
T_crit = PropsSI("T_CRITICAL", "Water")         # ~647.096 K

T_sat = np.linspace(T_min, T_crit, 600)

s_l = np.array([PropsSI("S", "T", T, "Q", 0, "Water") for T in T_sat])
s_v = np.array([PropsSI("S", "T", T, "Q", 1, "Water") for T in T_sat])

# --- ISÓBARAS DEL CICLO ---
T_iso = np.linspace(300, 850, 300)

s_iso_boiler = [PropsSI("S", "P", P_boiler, "T", T, "Water") for T in T_iso]
s_iso_reheat = [PropsSI("S", "P", P_reheat, "T", T, "Water") for T in T_iso]

# --- NUEVA ISÓBARA DE 10 000 Pa (0.01 MPa) ---
P_low = 10_000  # Pa
s_iso_low = [PropsSI("S", "P", P_low, "T", T, "Water") for T in T_iso]

# --- CICLO ---
s_cycle = [s1, s2, s3, s4, s5, s6, s1]
T_cycle = [T1, T2, T3, T4, T5, T6, T1]

plt.figure(figsize=(11, 8))

# DOMO COMPLETO RELLENO
plt.fill_betweenx(T_sat, s_l, s_v, color="lightgray", alpha=0.5, label="Región bifásica")

# Líneas frontera del domo
plt.plot(s_l, T_sat, 'b', label="Líquido saturado (Q=0)")
plt.plot(s_v, T_sat, 'r', label="Vapor saturado (Q=1)")

# Isóbaras del ciclo
plt.plot(s_iso_boiler, T_iso, '--k', alpha=0.7, label="Isóbara caldera (15 MPa)")
plt.plot(s_iso_reheat, T_iso, '--g', alpha=0.7, label="Isóbara recalentamiento (3.5 MPa)")

# Nueva isóbara de baja presión
plt.plot(s_iso_low, T_iso, '--b', alpha=0.7, label="Isóbara 10 kPa")

# Ciclo base
plt.plot(s_cycle, T_cycle, "o-", linewidth=2.5, color='purple', label="Ciclo Rankine base")

# Etiquetas de puntos
for i, (si, Ti) in enumerate(zip(s_cycle[:-1], T_cycle[:-1]), start=1):
    plt.text(si, Ti, f" {i}", fontsize=13, fontweight='bold')

plt.xlabel("s [J/kg·K]")
plt.ylabel("T [K]")
plt.title("Diagrama T–s — Ciclo Rankine con Recalentamiento (Domo Completo de Saturación)")
plt.grid(True)
plt.legend()
plt.show()
```

##5. Regeneración

Otra manera de mejorar el rendimiento del ciclo de Rankine consiste en incrementar la temperatura del agua que se ha condensado y que se encuentra como líquido subenfriado a la presión del condensador. Por ejemplo, si la presión de condensación fuese de 1 pulgada de mercurio absoluto (1/30 atm), la temperatura del líquido saturado sería relativamente baja (aproximadamente 46 °C). Por otra parte, si la presión de la caldera fuese igual a 4 MPa, la temperatura de saturación sería de unos 190 °C.

Esto quiere decir que se podría calentar el líquido antes de entrar a la caldera a través de un diferencial importante de temperatura. Este calentamiento se podría realizar en una turbina especial en la que el vapor que se expande fluyera en contracorriente con el agua que se está calentando. Teóricamente, se podría obtener agua en estado de saturación a la entrada de la caldera.

De este modo, se disminuiría el aporte de calor en esta última al mismo tiempo que la temperatura de adición de calor del ciclo ideal se haría constante e igual a la temperatura de saturación del vapor a alta presión. Como todo el calor extraído en el condensador fluiría a temperatura constante, la eficiencia térmica del ciclo sería, en teoría, igual a la del ciclo de Carnot.

Tal equipo es prácticamente imposible de construir; no obstante, en la actualidad se puede conseguir un efecto parecido mediante el esquema explicado a continuación.


Eleva la **temperatura del agua de alimentación** antes de caldera, reduciendo $q_H$ y acercando el perfil térmico al **Carnot** ideal.  

Implementación práctica mediante **extracciones** (sangrías) y **calentadores**:

- **Abiertos (contacto directo):** el vapor de sangría se mezcla con el agua → **misma presión**.  
- **Cerrados (superficie):** intercambio sin mezcla → **presiones distintas** posibles.

> Múltiples etapas de calentamiento permiten acercar $T_{\text{agua}}$ a la **temperatura de saturación** de la caldera, reduciendo el **salto térmico** en evaporador y mejorando $\eta$.

*Figura 5 :* Diagrama de una planta con regeneración por extración de vapor, con un calentador de tipo abierto o de contacto directo

En la práctica, la regeneración se logra mediante la extracción de una parte del vapor durante la expansión en la turbina; esta se utiliza para precalentar el agua de alimentación, haciéndola circular por un intercambiador de calor (regenerador), como se puede ver en la figura 5. El intercambiador puede ser de superficie, es decir, que no hay contacto entre el vapor usado para calentar el líquido que circula por el otro lado del equipo; en este caso, las presiones del vapor extraído y la del líquido pueden ser diferentes. El intercambio térmico también puede llevarse a cabo mediante un intercambiador de contacto directo o abierto, en el que el vapor de extracción entra en contacto con el líquido, elevando su temperatura.

Este tipo de intercambiadores se utiliza, además, para extraer el aire que pueda contener el agua de alimentación y, por lo general, se colocan en la zona de baja presión de la planta. El proceso de calentamiento del agua se puede repetir una o muchas veces, extrayendo vapor de la turbina y utilizándolo para precalentar el agua de alimentación. De esta forma, el vapor intercambia calor en contraflujo con el agua de alimentación hasta llevarla a una temperatura muy cercana a la de saturación correspondiente a la presión de la caldera.


## Notas de diseño y operación

- **Aumentar $P/T$ de caldera** y **reducir $p_{\text{cond}}$** (vacío) mejora $\eta$, limitado por **materiales** y **termohidráulica**.  
- **Recalentamiento**: disminuye **humedad** en LP y mejora $\eta$ a costa de mayor complejidad.  
- **Regeneración**: disminuye $q_H$ en caldera y mejora el **perfil térmico**.  
- **Tratamiento de agua**: sílice, oxígeno disuelto, pH, **corrosión/ensuciamiento**.  
- **Auxiliares** (bombas/ventiladores): su consumo impacta **$w_{\text{neto}}$**.  
- **Condensador/torre**: control de **vacío**, **fugas de aire**, **ensuciamiento** y **rechazo de calor**.
