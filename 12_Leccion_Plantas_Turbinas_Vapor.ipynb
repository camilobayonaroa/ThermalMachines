##1. Plantas Térmicas de Turbinas de vapor (Rankine)

- **Bomba:** Impulsa el agua a través del sistema.
- **Caldera:** Donde se calienta el agua para convertirla en vapor.
- **Intercambiadores de Calor:** Como los generadores de vapor de recuperación de calor (HRSG) que utilizan el calor de procesos industriales.

##2. Fuentes Energéticas:

###2.1. Carbón


Las centrales eléctricas de carbón, también conocidas como centrales termoeléctricas de carbón, son instalaciones que queman carbón para generar electricidad. Estas centrales desempeñan un papel fundamental en la producción de electricidad a nivel mundial, pero también tienen un impacto significativo en el medio ambiente y la salud pública.

![Central eléctrica de carbón en Tuticorin, India](https://upload.wikimedia.org/wikipedia/commons/e/ee/NTPL_Tuticorin_NLCIL.jpg)

El proceso implica la pulverización del carbón, que luego se quema en una caldera de carbón pulverizado. El calor generado en el horno convierte el agua de la caldera en vapor, que posteriormente se utiliza para mover turbinas que, a su vez, accionan generadores eléctricos.


![Central eléctrica de carbón en Tuticorin, India](https://upload.wikimedia.org/wikipedia/commons/9/9c/Coal_Conveyors_at_Salem_Harbor_Station.jpg)


###2.2. Gas Natural


Las centrales térmicas son instalaciones que utilizan el calor generado por la combustión para producir electricidad. Se dividen principalmente en centrales de carbón y centrales de ciclo combinado.
Las centrales de ciclo combinado, por otro lado, son más eficientes y emiten menos gases de efecto invernadero. Estas centrales transforman la energía térmica del gas natural en electricidad a través de dos mecanismos: una turbina de gas y otra de vapor. Son más eficientes, tienen menores emisiones, consumen menos agua de refrigeración y ahorran energía en forma de combustible. Además, su construcción es más rápida y requieren menos superficie por megavatio (MW) instalado, reduciendo el impacto visual.

El gas natural es una mezcla de gases, principalmente metano, pero también contiene nitrógeno, etano, CO2, H2O, butano, propano, mercaptanos y otros hidrocarburos. El metano suele constituir el 95% del gas natural. Este recurso se formó hace millones de años a partir de depósitos de materia orgánica entre sedimentos en un ambiente con bajo contenido de oxígeno. Con el tiempo, el gas ascendió entre las capas de terreno y se acumuló en yacimientos o reservas.

![Planta Solar PS10](https://upload.wikimedia.org/wikipedia/commons/1/14/Metz_biomass_power_station.jpg)



###2.3. Biomasa

La biomasa se refiere a un tipo de energía que las plantas transforman de la energía radiante del Sol en energía química a través de la fotosíntesis. Parte de esa energía química queda almacenada en forma de materia orgánica. Esta energía química puede recuperarse quemando la biomasa directamente o transformándola en combustible.

![Biomasa](https://upload.wikimedia.org/wikipedia/commons/7/76/Biochar.jpg)

Los tipos de Biomasa son:

1. Biomasa Natural: Producida en la naturaleza sin intervención humana, como la caída natural de ramas, hojas y otros materiales de los árboles o animales en los bosques.
2. Biomasa Residual: Es el subproducto o residuo generado en actividades agrícolas, silvícolas, ganaderas, y en la industria agroalimentaria y de transformación de la madera.
3. Cultivos Energéticos: Cultivos destinados a la producción de biocombustibles. Estos pueden ser cereales y remolacha para producción de bioetanol y oleaginosas para producción de biodiésel, entre otros.

Los procesos de obtención son:

1. Uso Directo: La biomasa se emplea directamente, como en el caso de la madera o la paja.
2. Fermentación Alcohólica: Proceso utilizado para producir bebidas alcohólicas, donde una mezcla de azúcares y agua se transforma en una mezcla de alcohol y agua.
3. Transformación de Ácidos Grasos: Aceites vegetales y grasas animales se transforman en una mezcla de hidrocarburos similar al diésel.
4. Descomposición Anaeróbica: Proceso que permite obtener metano en forma de Biogás a partir de residuos orgánicos, principalmente excrementos animales.
5. Procesos de secado de materiales vegetales.
6. Procesos de pirólisis de materiales vegetales.

La biomasa puede considerarse neutra en términos de emisiones netas si solo se emplea en cantidades iguales a la producción neta de biomasa del ecosistema que se explota.


###2.4. Energía Solar Térmica

La central se basa en un diseño de torre central, que es esencial para su funcionamiento. La torre central mide 115 metros de altura y está rodeada por 624 heliostatos, espejos que siguen el movimiento del sol, cubriendo una superficie de 60 hectáreas. Cada heliostato tiene una superficie de 120 m².

El principio de funcionamiento es sencillo pero efectivo: los heliostatos reflejan la luz solar hacia un receptor situado en la parte superior de la torre central. Este reflejo de alta intensidad calienta el receptor. El calor generado produce vapor a 257ºC a partir del agua a presión que circula por tuberías. Este vapor se utiliza para mover una turbina que genera electricidad. Además, parte del vapor se almacena en tanques acumuladores para su uso cuando la producción solar es insuficiente.

La PS10 convierte aproximadamente el 17% de la energía solar reflejada en electricidad, produciendo 11 megavatios. Esto es suficiente para ahorrar la emisión de 6.700 toneladas de dióxido de carbono (CO₂) al año, según datos de Abengoa. La Planta Solúcar, que incluye PS10, PS20 y Solnovas, produce en conjunto 183 megavatios anuales, lo que equivale a la energía necesaria para 94.000 hogares y evita la emisión de unas 114.000 toneladas de CO₂.

###2.5. Energía Geotérmica

La energía geotérmica es una forma de energía renovable que se obtiene aprovechando el calor del interior de la Tierra. Este calor se transmite a través de los cuerpos de roca o piedras calientes mediante procesos de conducción y convección. En estos procesos, ocurre una interacción entre el agua subterránea y las rocas, dando origen a los sistemas geotérmicos. El término "geotérmico" proviene del griego *geo* («Tierra») y *thermos* («calor»), lo que se traduce como "calor de la Tierra".

El interior de nuestro planeta está caliente, y esta temperatura aumenta con la profundidad. En las capas profundas, las altas temperaturas a menudo calientan las capas freáticas, y cuando este agua caliente o vapor asciende, produce manifestaciones en la superficie, como géiseres o fuentes termales, que han sido utilizadas para baños desde tiempos antiguos.

![Imagen Energía Geotérmica](https://upload.wikimedia.org/wikipedia/commons/9/9f/NesjavellirPowerPlant_edit2.jpg)


![Imagen Energía Geotérmica](https://upload.wikimedia.org/wikipedia/commons/c/c6/EGS_diagram.svg)

Los tipos de yacimientos geotérmicos son:
1. De agua caliente: Estos yacimientos pueden ser fuentes o subterráneos contenidos en un acuífero. Los que forman fuentes se han aprovechado desde tiempos antiguos como baños termales. Los subterráneos, con aguas termales muy calientes a poca o media profundidad, se utilizan para aprovechar el calor del interior de la tierra.
2. Secos: En estos yacimientos, hay una zona bajo tierra con piedras calientes en seco. Se inyecta agua por una perforación y se recupera ya caliente por otra, aprovechando el calor mediante un intercambiador.
3. Géiseres: Son fuentes termales que expulsan periódicamente una columna de agua caliente y vapor al aire.
4. Vapor seco: Yacimientos donde el agua sale en forma de vapor, aprovechado directamente para obtener energía mecánica mediante una turbina.

La clasificación según la temperatura del agua:
1. Alta temperatura (150-400°C): Se produce vapor en la superficie a altas temperaturas, capaces de ser aprovechadas directamente con una turbina.
2. Temperaturas medias (70-150°C): La conversión vapor-electricidad se realiza con un rendimiento menor, y se explota mediante un fluido volátil.
3. Baja temperatura (50-70°C): Aprovechable en zonas residenciales, debido al gradiente geotérmico.
4. Muy baja temperatura (20-50°C): Se utiliza para necesidades domésticas, urbanas o agrícolas, como la climatización geotérmica.

Algunos de los métodos de generación y aprovechamiento de la Energía Geotérmica:
1. Vapor seco: Se filtra agua de menor temperatura a la corteza terrestre, se calienta y emana vapor a la superficie, que se utiliza para mover las turbinas de los generadores eléctricos.
2. Vapor flash: Se extrae agua a altas temperaturas hacia la superficie y se convierte inmediatamente en vapor para impulsar las turbinas.
3. Ciclo binario: El agua pasa por un intercambiador de calor y se combina con isobutano, que se convierte en vapor para mover las hélices de las turbinas.

Ventajas:
- Disminuye la dependencia energética de los combustibles fósiles.
- Eliminación del riesgo de transmisión de toxinas.
- Obtención de calefacción, refrigeración y agua potable con un mismo sistema.
- Poca afectación del paisaje.
- Recurso disponible todo el tiempo, independientemente de las condiciones meteorológicas.
- Produce mínimos residuos con menor impacto ambiental.
- Es una fuente de energía prácticamente inagotable a escala humana.
- No está sujeta a precios internacionales.
- Emite menos CO2 en comparación con la combustión.

Desventajas:
- En yacimientos secos, pueden producirse microsismos.
- Emisión de gases como el ácido sulfhídrico.
- Contaminación de aguas próximas con sustancias como arsénico.
- Deterioro del paisaje.
- No está disponible en todos los lugares.

La energía geotérmica ha sido utilizada para generar electricidad desde 1904, cuando se produjo por primera vez en Larderello, Italia. Actualmente, su uso ha crecido a nivel mundial, siendo países como Estados Unidos, Indonesia, Filipinas, Turquía y Nueva Zelanda los principales productores.

###3.1. Funcionamiento General de la Central:
- El combustible se quema en la caldera, liberando calor que se utiliza para calentar agua.
- El agua calentada se transforma en vapor con una alta presión, que hace girar una turbina de vapor, convirtiendo la energía interna del vapor en energía mecánica.
- La electricidad se genera en el alternador mediante la rotación del rotor (que comparte el mismo eje que la turbina de vapor) y mediante la inducción electromagnética.
- El vapor que sale de la turbina se envía a un condensador para transformarlo en líquido y retornarlo a la caldera para comenzar un nuevo ciclo de producción de vapor.

###3.2. Componente: Bomba

![texto del enlace](https://pumpsolutions.com.au/wp-content/uploads/2016/05/power-station-pumps.jpg)


En una central termoeléctrica, la bomba juega un papel crucial en el proceso de generación de energía. Su función principal es impulsar el agua a través del sistema. A continuación, se detalla la función y el papel de la bomba en una central termoeléctrica:

1. Impulsar el Agua: La bomba hidráulica impulsa el agua a través del sistema. Esta agua es esencial para generar vapor que, a su vez, accionará las turbinas y producirá electricidad.
   
2. Circulación del Agua: Una vez que el vapor ha pasado por las turbinas, se condensa y se convierte nuevamente en agua. Esta agua necesita ser recirculada de nuevo al sistema para ser calentada y convertida en vapor. La bomba asegura que este ciclo de agua continúe sin interrupciones.

3. Mantenimiento de la Presión: La bomba también ayuda a mantener la presión adecuada en el sistema. Una presión constante es esencial para garantizar que el agua se convierta en vapor a la temperatura deseada y que el vapor tenga la energía suficiente para mover las turbinas.

###3.3.Componente: Caldera

Una caldera de combustión es uno de los equipos esenciales en las centrales térmicas. Su principal función es extraer la energía calorífica del combustible y transferirla al fluido que circula por su interior, que puede ser agua o vapor.

![Imagen del Condensador](https://upload.wikimedia.org/wikipedia/commons/5/59/Turmkessel02.png)

Los componentes de una caldera típica son:
1. Hogar o Cámara de Combustión: Es el lugar donde se quema el combustible.
2. Intercambiadores de calor: Están formados por haces de tubos. Los gases a alta temperatura circulan por el exterior de estos tubos, transfiriendo calor al fluido que circula por su interior. Este fluido puede ser agua, que se calienta o vaporiza, o vapor, que se sobrecalienta o recalentiza. Los principales intercambiadores de calor son los economizadores, las pantallas vaporizadoras o paredes de agua, los sobrecalentadores y los recalentadores del vapor.
3. Ceniceros: Son la parte inferior de la caldera donde se recogen las escorias. Estas escorias luego son transportadas a un silo de recogida y almacenamiento.
4. Chimeneas: Permiten una efectiva descarga de gases hacia el ambiente exterior. Usualmente permiten forzar el flujo de gases desde el hogar hasta el exterior.

Los combustibles se introducen en el hogar para alimentar directamente la llama. Los gaseosos se introducen con facilidad, los líquidos se pulverizan en forma de gotitas, y los sólidos se muelen a tamaños finos e inyectan en la caldera arrastrados por una corriente de aire.

Además de generar vapor, en las calderas actuales, se sobrecalienta el vapor hasta una determinada temperatura. En centrales con ciclo recalentado, se recalienta el vapor que regresa a la caldera después de haberse expandido en las etapas de alta presión de la turbina.

###3.4. Componente: Turbina

Las turbinas son máquinas que transforman la energía contenida en un fluido en movimiento (en este caso, vapor) en energía mecánica mediante la rotación de un eje. En las centrales termoeléctricas, las turbinas de vapor juegan un papel crucial en la conversión de la energía térmica en energía eléctrica.

![Imagen del Condensador](https://upload.wikimedia.org/wikipedia/commons/7/79/Dampfturbine_Montage01.jpg)

Al expandirse el vapor dentro de la turbina, este actúa sobre las palas de la turbina, provocando su movimiento rotativo. Esta rotación se transforma en energía mecánica, que se utiliza para mover un eje conectado a un alternador. El alternador, mediante la inducción electromagnética, convierte esta energía mecánica en energía eléctrica. Una vez que el vapor ha pasado por la turbina, se dirige a un condensador donde se enfría y se transforma nuevamente en agua líquida, para ser bombeada de nuevo a la caldera y comenzar el ciclo nuevamente.

Los tipos de turbinas en centrales termoeléctricas son usualmente llamadas:
1. Turbina de Alta Presión: Recibe el vapor directamente desde la caldera a muy alta presión y temperatura.
2. Turbina de Media Presión: Actúa después de la turbina de alta presión, recibiendo el vapor a una presión y temperatura intermedias.
3. Turbina de Baja Presión: Es la última etapa, donde el vapor tiene la menor presión antes de entrar al condensador.

###3.5. Componente: Condensador

![texto del enlace](https://upload.wikimedia.org/wikipedia/commons/f/fe/Surface_condenser_Anadrian_MMM_n01.jpg)



**Funcionamiento del Condensador:**
- La función principal del condensador en una central térmica es actuar como el foco frío o sumidero de calor dentro del ciclo termodinámico del grupo térmico.
- Su misión es condensar el vapor que proviene del escape de la turbina de vapor en condiciones próximas a la saturación y evacuar el calor de condensación (calor latente) al exterior mediante un fluido de intercambio, que puede ser aire o agua.

- Las condiciones en el interior del condensador son de saturación, es decir, está a la presión de saturación correspondiente a la temperatura de condensación del vapor. Esta presión es siempre inferior a la atmosférica, lo que se traduce en un vacío en el condensador.

**Tipos de Condensadores en Centrales Térmicas:**
1. **Axiales:** Están situados al mismo nivel que la turbina de vapor. Son típicos de turbinas de vapor hasta 150 MW.
2. **Laterales:** Están situados al mismo nivel que la turbina de vapor. El cuerpo de baja presión de la turbina de vapor es de dos flujos.
3. **Inferiores:** Están situados debajo de la turbina de vapor de baja presión, lo que les obliga a estar metidos en un foso. Dadas las potencias de las centrales convencionales actuales, este es el tipo de condensador más usualmente empleado.

![Imagen del Condensador](https://upload.wikimedia.org/wikipedia/commons/8/8b/Surface_Condenser.png)



**Disposición Constructiva de un Condensador en Centrales Térmicas:**
- Los condensadores que emplean aire como fluido refrigerante, llamados aerocondensadores, tienen un bajo rendimiento y, por tanto, necesitan de grandes superficies para ser instalados.
- Los condensadores de las centrales térmicas son cambiadores de calor tubulares, de superficie, del tipo carcasa y tubo.
- Las partes más significativas de un condensador son: Cuello, Carcasa o cuerpo, Cajas de agua, Tubos, Placas de tubos, Placas soporte, Pozo caliente, Zona de enfriamiento de aire, Sistema de extracción de aire.

###3.6. Torres de Enfriamiento

![Imagen del Condensador](https://upload.wikimedia.org/wikipedia/commons/9/95/Evaporative_Cooling_Tower.jpg)


Las torres de refrigeración o torres de enfriamiento son estructuras diseñadas para disminuir la temperatura del agua y otros medios. Su uso principal es rebajar la temperatura del agua de refrigeración utilizada en diversas instalaciones industriales, como plantas de energía, refinerías de petróleo, plantas petroquímicas y plantas de procesamiento de gas natural.


**Tipos de Torres de Enfriamiento:**

1. **Torres de refrigeración húmedas:** Funcionan por el principio de evaporación. En estas torres, el agua caliente puede ser enfriada a una temperatura inferior a la del ambiente si el aire es relativamente seco.

![Imagen de torre de refrigeración](https://upload.wikimedia.org/wikipedia/commons/3/3a/Power_station_Westfalen._Cooling_towers.jpg)

![Imagen de torre de refrigeración](https://www.powermag.com/wp-content/uploads/2008/09/520004dd8f419-56-12.jpg)

2. **Torres de refrigeración secas:** Funcionan transmitiendo el calor a través de una superficie que separa el fluido a refrigerar del aire ambiente.

3. **Tiro de aire en la torre:**
   - **Tiro natural:** Utiliza una chimenea alta.
   - **Tiro inducido:** El ventilador se coloca en la parte superior de la torre.
   - **Tiro mecánico (o tiro forzado):** Utiliza la potencia de motores de ventilación para impulsar el aire a la torre.

![Imagen de torre de refrigeración](https://upload.wikimedia.org/wikipedia/commons/e/e7/Didcot_power_station_cooling_tower_zootalures.jpg)

Las torres de enfriamiento usan la evaporación del agua para rechazar el calor de un proceso, como la generación de energía eléctrica. Pueden variar en tamaño, desde estructuras pequeñas hasta muy grandes que pueden sobrepasar los 220 metros de altura y 100 metros de longitud.



##4. Representaciones gráficas
- **T–s:** área bajo expansión ≈ **trabajo de turbina**; recta vertical isentrópica ideal; efecto de **recalentamiento** (bucle) y **regeneración** (desplaza 2→2′).  
- **p–h:** visualiza **caídas de presión** y **aportes/rechazos** de calor.



##5. Mejoras de eficiencia
- **Recalentamiento (Reheat):** 3→4 (HP), reheat a 3′, 3′→4′ (MP/LP). Reduce **humedad** y eleva **η**.  
- **Regeneración:** **calentadores de agua de alimentación** (abiertos/cerrados) usando **sangrías** de la turbina → reduce \( q_{in} \).  
- **Mayor vacío** en condensador (menor \( p_{cond} \)) y **mayor T/P** de caldera (materiales y límites metalúrgicos).  
- **Ciclo combinado:** TG (**Brayton**) + HRSG + TV (**Rankine**) → **η global** > 55–60% (según configuración).





##6. Apéndice: balances y notas prácticas

- **Bomba/Condensado:** verificar **NPSH**, evitar cavitación, **sellos mecánicos** y **consumo auxiliar**.  
- **Caldera:** control de **exceso de aire**, **fouling/slagging**, **tratamiento químico** del agua (O₂, sílice, pH).  
- **Turbina:** límites de **humedad** en últimas etapas, **sellos**, **alineación** y **vibración**.  
- **Condensador/Torre:** **ensuciamiento** de tubos, **aire disuelto**, **purga** y **tratamiento del agua**.  
- **Medio ambiente:** **NOx/SOx/PM**, **CO₂**, control de **plumas** en torres, **ruido** y **uso de agua**.
