## 3. Potencia al freno $(P_{\text{freno}})$ o al eje
El primer valor que se requiere calcular es la potencia al freno, es decir la **potencia que aplica el freno** para **mantener la velocidad rotacional de motor constante**, esta magnitud viene relacionado al Par T por la velocidad rotacional.

Esta potencia se determina multiplicando el par $T$ por la velocidad angular $\omega$ en radianes por segundo:

$
P_{\text{freno}} = T \cdot \omega = T \cdot \frac{2\pi N}{60}
$





## 4. Flujo de combustible y CEC al freno

El flujo de combustibles se calcula mediante (en kg/s):

$$
\dot{m}_c = \frac{M}{t}
$$



### Consumo específico de combustible al freno

El **consumo específico de combustible al freno** $cec$ es una medida indirecta del rendimiento del motor y se determina haciendo la razón entre el flujo de combustible y la potencia al freno:

$$
cec = \frac{\dot{m}_c}{P_{\text{freno}}}
$$




## 5. Eficiencia térmica al freno

Para determinar la eficiencia térmica al freno, se utiliza el consumo específico multiplicado por el poder calorífico inferior (PCI):

$$
\eta_{\text{freno}} = (cec \cdot \text{PCI})^{-1}
$$




## 6. Flujo teórico de aire $(\dot{V}_{\text{teo}})$

El flujo teórico de aire se determina mediante el producto del desplazamiento volumétrico multiplicado por el número de ciclos de aspiración por segundo.

Para motores de **dos tiempos**:

$$
\dot{V}_{\text{teo}} = \frac{N_c}{4} \cdot \frac{\pi d^2}{4} \cdot L \cdot \frac{N}{60}
$$

Para motores de **cuatro tiempos**:

$$
\dot{V}_{\text{teo}} = \frac{N_c}{2} \cdot \frac{\pi d^2}{4} \cdot L \cdot \frac{N}{120}
$$

donde \(N_c\) es el **número de cilindros**.

> **Rendimiento volumétrico $(\eta_v)$**: el flujo **real** de aire suele ser $\dot{V}_{\text{real}}\approx \eta_v\,\dot{V}_{\text{teo}}$, con $\eta_v$ entre **0.75–1.10** según régimen, sobrealimentación y diseño de admisión.


````python
import math
# definimos variables
N = 3000  # rpm
M = 0.5  # kg
t = 120  # s
T = 150  # N·m
PCI = 45000  # kJ/kg
d = 0.1  # m
L = 0.15  # m
# Calculamos los parametros anteriormente mencionados:
P_freno = T * N * 2 * math.pi / 60 #W
m_c = M/t #kg/s
cec=m_c/P_freno #kg/Ws
eff_freno = (cec*PCI*1000)**(-1)
V_teo = 1*math.pi*(d**2)/4 * L * N/120
# Finally, we print the result:
print("La potencia de freno es:", round(P_freno,2), "kW")
print("El flujo de combustible es:", round(m_c*1000,2), "x10^-3 kg/s")
print("El consumo específico del combustible al freno es:", round(cec*10**8,2), "x10^-8 kg/W")
print("La eficiencia termica al freno es:", round(eff_freno*100,2),"%")
print("El flujo teorico de aire es:", round(V_teo,2), "m3/s")
````

## 7. Presión Media Efectiva (PME)
La presion media efectiva de un motor la presión promedio que actúa sobre el pistón durante el ciclo de trabajo del motor. Representa la presión que, si se aplicara de manera uniforme sobre toda la superficie del pistón, produciría la misma cantidad de trabajo que el motor produce en condiciones reales de funcionamiento.

La presión media efectiva es una herramienta que permite comparar el rendimiento de diferentes motores y para evaluar mejoras en la eficiencia y la potencia. Se calcula mediante la integración de las presiones instantáneas durante el ciclo de trabajo y se expresa generalmente en unidades de presión, como bares o psi. Una presión media efectiva más alta indica un motor más eficiente y potente, mientras que una presión más baja puede indicar pérdidas en el rendimiento del motor.



````python
import math
import matplotlib.pyplot as plt

def engine_kinematics(Bore, stroke, con_rod, cr, start_crank, end_crank):

	#Geometric parameters
	a = stroke/2		# Crank pin radius
	R = con_rod/a 		# constant

	#Volume computation
	V_s = (math.pi/4)*pow(Bore,2)*stroke		# swept volume
	V_c = V_s/(cr-1)	# clearance volume

	sc = math.radians(start_crank)
	ec = math.radians(end_crank)

	num_values = 50

	dtheta = (ec-sc)/(num_values-1)
	V = []

	for i in range(0,num_values):

		theta = sc + i*dtheta

		term1 = 0.5*(cr-1)
		term2 = R+1-math.cos(theta)
		term3 = pow(R,2)-pow(math.sin(theta),2)
		term3 = pow(term3,0.5)

		V.append ((1+term1*(term2-term3))*V_c)
	return V


# inputing the initial parameters
P1 = 0.1	 # pressure at state-1
t1 = 500	 # temperature at state-1
gamma = 1.4
t3 = 2300

#geometric parameters
Bore = 0.10
stroke = 0.1
con_rod = 0.15
cr = 12

#Volume computation
V_s = (math.pi/4)*pow(Bore,2)*stroke		# swept volume
V_c = V_s/(cr-1)	# clearance volume
V1 = V_c + V_s

# at state point-2
V2 = V_c

# P2V2^gamma = P1V1^gamma
P2 = P1*pow(V1,gamma)/pow(V2,gamma)

# P2V2/t2 = P1V1/t1 | rhs = P1V1/t1 | P2V2/t2 = rhs | t2 = P2V2/rhs

rhs = P1*V1/t1
t2 = P2*V2/rhs
V_compression = engine_kinematics(Bore,stroke,con_rod,cr,180,0)
constant = P1*pow(V1,gamma)
P_compression = []

for V in V_compression:
	P_compression.append(constant/pow(V,gamma))

# at state point 3
V3 = V2
# P3V3/t3 = P2V2/t2 | rhs = P2V2/t2 | P2 = rhs*t3/V3
rhs = P2*V2/t2
P3 = rhs*t3/V3

V_expansion = engine_kinematics(Bore,stroke,con_rod,cr,0,180)
constant = P3*pow(V3,gamma)
P_expansion = []

for V in V_expansion:
	P_expansion.append(constant/pow(V,gamma))

# at state point 4
V4 = V1
# P4V4^gamma=P3V3^gamma
P4 = P3*pow(V3,gamma)/pow(V4,gamma)
# P4V4/t4 = rhs
t4 = P4*V4/rhs

#Thermal efficiency of the Otto cycle
#thermal efficiency is referred as Ther_eff
Ther_eff = (1 - ((t4-t1)/(t3-t2)))*100
print(Ther_eff)

#plotting the P_V diagram from otto cycle
plt.grid()
plt.plot([V2,V3],[P2,P3])
plt.plot(V_compression,P_compression)
plt.plot(V_expansion,P_expansion)
plt.plot([V4,V1],[P4,P1])
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.title('Otto cycle')
plt.show()

````

El valor de la presion media efectiva si viene dado por el diagrama PV se llama presion *media efectiva especifica indicada*, la cual esta definica como el  trabajo por unidad de masa y la diferencia de volumenes y hace referencia a las lecturas realizadas dentro del cilindro para varias posiciones angulares del cigueñal del motor.


$P_{mei} = (Areadeldiagrama)/L$

Donde L es la representacion arbitraria  del volumen.

Para este caso podemos calcular la presion media efectiva indicada:

````python
import numpy as np
# calculamos las areas de compresion y de expansion adiabaticas
area_comp = np.trapz(P_compression, V_compression)
area_exp = np.trapz(P_expansion, V_expansion)
# calculamos el trabajo generado mostrado en el diagrama PV
W_neto = area_exp - area_comp
# Estimamos la Pmei
Pmei = W_neto/(max(V_compression)-min(V_compression))
print("El trabajo neto es: ",W_neto)
print("La presion media efectiva indicada es: ",Pmei)
````

Si la presion es medida sobre la potencia en el eje del motor, o la potencia al freno, se denomina presión media efectiva al freno ($P_{mef}$) la cual esta relacionada con al eficiencia del motor de la siguiente manera:
$P_{mef} = P_{mei}*\eta_{th,otto}$

donde $\eta_m$ es la **eficiencia mecánica** (efecto de fricción y auxiliares).  
Si se usa la eficiencia térmica del ciclo ideal (Otto/Diesel) como aproximación de conversión termo-mecánica en ejercicios introductorios:
$
\text{PME}_f \;\approx\; \text{PME}_i \cdot \eta_{\text{th}}
$
