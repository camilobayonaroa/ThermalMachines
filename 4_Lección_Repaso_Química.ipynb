## 1. Moléculas, Moles y Átomos

- **Átomo:** Unidad básica de un elemento químic.

![orgle](https://upload.wikimedia.org/wikipedia/commons/thumb/2/23/Helium_atom_QM.svg/640px-Helium_atom_QM.svg.png)


- **Molécula:** Conjunto de dos o más átomos unidos mediante enlaces químicos.

![orgle](https://upload.wikimedia.org/wikipedia/commons/thumb/9/90/Thermally_Agitated_Molecule_%28grayscale%29.gif/640px-Thermally_Agitated_Molecule_%28grayscale%29.gif)

- **Mol:** Unidad que representa $6.022 \times 10^{23}$ entidades (número de Avogadro). Es la cantidad de sustancia en la que la masa en gramos equivale al peso molecular del compuesto.

![orgle](https://upload.wikimedia.org/wikipedia/commons/a/ac/Mass_versus_moles_of_iron_vs_gold.svg)


**Ejemplo de cálculo:**

Si tienes 18 g de agua (H₂O, masa molar ≈18 g/mol), tienes 1 mol de agua, lo que equivale a $6.022 \times 10^{23}$ moléculas de agua.


![Petable](https://upload.wikimedia.org/wikipedia/commons/4/4d/Periodic_table_large.svg)



## 2. Enlaces Químicos

Los enlaces químicos son las fuerzas que mantienen unidos a los átomos en una molécula. Se pueden clasificar en:

- **Enlace Iónico:**
  - Ocurre por la transferencia completa de electrones entre átomos.
  - Ejemplo: NaCl.
  - Características: Alta energía de reticulación, alta solubilidad en agua.


![orgle](https://upload.wikimedia.org/wikipedia/commons/7/75/Ionic_bonding_animation.gif)


- **Enlace Covalente:**
  - Se forma por el compartimiento de electrones entre átomos.
  - Ejemplo: H₂O, CH₄.
  - Puede ser polar (cuando hay diferencia de electronegatividad) o no polar.

![orgle](https://upload.wikimedia.org/wikipedia/commons/c/c2/Covalent_bond_h.png)

- **Enlace Metálico:**
  - Se produce en metales, donde los electrones de valencia se comparten colectivamente.
  - Propiedades: Alta conductividad eléctrica y térmica, maleabilidad y ductilidad.

![orgle](https://upload.wikimedia.org/wikipedia/commons/a/aa/Metalic_bond_model.svg)

- **Interacciones Débiles:**
  - Fuerzas de Van der Waals, enlaces de hidrógeno y fuerzas dipolo-dipolo, que aunque son más débiles, juegan un papel crucial en la estructura y propiedades de sustancias orgánicas.

![orgle](https://www.daviddarling.info/images/hydrogen_bonding_in_water.gif)



## 3. Electrones de Valencia

- **Definición:** Son los electrones que se encuentran en la capa más externa de un átomo y son los responsables de la formación de enlaces químicos.
- **Importancia:**
  - Determinan la reactividad química del átomo.
  - Dictan la geometría molecular según la teoría de repulsión de pares de electrones (VSEPR).


![orgle](https://upload.wikimedia.org/wikipedia/commons/0/0b/Electrones_de_valencia_del_Grupo_IVA.png)

## 4. Electronegatividad

- **Definición:** Es la capacidad de un átomo para atraer los electrones compartidos en un enlace covalente.
- **Tendencias:**
  - Aumenta de izquierda a derecha a lo largo de un período.
  - Disminuye de arriba hacia abajo en un grupo.
- **Ejemplo:** El flúor es el elemento más electronegativo, lo que explica la polaridad de los enlaces C–F.

![orgle](https://upload.wikimedia.org/wikipedia/commons/5/57/Periodic_table_electronegativity.png)


## 5. Química Inorgánica

La química es la ciencia que estudia la composición, estructura, propiedades y cambios de la materia. Se divide en dos grandes ramas:

- **Química Inorgánica:** Estudia compuestos sin enlaces carbono-hidrógeno (salvo algunas excepciones). Se enfoca en minerales, metales, sales, ácidos y bases, entre otros.
- **Química Orgánica:** Se centra en compuestos que contienen carbono y, usualmente, hidrógeno. Incluye una enorme variedad de sustancias, desde moléculas simples hasta macromoléculas complejas.


### 5.1. Química Inorgánica vs. Química Orgánica

- **Química Inorgánica:**
  - Predomina en minerales, metales y sales.
  - Ejemplos: agua (H₂O), cloruro de sodio (NaCl), óxidos metálicos.
  - Menor complejidad en estructuras moleculares en comparación con la orgánica.


![inorgle](https://prod-cat-files.macmillan.cloud/MediaResources/Jackets/258W/9781464176807.jpg)
  






- **Química Orgánica:**
  - Estudia compuestos del carbono, fundamentalmente en combinación con hidrógeno, oxígeno, nitrógeno, halógenos, etc.
  - Ejemplos: metano (CH₄), etanol (C₂H₅OH), ácidos orgánicos.
  - Gran diversidad estructural y funcional, con la presencia de grupos funcionales.

![orgle](https://upload.wikimedia.org/wikipedia/commons/7/70/Periodic_Table_Organic_Chemistry_01.png)

## 6. Química Orgánica
La química orgánica abarca una amplia gama de compuestos, clasificados según su estructura y funcionalidad:

- **Alquilos y Acilos:** Derivados de los alcanos y ácidos carboxílicos, respectivamente.
- **Alcoholes:** Compuestos que contienen uno o más grupos -OH.
- **Éteres:** Moléculas donde un átomo de oxígeno conecta dos grupos alquilo o arilo.
- **Aldehídos y Cetonas:** Contienen el grupo funcional carbonilo (C=O); en los aldehídos el grupo está en el extremo de la cadena, en las cetonas en posición interna.
- **Ácidos Carboxílicos:** Presentan el grupo carboxilo (-COOH).
- **Esteres y Amidas:** Derivados de los ácidos carboxílicos, formados mediante la reacción con alcoholes (ésteres) o aminas (amidas).
- **Aminas:** Compuestos que contienen grupos funcionales derivados del amoníaco (NH₃).

Cada grupo funcional confiere propiedades químicas y físicas específicas, y la diversidad de estructuras orgánicas permite la gran variedad de aplicaciones en la industria, biología y materiales.

### 6.1. Hidrocarburos

Los hidrocarburos son compuestos formados únicamente por carbono e hidrógeno. Se clasifican principalmente en:

- **Alcanos:** Hidrocarburos saturados con enlaces simples.  
  - Fórmula general: $C_nH_{2n+2}$.
  - Ejemplo: Metano (CH₄).

![orgle](https://upload.wikimedia.org/wikipedia/commons/thumb/9/94/Methane_CH4.png/640px-Methane_CH4.png)

- **Alquenos:** Hidrocarburos insaturados con uno o más enlaces dobles.  
  - Fórmula general: $C_nH_{2n}$ (para alquenos lineales).
  - Ejemplo: Eteno (C₂H₄).

![orgle](https://upload.wikimedia.org/wikipedia/commons/thumb/7/72/Ethylene_8147.JPG/640px-Ethylene_8147.JPG)

- **Alquinos:** Hidrocarburos insaturados con uno o más enlaces triples.  
  - Fórmula general: $C_nH_{2n-2}$.
  - Ejemplo: Etino (C₂H₂).

![orgle](https://upload.wikimedia.org/wikipedia/commons/thumb/c/c4/Acetylene_8163.JPG/640px-Acetylene_8163.JPG)

- **Hidrocarburos Aromáticos:** Compuestos con anillos de benceno, caracterizados por su estabilidad y deslocalización de electrones.  
  - Ejemplo: Benceno (C₆H₆).

  ![orgle](https://upload.wikimedia.org/wikipedia/commons/8/82/Benzene-non-aromatic-3D-balls.png)

```python
# Instalar Graphviz (si es necesario) en Colab/Jupyter:
!apt-get update -qq
!apt-get install graphviz -y
!pip install graphviz

from graphviz import Digraph

dot = Digraph(comment='Clasificación de Hidrocarburos', format='png')
dot.attr(rankdir='TB', size='10,8', fontsize='10')

# Nodo raíz
dot.node('H', 'Hidrocarburos\n(Compuestos de C y H)', shape='ellipse', style='filled', fillcolor='lightblue')

# Nodos para cada grupo
dot.node('A', 'Alcanos\nFórmula: C_nH₂n₊₂\nEjemplo: Metano (CH₄)', shape='box', style='filled', fillcolor='lightyellow')
dot.node('B', 'Alquenos\nFórmula: CnH₂n\nEjemplo: Eteno (C₂H₄)', shape='box', style='filled', fillcolor='lightgreen')
dot.node('C', 'Alquinos\nFórmula: CnH₂n₋₂\nEjemplo: Etino (C₂H₂)', shape='box', style='filled', fillcolor='lightcoral')
dot.node('D', 'Hidrocarburos\nAromáticos\nEjemplo: Benceno (C₆H₆)', shape='box', style='filled', fillcolor='wheat')

# Conectar el nodo raíz con cada grupo
dot.edge('H', 'A')
dot.edge('H', 'B')
dot.edge('H', 'C')
dot.edge('H', 'D')

dot.render('clasificacion_hidrocarburos', view=True)
dot
```

Diagrama 3D: Proyección Cuña-Dash para un Centro Quiral

Este diagrama muestra un centro quiral (representado como C*), al que se unen cuatro grupos (R₁, R₂, R₃, R₄) con distintos estilos para representar la orientación espacial:

    Cuña (saliente): Representa un enlace que sale del plano hacia el observador (se dibuja con línea gruesa o "bold").
    Línea (en plano): Representa un enlace que se proyecta en el plano de la pantalla o papel.
    Discontinua (entrante): Representa un enlace que se aleja del observador (se dibuja con línea discontinua).

```python
from graphviz import Graph

# Usamos el motor "neato" para posicionamiento manual
dot3d = Graph('3D_Enlaces', engine='neato', format='png', comment='Representación 3D (Proyección Cuña-Dash)')
dot3d.attr(overlap='false', splines='true')

# Definir nodos con posiciones manuales (la notación "pos" requiere el flag "!" para posiciones fijas)
dot3d.node('C', 'C*', pos='0,0!')
dot3d.node('R1', 'R₁', pos='1,1!')   # Enlace saliente (cuña)
dot3d.node('R2', 'R₂', pos='-1,1!')  # Enlace en plano
dot3d.node('R3', 'R₃', pos='-1,-1!') # Enlace entrante (discontinua)
dot3d.node('R4', 'R₄', pos='1,-1!')  # Enlace en plano

# Agregar aristas con estilos para simular la proyección cuña-dash
# Enlace cuña: estilo "bold" para indicar que sale del plano
dot3d.edge('C', 'R1', label='Cuña', style='bold', color='black')
# Enlaces en plano: línea sólida
dot3d.edge('C', 'R2', label='Línea', style='solid', color='black')
dot3d.edge('C', 'R4', label='Línea', style='solid', color='black')
# Enlace discontinua: línea discontinua para indicar que va hacia atrás
dot3d.edge('C', 'R3', label='Discontinua', style='dashed', color='black')

# Renderizar y mostrar el diagrama
dot3d.render('enlaces_3d', view=True)
dot3d
```

## 10. Reacciones

Ejemplo 1: Hidrogenación del Eteno (C₂H₄ + H₂ → C₂H₆)

Este diagrama ilustra cómo el eteno reacciona con hidrógeno. En el proceso se rompe el enlace π del eteno y se forman nuevos enlaces σ para producir etano.

```python
# Instalar Graphviz (si es necesario) en Colab/Jupyter:
!apt-get update -qq
!apt-get install graphviz -y
!pip install graphviz

from graphviz import Digraph

dot = Digraph(comment='Hidrogenación del Eteno', format='png')
dot.attr(rankdir='LR', size='10,5')

# Cluster para Reactivos
with dot.subgraph(name='cluster_reactivos') as c:
    c.attr(label='Reactivos', style='filled', color='lightgrey')
    c.node('E', 'Eteno\n(C₂H₄)')
    c.node('H', 'Hidrógeno\n(H₂)')

# Cluster para el Proceso de Recambio de Enlace
with dot.subgraph(name='cluster_proceso') as p:
    p.attr(label='Proceso\n(Recambio de enlace)', style='filled', color='lightblue')
    p.node('P', 'Romper enlace π\nFormar enlaces σ')

# Cluster para Productos
with dot.subgraph(name='cluster_productos') as pr:
    pr.attr(label='Productos', style='filled', color='lightgreen')
    pr.node('A', 'Etano\n(C₂H₆)')

# Conexiones: Reactivos → Proceso → Productos
dot.edge('E', 'P', arrowhead='normal')
dot.edge('H', 'P', arrowhead='normal')
dot.edge('P', 'A', arrowhead='normal')

dot.render('hidrogenacion_eteno', view=True)
dot
```

A continuación, se muestra un ejemplo de diagrama en Graphviz que representa la combustión del metano (CH₄), en el que se ilustran los reactivos (con sus diagramas moleculares), el proceso de combustión (recambio de enlaces) y los productos formados. La reacción es:

CH4+2 O2→CO2+2 H2O

```python
# Instalar Graphviz (si es necesario) en Colab/Jupyter:
!apt-get update -qq
!apt-get install graphviz -y
!pip install graphviz

from graphviz import Digraph

# Crear el diagrama de la combustión del metano
dot = Digraph(comment='Combustión del Metano (CH₄)', format='png')
dot.attr(rankdir='LR', size='12,8')  # Disposición de izquierda a derecha

# Cluster para los Reactivos
with dot.subgraph(name='cluster_reactivos') as r:
    r.attr(label='Reactivos', style='filled', color='lightgrey')
    # Metano (CH₄) con un dibujo sencillo en ASCII
    r.node('CH4', 'CH₄\n\n   H\n   |\nH–C–H\n   |\n   H', shape='box', fontname='Courier')
    # Oxígeno (O₂) representado como O=O
    r.node('O2a', 'O₂\n\nO=O', shape='box', fontname='Courier')
    r.node('O2b', 'O₂\n\nO=O', shape='box', fontname='Courier')

# Cluster para el Proceso de Combustión
with dot.subgraph(name='cluster_proceso') as p:
    p.attr(label='Proceso: Combustión', style='filled', color='lightblue')
    p.node('Process', 'Combustión:\nRomper enlaces π\ny formar enlaces σ', shape='box', fontname='Courier')

# Cluster para los Productos
with dot.subgraph(name='cluster_productos') as pr:
    pr.attr(label='Productos', style='filled', color='lightgreen')
    # Dióxido de carbono (CO₂) representado como O=C=O
    pr.node('CO2', 'CO₂\n\nO=C=O', shape='box', fontname='Courier')
    # Agua (H₂O) representada como H–O–H
    pr.node('H2O1', 'H₂O\n\nH–O–H', shape='box', fontname='Courier')
    pr.node('H2O2', 'H₂O\n\nH–O–H', shape='box', fontname='Courier')

# Conectar los reactivos con el proceso
dot.edge('CH4', 'Process', label='Combustión', arrowhead='normal')
dot.edge('O2a', 'Process', arrowhead='normal')
dot.edge('O2b', 'Process', arrowhead='normal')

# Conectar el proceso con los productos
dot.edge('Process', 'CO2', arrowhead='normal')
dot.edge('Process', 'H2O1', arrowhead='normal')
dot.edge('Process', 'H2O2', arrowhead='normal')

# Renderizar y mostrar el diagrama
dot.render('combustion_metano_diagram', view=True)
dot
```

# Cinética de Reacción

La cinética química estudia la velocidad a la que ocurren las reacciones químicas y los factores que influyen en dichas velocidades. En este repaso se analizan los conceptos fundamentales, tales como la tasa de reacción, la ley de velocidad, la dependencia de la temperatura (Ecuación de Arrhenius) y la representación gráfica de una reacción.

## 1. Tasa de Reacción

La **tasa de reacción** se define como el cambio de concentración de un reactivo (o producto) por unidad de tiempo. Para un reactivo A:

$
\text{Tasa} = -\frac{d[A]}{dt}
$

El signo negativo indica que la concentración de A disminuye a medida que avanza la reacción.

## 2. Ley de Velocidad

Para una reacción general:

$
aA + bB \rightarrow cC + dD
$

La ley de velocidad se expresa habitualmente como:

$
\text{Velocidad} = k [A]^m [B]^n
$

donde:
- $ k $ es la constante de velocidad.
- $ [A] $ y $ [B] $ son las concentraciones de los reactivos.
- $ m $ y $ n $ son los órdenes de reacción respecto a A y B, determinados experimentalmente (no necesariamente iguales a los coeficientes estequiométricos $ a $ y $ b $).

## 3. Ejemplo: Combustión del Metano

La combustión del metano se puede escribir como:

$
\text{CH}_4 + 2\,\text{O}_2 \rightarrow \text{CO}_2 + 2\,\text{H}_2\text{O}
$

La ley de velocidad para esta reacción (según un estudio experimental hipotético) podría ser:

$
\text{Velocidad} = k [\text{CH}_4]^m [\text{O}_2]^n
$

Por ejemplo, si se determina que la reacción es de primer orden respecto al metano ($ m = 1 $) y de segundo orden respecto al oxígeno ($ n = 2 $), la expresión queda:

$
\text{Velocidad} = k [\text{CH}_4][\text{O}_2]^2
$

## 4. Ecuación de Arrhenius

La constante de velocidad $ k $ depende de la temperatura según la ecuación de Arrhenius:

$
k = A \, e^{-E_a/(RT)}
$

donde:
- $ A $ es el factor pre-exponencial (o de frecuencia).
- $ E_a $ es la energía de activación.
- $ R $ es la constante de los gases (8.314 J/(mol·K)).
- $ T $ es la temperatura en Kelvin.

Esta ecuación muestra que, al aumentar la temperatura, $ k $ aumenta, acelerando la reacción.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
R = 8.314  # Constante de los gases en J/(mol*K)
A = 1e12   # Factor pre-exponencial (s^-1), valor arbitrario para la comparación

# Rango de temperaturas en Kelvin (por ejemplo, de 300K a 1500K)
T = np.linspace(300, 1500, 300)

# Lista de energías de activación en J/mol para comparar (50, 75, 100, 125 kJ/mol)
Ea_values = [50e3, 75e3, 100e3, 125e3]

plt.figure(figsize=(8,6))

for Ea in Ea_values:
    k = A * np.exp(-Ea / (R * T))
    plt.plot(T, k, label=f'$E_a$ = {Ea/1000:.0f} kJ/mol')

plt.xlabel('Temperatura (K)')
plt.ylabel('Constante de velocidad, k (s⁻¹)')
plt.title('Gráfica de Arrhenius para la Combustión del Metano')
plt.legend()
plt.grid(True)
plt.yscale('log')  # Escala logarítmica para k
plt.tight_layout()
plt.show()
```

# Derivación de la Ecuación de Concentración en Función del Tiempo

Consideremos una reacción de primer orden, donde la tasa de disminución de la concentración es proporcional a la concentración misma. La ecuación diferencial es:

$
\frac{dC}{dt} = -k C,
$

donde:
- $ C(t) $ es la concentración en el tiempo $ t $.
- $ k $ es la constante de velocidad de la reacción.

## 1. Separación de Variables

Reorganizamos la ecuación para separar las variables $ C $ y $ t $:

$
\frac{dC}{C} = -k \, dt.
$

## 2. Integración

Integramos ambos lados de la ecuación:

$
\int \frac{1}{C}\, dC = -k \int dt.
$

La integral del lado izquierdo es:

$
\int \frac{1}{C}\, dC = \ln |C| + C_1,
$

y la integral del lado derecho es:

$
\int dt = t + C_2.
$

Podemos combinar las constantes de integración ($ C_1 $ y $ C_2 $) en una única constante. Escribimos:

$
\ln |C| = -k t + \ln C_0,
$

donde $ \ln C_0 $ es la constante de integración determinada por la condición inicial $ C(0) = C_0 $.

## 3. Resolviendo para $ C(t) $

Exponenciamos ambos lados de la ecuación para despejar $ C $:

$
|C| = e^{-k t + \ln C_0} = C_0 \, e^{-k t}.
$

Como la concentración $ C $ es siempre positiva, podemos omitir el valor absoluto:

$
C(t) = C_0 \, e^{-k t}.
$

## Resultado Final

La concentración en función del tiempo para una reacción de primer orden se expresa como:

$
\boxed{C(t) = C_0 \, e^{-k t}},
$

donde:
- $ C_0 $ es la concentración inicial.
- $ k $ es la constante de velocidad.
- $ t $ es el tiempo.


```python
import numpy as np
import matplotlib.pyplot as plt

# Concentración inicial
C0 = 1.0  # unidad arbitraria

# Rango de tiempo (en segundos)
t = np.linspace(0, 10, 200)

# Constantes de velocidad
k_oxidacion = 0.1   # Oxidación: reacción lenta
k_combustion = 1.0   # Combustión: reacción rápida

# Evolución de la concentración en cada proceso (modelo de primer orden)
C_oxidacion = C0 * np.exp(-k_oxidacion * t)
C_combustion = C0 * np.exp(-k_combustion * t)

# Crear la gráfica
plt.figure(figsize=(8, 5))
plt.plot(t, C_oxidacion, label="Oxidación (k = 0.1 s⁻¹)", linewidth=2)
plt.plot(t, C_combustion, label="Combustión (k = 1.0 s⁻¹)", linewidth=2)

plt.xlabel("Tiempo (s)")
plt.ylabel("Concentración del reactante (C)")
plt.title("Comparación de velocidades: Oxidación vs. Combustión")
plt.legend()
plt.grid(True)
plt.show()
```
