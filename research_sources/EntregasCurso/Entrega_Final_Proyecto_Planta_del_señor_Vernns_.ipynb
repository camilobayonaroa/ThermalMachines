# 1. Introducción

La energía nuclear ha ocupado un lugar central en el desarrollo científico y tecnológico de la humanidad durante el último siglo. A partir del descubrimiento de la radiactividad y la formulación de principios fundamentales como la equivalencia masa–energía, la humanidad logró dominar dos procesos fundamentales: la fisión nuclear, que consiste en la división de núcleos pesados liberando gran cantidad de energía, y la fusión nuclear, que busca reproducir las reacciones que ocurren en el interior de las estrellas mediante la unión de núcleos ligeros.

La fisión nuclear ha sido la base de las plantas nucleares actuales, con un papel destacado en la producción de electricidad en múltiples países. No obstante, también ha estado asociada a debates sociales, económicos y ambientales, debido tanto a la gestión de residuos radiactivos como a los riesgos de seguridad. Por su parte, la fusión nuclear se presenta como una alternativa con un potencial inmenso: promete una fuente de energía más limpia, segura y prácticamente inagotable, aunque aún enfrenta grandes desafíos técnicos para su implementación a gran escala.

### 3.2 Cinética y energía de combustión

En la planta nuclear no hay *combustión* química; el calor proviene de la **fisión** de núcleos fisibles (p. ej., $^{235}$U).  
Una reacción representativa es:

$$
^{235}\text{U} + n \;\rightarrow\; \text{fragmentos de fisión} + \nu n + \text{energía}\ (\approx 200\ \mathrm{MeV})
$$

La **energía liberada por fisión** de $^{235}$U es del orden de $200$–$207\ \mathrm{MeV}$ por evento [2].  
A continuación se muestra la conversión a unidades de ingeniería:

$$
\begin{aligned}
E_f &\approx 200\,\mathrm{MeV}\times 1.602\times 10^{-13}\,\frac{\mathrm{J}}{\mathrm{MeV}}\\
&= 3.204\times 10^{-11}\ \mathrm{J/fisión},\\[6pt]
E_{\text{mol}} &= E_f\,N_A
\approx 3.204\times 10^{-11}\times 6.022\times 10^{23}\\
&\approx 1.93\times 10^{13}\ \mathrm{J/mol},\\[6pt]
E_{\text{kg}} &= E_{\text{mol}}\times \frac{1000}{235}\\
&\approx 8.2\times 10^{13}\ \mathrm{J/kg} \quad (\sim 82\ \mathrm{TJ/kg}).
\end{aligned}
$$

---

#### Reparto de la energía de fisión

La energía de $\sim 200$ MeV no se destina toda a calor. La distribución aproximada es:

- $\sim 168$ MeV → energía cinética de fragmentos de fisión (se transforma en calor en el combustible).  
- $\sim 5$ MeV → energía cinética de neutrones.  
- $\sim 7$ MeV → rayos gamma inmediatos.  
- $\sim 20$ MeV → neutrinos (escapan y no contribuyen al calor).  

De este modo, **aprox. el 90% de la energía total (≈ 180 MeV)** es aprovechable como calor útil:

$$
E_{f,\ \text{térmica}} \approx 0.9\,E_f
= 2.88\times 10^{-11}\ \mathrm{J/fisión}.
$$

En magnitudes macroscópicas:

$$
\begin{aligned}
E_{\text{mol,térmica}} &= E_{f,\ \text{térmica}}\,N_A
\approx 1.74\times 10^{13}\ \mathrm{J/mol},\\[6pt]
E_{\text{kg,térmica}} &= E_{\text{mol,térmica}}\times \frac{1000}{235}
\approx 7.4\times 10^{13}\ \mathrm{J/kg} \quad (\sim 74\ \mathrm{TJ/kg}).
\end{aligned}
$$

---

#### Conversión a temperatura equivalente (constante de Boltzmann)

Si se interpreta la energía de una fisión como energía cinética de una partícula, puede asociarse a una temperatura equivalente mediante:

$$
T = \frac{2E}{3k_B},
$$

con $k_B = 1.3806\times 10^{-23}\ \mathrm{J/K}$.  

Para la energía total:

$$
T_{\text{eq}} \approx \frac{2\,(3.204\times 10^{-11})}{3\,(1.3806\times 10^{-23})}
\approx 1.55\times 10^{12}\ \mathrm{K}.
$$

Para la fracción térmica:

$$
T_{\text{eq,térmica}} \approx \frac{2\,(2.88\times 10^{-11})}{3\,(1.3806\times 10^{-23})}
\approx 1.39\times 10^{12}\ \mathrm{K}.
$$

---

#### Estimación práctica (1 kg de $^{235}$U)

En la práctica, la energía se reparte y se disipa en el combustible y el refrigerante.  
Si **1 kg de U-235** liberara toda su energía térmica y la absorbiéramos solo en el propio material (con calor específico $c \approx 116\ \mathrm{J/(kg\cdot K)}$), el aumento de temperatura sería:

$$
\Delta T \approx \frac{E_{\text{kg,térmica}}}{m\,c}
= \frac{7.4\times 10^{13}}{1\times 116}
\approx 6.4\times 10^{11}\ \mathrm{K}.
$$

---

Este valor es **puramente teórico**: en un reactor real, la energía se transfiere al refrigerante y las temperaturas típicas de operación son de apenas **600–800 K (300–500 °C)**.


### Reacción canónica de fisión y conservación

Una vía representativa de fisión térmica en U-235 es:

$$
{}^{235}_{92}\mathrm{U}+{}^{1}_{0}\mathrm{n}\;\rightarrow\;{}^{141}_{56}\mathrm{Ba}+{}^{92}_{36}\mathrm{Kr}+3\,{}^{1}_{0}\mathrm{n}
$$

- **Conservación**:  
  $A$: $235+1=236=141+92+3\cdot1$;  
  $Z$: $92=56+36$.  

- **Energía liberada**: $\approx 200~\mathrm{MeV}$ por fisión $\;\Rightarrow\;$ calor del reactor.  

- **Multiplicación**: emite $\nu$ neutrones/evento ($\bar\nu\!\approx\!2.43$ térmico en U-235), habilitando la **reacción en cadena**.  

- **Conexión con nuestro modelo**:  
  $$
  \dot n=\frac{\rho-\beta}{\ell}n+\sum\lambda_i C_i
  $$
  (producción *prompt* + *retardada*).  

  Usamos $E_f\!\approx\!200~\mathrm{MeV}$ para relacionar fisiones/s $\leftrightarrow$ potencia.

### Formulación cinética (neutrónica) para la potencia  

En un reactor nuclear, la variación del número de neutrones presentes en el tiempo se explica como un balance sencillo: por un lado, se generan neutrones a través de los procesos de fisión, y por otro, se pierden debido a fenómenos como la absorción en materiales no fisibles y la fuga fuera del núcleo. De esta manera, la ecuación general puede expresarse como la tasa de cambio de neutrones en función del tiempo.
$$\frac{dn(t)}{dt}=na-nm$$
	​

Que corresponde a la diferencia entre los “nacimientos”(na) de neutrones y las “muertes” o pérdidas que ocurren en el sistema(nm). Por ello, los parámetros fundamentales para describir la evolución de los neutrones en un reactor son: el tiempo de vida prompt del neutrón(ℓ)  ,definido como el tiempo promedio entre la creación y la absorción de un neutrón; el factor de multiplicación efectivo (keff) que determina si la reacción en cadena aumenta, permanece estable o disminuye; y la reactividad (𝜌), una medida adimensional que expresa la desviación del reactor respecto a la criticidad y que se define como :

$$ ρ=\frac{keff​−1}{keff}$$
A partir de estas definiciones, formula la ecuación que describe la población de neutrones (sin incluir aún el efecto de los neutrones retardados):
$$\frac{dn(t)}{dt}=\frac{ρ}{ℓ}n(t)$$
la cual indica que, si 𝜌>0 ,el número de neutrones crece exponencialmente, mientras que, si 𝜌<, dicho número decrece con el tiempo.
En los reactores nucleares, la mayoría de los neutrones se producen inmediatamente tras la fisión, conocidos como neutrones prompt. Sin embargo, una pequeña fracción de neutrones, aproximadamente 𝛽 ≈ 0.0065 en U-235se origina a partir del decaimiento beta de fragmentos de fisión inestables; estos son los neutrones retardados. La importancia de los neutrones retardados radica en que permiten que la reacción en cadena del reactor sea controlable, ya que sin ellos los tiempos característicos serían del orden de microsegundos, imposibles de manejar con sistemas de control. Para modelar la dinámica de los neutrones retardados, se definen variables Ci(t) correspondientes a cada grupo de precursores, cuya evolución está dada por la ecuación diferencial:
$$\frac{dCi(t)}{dt}=\frac{βi}{ℓ}n(t)-λiCi(t)$$

Por ende, La evolución temporal de la potencia a baja potencia se modela con las **ecuaciones de cinética puntual** con 6 grupos de neutrones retardados:

$$
\begin{aligned}
\frac{dn(t)}{dt} &= \frac{\rho(t)-\beta}{\ell}\,n(t) + \sum_{i=1}^{6}\lambda_i\,C_i(t), \\[6pt]
\frac{dC_i(t)}{dt} &= \frac{\beta_i}{\ell}\,n(t) - \lambda_i\,C_i(t), \quad i=1,\dots,6.
\end{aligned}
$$

donde $n(t)$ es proporcional a la potencia, $\rho$ es la *reactividad*,  
$\beta=\sum_i\beta_i$ la fracción de retardados, $\ell$ el tiempo de vida *prompt* y $\lambda_i$ las constantes de decaimiento de los precursores [2][3].  

Valores típicos para fisión térmica de $^{235}$U:  
$\beta \approx 0.0065$ y $\ell$ en el orden de $10^{-5}$–$10^{-3}\ \mathrm{s}$.  
Las constantes $(\beta_i,\lambda_i)$ se encuentran tabuladas en textos clásicos [2].

---

La potencia térmica puede relacionarse con el flujo neutrónico[5]
$$
Potencia=(\text{Tasa liberada por fision})*(\text{Enegia liberada por fision})
$$
$$
P = \Phi\,\Sigma_f\,E_f\,V \;=\; \Phi\,N_{^{235}\!U}\,\sigma_{f,235}\,E_f\,V
$$

donde:  
- $\Phi$ es el flujo neutrónico,  
- $\Sigma_f$ la sección eficaz macroscópica de fisión,  
- $N_{^{235}\!U}$ la densidad atómica fisible,  
- $\sigma_{f,235}$ la sección eficaz microscópica,  
- $V$ el volumen efectivo.

### Cálculo de magnitudes derivadas de la reacción

A partir de la energía por fisión ($\sim 200 \,\text{MeV}$) y la potencia térmica de diseño ($3000 \,\text{MWt}$), se estiman:

- **Fisiones por segundo** en el reactor.  
- **Neutrones por segundo** emitidos (usando $\nu \approx 2.43$).  
- **Consumo de masa de U-235** en kg/h y kg/día.  

Esto permite cuantificar en términos de ingeniería lo que significa la ecuación nuclear, dando una idea tangible de cuánta materia se transforma y qué flujo neutrónico se maneja.




```python
import numpy as np
from math import ceil

# Parámetros físicos
E_f_MeV = 200.0
J_per_MeV = 1.602176634e-13
E_f = E_f_MeV * J_per_MeV               # J/fisión
nu_bar = 2.43                           # neutrones por fisión (promedio U-235 térmico)
NA = 6.02214076e23

# === Parámetros del reactor ===
MWt = 3000                              # Potencia térmica [MWt]
eta = 0.33                              # Eficiencia térmica → eléctrica
MW_e = MWt * eta                        # Potencia eléctrica [MWe]

# Potencia térmica a n=1 (ya definida arriba): Wth = MWt*1e6
Pth_W = MWt * 1e6                       # W

# Tasas
R_f = Pth_W / E_f                       # fisiones por segundo a n=1
R_n = nu_bar * R_f                      # neutrones/s

# Consumo de U-235 equivalente (si todo el calor fuese de U-235)
m_fission_kg = 0.235 / NA               # kg por fisión (235 g/mol)
burn_kg_s  = R_f * m_fission_kg
burn_kg_h  = burn_kg_s * 3600
burn_kg_d  = burn_kg_s * 86400

# === Resultados ===
print("=== Magnitudes derivadas (a n=1) ===")
print(f"Potencia térmica:      {MWt:,.0f} MWt  (~ {MW_e:,.0f} MWe)")
print(f"Energía por fisión:    {E_f:.3e} J")
print(f"Fisiones por segundo:  {R_f:,.3e} 1/s")
print(f"Neutrones por segundo: {R_n:,.3e} 1/s (con ν̄={nu_bar})")
print(f"Consumo U-235:         {burn_kg_h:,.3f} kg/h  (~{burn_kg_d:,.3f} kg/día)")
```

### Desglose de la energía liberada por fisión

```python
import numpy as np
import matplotlib.pyplot as plt

# Desglose típico aproximado (MeV/fisión)
parts = {
    "Fragmentos (Ec)": 168.0,
    "Neutrones (Ec)":   7.0,
    "γ prompt":         5.0,
    "β/γ retardados":  12.0,
    "Antineutrinos":    8.0    # no recuperable
}
total = sum(parts.values())
heat_recoverable = total - parts["Antineutrinos"]

print(f"E_total ≈ {total:.0f} MeV/fisión; recuperable ≈ {heat_recoverable:.0f} MeV (≈ {heat_recoverable/total*100:.0f}%).")

# Gráfico simple
labels = list(parts.keys())
vals   = list(parts.values())

plt.figure(figsize=(7,4))
plt.bar(labels, vals)
plt.ylabel("MeV por fisión")
plt.title("Desglose energético aproximado por fisión (U-235)")
plt.xticks(rotation=20)
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
```

### 3.3 Parámetros cinéticos y fuentes

Parámetros de cinética puntual (reactor térmico, $^{235}$U). En el modelado a baja potencia, la dinámica se describe con *cinética puntual* y 6 grupos de neutrones retardados:

$$
\begin{aligned}
\frac{dn(t)}{dt} &= \frac{\rho(t)-\beta}{\ell}\,n(t) + \sum_{i=1}^{6}\lambda_i\,C_i(t), \\[6pt]
\frac{dC_i(t)}{dt} &= \frac{\beta_i}{\ell}\,n(t) - \lambda_i\,C_i(t),
\qquad i=1,\dots,6.
\end{aligned}
$$

- $\rho(t)$: reactividad (adimensional). En práctica, $1\ \text{pcm}=10^{-5}$.  
- $\beta=\sum_i \beta_i$: fracción total de neutrones retardados.  
- $\ell$: tiempo de vida *prompt* (orden $10^{-5}$–$10^{-3}$ s según diseño).  
- $(\beta_i,\lambda_i)$: fracciones y constantes de decaimiento de los precursores retardados.  

---

**Relaciones útiles para transitorios:**

- *Ecuación inhour* (raíces $s$ de un escalón de $\rho$):

$$

$$

- *Salto prompt* (para $0<\rho<\beta$):

$$
\frac{P(0^+)}{P(0^-)}=\frac{\beta}{\beta-\rho}
$$

- **Supercrítico**: $\rho>\beta$ (p. ej. $+800\ \mathrm{pcm}$); **sin** Doppler para ilustrar condición peligrosa → mostrar **WARNING** visual.  
- **Normal**: $\rho_0=+100\ \mathrm{pcm}=10^{-3}$ ($0<\rho<\beta$); con **Doppler**, tiempo de simulación = 1 hora.  
- **Subcrítico**: $\rho=-200\ \mathrm{pcm}$; con **Doppler**, tiempo de simulación = 10 minutos.  

---

A continuación se simula y grafican \(P(t)\), \(Q(t)\) (MWh) y el costo acumulado (COP) con una tarifa dada.


```python
# PlantaNuclear_cases_full.py
# Traducción del script MATLAB a Python
import numpy as np
from scipy.integrate import solve_ivp, cumulative_trapezoid
import matplotlib.pyplot as plt

# ======== Parámetros base ========
beta_i = np.array([0.000215, 0.001424, 0.001274, 0.002568, 0.000748, 0.000273])[:, None]  # columna
lambda_i = np.array([0.0124, 0.0305, 0.1110, 0.3010, 1.1400, 3.0100])[:, None]
beta = beta_i.sum()
l = 5e-5                 # [s] tiempo de vida prompt
n0 = 1.0                 # potencia normalizada al inicio

# Módulo térmico simple para retroalimentación Doppler
MWt = 3000               # [MWt] a n=1
MW_e = 1000              # [MWe] equivalente (≈33% eficiencia)
Wth = MWt * 1e6          # [W] térmicos a n=1
alpha_comb = -3e-5       # [Δρ/K]
T0 = 600.0               # [K] temperatura combustible de referencia
Tcool = 580.0            # [K] refrigerante (referencia)
Mcp = 5e10               # [J/K]
UA = 5e7                 # [W/K]

# Economía
COP_kWh = 600            # [COP/kWh]

# Integrador: tolerancias (similar a odeset en MATLAB)
rtol = 1e-7
atol = 1e-9

# Umbral visual de WARNING
n_limit = 50.0
T_limit = 2000.0

# ======== Función del modelo (cinética puntual + nudo térmico) ========
def kinetics_thermal(t, y, rho_ext_fun, beta_i, lambda_i, beta, l,
                     alpha_comb, Wth, Mcp, UA, Tcool, T0):
    # y = [n, C1..C6, T]
    n = y[0]
    Ci = y[1:7].reshape((-1,1))
    T = y[7]
    rho_ext = rho_ext_fun(t)
    rho = rho_ext + alpha_comb*(T - T0)
    # Cinética puntual (6 grupos)
    dn = ((rho - beta)/l)*n + float((lambda_i * Ci).sum())
    dCi = (beta_i * n)/l - (lambda_i * Ci)
    # Nudo térmico (balance 1 nodo)
    Pth = max(n, 0.0) * Wth
    dT = (Pth - UA*(T - Tcool)) / Mcp
    dydt = np.empty(8)
    dydt[0] = dn
    dydt[1:7] = dCi.flatten()
    dydt[7] = dT
    return dydt

# ======== CASO 1: SUPERCRÍTICO (sin detener) ========
tspan_sup = (0.0, 0.7)      # [s]
rho_sup = 0.008             # 800 pcm
rho_ext_fun_sup = lambda t: rho_sup

y0_sup = np.concatenate(([n0],
                         (beta_i * n0 / (l * lambda_i)).flatten(),
                         [T0]))

sol_sup = solve_ivp(fun=lambda t, y: kinetics_thermal(t, y, rho_ext_fun_sup,
                                                      beta_i, lambda_i, beta, l,
                                                      0.0, Wth, Mcp, UA, Tcool, T0),
                    t_span=tspan_sup, y0=y0_sup, method='BDF',
                    rtol=rtol, atol=atol, dense_output=False)

tS = sol_sup.t
yS = sol_sup.y.T
P_S = yS[:,0]
T_S = yS[:,7]

# Energía y costo acumulados
E_S_MWh_t = cumulative_trapezoid(P_S, tS, initial=0.0) * MW_e / 3600.0
Cost_S_t = E_S_MWh_t * 1000.0 * COP_kWh

# Índice de cruce del umbral WARNING
idxW = np.argmax(P_S >= n_limit) if np.any(P_S >= n_limit) else None
if idxW is not None and not (P_S[idxW] >= n_limit):
    idxW = None

# ======== CASO 2: NORMAL (delayed-supercritical, controlable) ========
tspan_norm = (0.0, 3600.0)   # 1 hora
rho_norm = 0.001             # 100 pcm
rho_ext_fun_norm = lambda t: rho_norm

y0_norm = y0_sup.copy()
sol_norm = solve_ivp(fun=lambda t, y: kinetics_thermal(t, y, rho_ext_fun_norm,
                                                       beta_i, lambda_i, beta, l,
                                                       alpha_comb, Wth, Mcp, UA, Tcool, T0),
                     t_span=tspan_norm, y0=y0_norm, method='BDF',
                     rtol=rtol, atol=atol)

tN = sol_norm.t
yN = sol_norm.y.T
P_N = yN[:,0]
T_N = yN[:,7]
E_N_MWh_t = cumulative_trapezoid(P_N, tN, initial=0.0) * MW_e / 3600.0
Cost_N_t = E_N_MWh_t * 1000.0 * COP_kWh

# ======== CASO 3: SUBCRÍTICO ========
tspan_sub = (0.0, 600.0)      # 10 minutos
rho_sub = -0.002
rho_ext_fun_sub = lambda t: rho_sub

y0_sub = y0_sup.copy()
sol_sub = solve_ivp(fun=lambda t, y: kinetics_thermal(t, y, rho_ext_fun_sub,
                                                      beta_i, lambda_i, beta, l,
                                                      alpha_comb, Wth, Mcp, UA, Tcool, T0),
                    t_span=tspan_sub, y0=y0_sub, method='BDF',
                    rtol=rtol, atol=atol)

tB = sol_sub.t
yB = sol_sub.y.T
P_B = yB[:,0]
T_B = yB[:,7]
E_B_MWh_t = cumulative_trapezoid(P_B, tB, initial=0.0) * MW_e / 3600.0
Cost_B_t = E_B_MWh_t * 1000.0 * COP_kWh

# ======== Resumen numérico ========
dur_S = tS[-1] - tS[0]
Pavg_S = E_S_MWh_t[-1] / (dur_S/3600.0) if dur_S > 0 else np.nan
dur_N = tN[-1] - tN[0]
Pavg_N = E_N_MWh_t[-1] / (dur_N/3600.0) if dur_N > 0 else np.nan
dur_B = tB[-1] - tB[0]
Pavg_B = E_B_MWh_t[-1] / (dur_B/3600.0) if dur_B > 0 else np.nan

print("\n=== RESUMEN ===")
print(f"SUPERCRITICO:  E = {E_S_MWh_t[-1]:.3f} MWh | Pprom = {Pavg_S:.1f} MWe | dur = {dur_S:.2f} s")
print(f"NORMAL      :  E = {E_N_MWh_t[-1]:.2f}  MWh | Pprom = {Pavg_N:.1f} MWe | dur = {dur_N:.1f} s | ΔT={T_N[-1]-T0:.1f} K")
print(f"SUBCRITICO  :  E = {E_B_MWh_t[-1]:.2f}  MWh | Pprom = {Pavg_B:.1f} MWe | dur = {dur_B:.1f} s | ΔT={T_B[-1]-T0:.1f} K\n")

# ======== Gráficas ========
plt.rcParams.update({'figure.max_open_warning': 0})

# ---------- SUPERCRÍTICO ----------
fig1 = plt.figure(num='Supercritico - Potencia'); plt.clf(); ax1 = fig1.gca(); ax1.grid(True)
if idxW is None:
    ax1.plot(tS, P_S, 'r-', linewidth=2)
else:
    ax1.plot(tS[:idxW+1], P_S[:idxW+1], 'r-', linewidth=2)
    ax1.plot(tS[idxW:], P_S[idxW:], 'm--', linewidth=2)
    yMax = P_S.max()*1.05
    ax1.fill_between(tS[idxW:], 0, yMax, facecolor=(1.0,0.8,0.9), alpha=0.35)
    ax1.axhline(n_limit, color='k', linestyle=':', label='Límite WARNING')
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('Potencia normalizada')
ax1.set_title('SUPERCRÍTICO: ρ > β (ej. +800 pcm). ¡Debe controlarse!')

fig2 = plt.figure(num='Supercritico - Energia'); plt.clf(); ax2 = fig2.gca(); ax2.grid(True)
ax2.plot(tS, E_S_MWh_t, 'r-', linewidth=2)
if idxW is not None:
    ax2.plot(tS[idxW:], E_S_MWh_t[idxW:], 'm--', linewidth=2)
ax2.set_xlabel('Tiempo (s)')
ax2.set_ylabel('Q(t) [MWh]')
ax2.set_title('Energía acumulada (supercrítico)')

fig3 = plt.figure(num='Supercritico - Costo'); plt.clf(); ax3 = fig3.gca(); ax3.grid(True)
ax3.plot(tS, Cost_S_t/1e6, 'r-', linewidth=2)
if idxW is not None:
    ax3.plot(tS[idxW:], (Cost_S_t[idxW:])/1e6, 'm--', linewidth=2)
ax3.set_xlabel('Tiempo (s)')
ax3.set_ylabel('Costo acumulado [Millones COP]')
ax3.set_title(f'Costo (tarifa = {COP_kWh} COP/kWh) - supercrítico')

# ---------- NORMAL ----------
fig4 = plt.figure(num='Normal - Potencia'); plt.clf(); ax4 = fig4.gca(); ax4.grid(True)
ax4.plot(tN/60.0, P_N, 'b-', linewidth=2)
ax4.set_xlabel('Tiempo (min)')
ax4.set_ylabel('Potencia normalizada')
ax4.set_title('NORMAL: ρ < β (ej. +100 pcm) con Doppler (estable)')

fig5 = plt.figure(num='Normal - Energia'); plt.clf(); ax5 = fig5.gca(); ax5.grid(True)
ax5.plot(tN/60.0, E_N_MWh_t, 'b-', linewidth=2)
ax5.set_xlabel('Tiempo (min)')
ax5.set_ylabel('Q(t) [MWh]')
ax5.set_title('Energía acumulada (normal)')

fig6 = plt.figure(num='Normal - Costo'); plt.clf(); ax6 = fig6.gca(); ax6.grid(True)
ax6.plot(tN/60.0, Cost_N_t/1e6, 'b-', linewidth=2)
ax6.set_xlabel('Tiempo (min)')
ax6.set_ylabel('Costo acumulado [Millones COP]')
ax6.set_title(f'Costo (tarifa = {COP_kWh} COP/kWh) - normal')

# ---------- SUBCRÍTICO ----------
fig7 = plt.figure(num='Subcritico - Potencia'); plt.clf(); ax7 = fig7.gca(); ax7.grid(True)
ax7.plot(tB/60.0, P_B, 'k-', linewidth=2)
ax7.set_xlabel('Tiempo (min)')
ax7.set_ylabel('Potencia normalizada')
ax7.set_title('SUBCRÍTICO: ρ < 0 (ej. -200 pcm), potencia decae')

fig8 = plt.figure(num='Subcritico - Energia'); plt.clf(); ax8 = fig8.gca(); ax8.grid(True)
ax8.plot(tB/60.0, E_B_MWh_t, 'k-', linewidth=2)
ax8.set_xlabel('Tiempo (min)')
ax8.set_ylabel('Q(t) [MWh]')
ax8.set_title('Energía acumulada (subcrítico)')

fig9 = plt.figure(num='Subcritico - Costo'); plt.clf(); ax9 = fig9.gca(); ax9.grid(True)
ax9.plot(tB/60.0, Cost_B_t/1e6, 'k-', linewidth=2)
ax9.set_xlabel('Tiempo (min)')
ax9.set_ylabel('Costo acumulado [Millones COP]')
ax9.set_title(f'Costo (tarifa = {COP_kWh} COP/kWh) - subcrítico')

plt.show()
```

# 4. Fusión Nuclear

La fusión nuclear es un proceso en el que dos núcleos atómicos ligeros se combinan para formar un núcleo más pesado, liberando una considerable cantidad de energía
[13][17]. Esta energía surge porque el núcleo resultante tiene menor masa que los núcleos que lo forman; esa “masa perdida” se transforma en energía según la ecuación de Einstein: $E = \Delta m c^2$[13].

La fusión es la fuente de energía del Sol y de las estrellas: en sus núcleos, los núcleos de hidrógeno se fusionan en helio y liberan energía que llega a la Tierra como luz y calor [14].

En cambio, la fisión nuclear consiste en dividir un núcleo pesado (como el de uranio o plutonio) en dos núcleos más ligeros [17]. El proceso también libera energía pero es esencialmente opuesto: en fisión dividimos, y en fusión unimos [14].

En ambos casos, se libera energía debido a la llamada energía de enlace nuclear, los nucleones (protones y neutrones) liberan energía cuando se acomodan en configuraciones más estables [14]. En núcleos más ligeros que el hierro, fusionar libera energía; en núcleos más pesados que el hierro, romperlos libera energía


Las diferencias clave entre ambos procesos son:


| Característica | Fusión Nuclear | Fisión Nuclear |
| :--- | :--- | :--- |
| **Proceso Principal** | Une núcleos ligeros para liberar energía. | Rompe núcleos pesados para liberar energía. |
| **Residuos Radiactivos** | Gases ligeros (helio) y materiales activados por neutrones de menor persistencia. No genera residuos de larga vida como los de fisión. | Genera subproductos radiactivos de larga vida (actínidos) que requieren gestión y almacenamiento prolongado. |
| **Seguridad y Control** | Intrínsecamente segura. La reacción se apaga si el confinamiento falla. No es una reacción en cadena autosostenida. | Requiere sistemas de control muy estrictos para gestionar la reacción en cadena autosostenida. |
| **Combustible** | Isótopos de hidrógeno (deuterio y tritio). El deuterio es abundante en el agua de mar. | Uranio o plutonio, que requieren procesos de enriquecimiento complejos. |





La fusión nuclear, además de ser un fenómeno físico fundamental, está íntimamente ligada a la historia misma del universo[5]. Desde el instante del Big Bang hasta los experimentos actuales en reactores como ITER [8], la fusión ha desempeñado un papel esencial en la evolución cósmica y en el desarrollo científico y tecnológico de la humanidad. ### 4.1 Reacción más prometedora (Composcición del combustible)

La reacción de fusión más estudiada y considerada como la más prometedora para aplicaciones energéticas en la Tierra es la que involucra a dos isótopos del hidrógeno: el deuterio ($^2$H) y el tritio ($^3$H) [13][15]. Esta elección no es casual, pues aunque en teoría existen muchas combinaciones posibles de núcleos ligeros que podrían fusionarse, la deuterio-tritio es la que requiere las condiciones menos extremas de temperatura y presión para superar la repulsión electrostática entre los núcleos (la llamada barrera de Coulomb) [13]. Mientras otras reacciones de fusión, como la deuterio-deuterio o la protón-boro, necesitan temperaturas del orden de cientos de millones de grados, la D-T puede producirse de manera más “accesible” a unos 100 millones de grados, lo cual la hace la opción más viable en los experimentos actuales.

La reacción puede escribirse de la siguiente forma [13] [17]:

$$^2_1H + ^3_1H \; \rightarrow \; ^4_2He \; (3.5 \; MeV) + n \; (14.1 \; MeV)$$

$$D + T \to {}^4\mathrm{He} + n + 17.6\ \mathrm{MeV}$$


```python
# Diagrama Simplificado de Reacción de Fusión (Deuterio-Tritio)

import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_nucleus(ax, x, y, protons, neutrons, label="", color_protons='red', color_neutrons='blue', size_factor=1.0):
    """Dibuja un núcleo simplificado con protones y neutrones."""
    radius = 0.15 * size_factor

    # Dibuja los neutrones
    for i in range(neutrons):
        nx = x + radius * 0.8 * (i - (neutrons - 1) / 2) if neutrons > 0 else x
        ny = y
        if neutrons > 1:
             nx = x + radius * 0.5 * (i % 2 * 2 - 1) * size_factor
             ny = y + radius * 0.5 * (i // 2 * 2 - 1) * size_factor if neutrons > 1 else y

        circle_n = patches.Circle((nx, ny), radius, color=color_neutrons, zorder=2)
        ax.add_patch(circle_n)

    # Dibuja los protones
    for i in range(protons):
        px = x + radius * 0.8 * (i - (protons - 1) / 2) if protons > 0 else x
        py = y + radius * 1.0 * size_factor
        if protons > 1:
            px = x + radius * 0.5 * (i % 2 * 2 - 1) * size_factor
            py = y + radius * 0.5 * (i // 2 * 2 - 1) * size_factor + (radius * 0.5 * size_factor if protons == 1 else 0)

        circle_p = patches.Circle((px, py), radius, color=color_protons, zorder=3)
        ax.add_patch(circle_p)
        ax.text(px, py, '+', color='white', ha='center', va='center', fontsize=8 * size_factor, fontweight='bold')

    # Añadir etiqueta principal
    if label:
        ax.text(x, y - radius * 2, label, ha='center', va='top', fontsize=12 * size_factor, fontweight='bold', color='darkblue')

def draw_single_particle(ax, x, y, particle_type, label="", color='blue', size_factor=1.0):
    """Dibuja una partícula individual (ej. neutrón)."""
    radius = 0.15 * size_factor
    circle = patches.Circle((x, y), radius, color=color, zorder=2)
    ax.add_patch(circle)
    if particle_type == 'proton':
        ax.text(x, y, '+', color='white', ha='center', va='center', fontsize=8 * size_factor, fontweight='bold')
    if label:
        ax.text(x, y - radius * 2, label, ha='center', va='top', fontsize=12 * size_factor, fontweight='bold', color='darkblue')

fig, ax = plt.subplots(figsize=(10, 8))
ax.set_aspect('equal', adjustable='box')
ax.set_xlim(-5, 5)
ax.set_ylim(-3.5, 3.5)
ax.axis('off')

# Factor para hacer las moléculas más grandes
size_factor = 1.5

# --- Reactantes (Deuterio y Tritio) ---
# Deuterio (D): 1 protón, 1 neutrón
draw_nucleus(ax, -3, 2, protons=1, neutrons=1, label="Deuterium\nD", size_factor=size_factor)
# Tritio (T): 1 protón, 2 neutrones
draw_nucleus(ax, 3, 2, protons=1, neutrons=2, label="Tritium\nT", size_factor=size_factor)

# Flechas hacia la reacción
ax.annotate('', xy=(-0.5, 0.5), xytext=(-2.5, 1.5), arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8))
ax.annotate('', xy=(0.5, 0.5), xytext=(2.5, 1.5), arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8))

# --- Reacción de Fusión (estado intermedio simplificado) ---
draw_nucleus(ax, 0, 0, protons=2, neutrons=3, label="Fusion Reaction", color_protons='red', color_neutrons='blue', size_factor=size_factor)

# Flechas desde la reacción
ax.annotate('', xy=(-2, -2.5), xytext=(-0.5, -0.5), arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8))
ax.annotate('', xy=(2, -2.5), xytext=(0.5, -0.5), arrowprops=dict(facecolor='black', shrink=0.05, width=1.5, headwidth=8))

# --- Productos (Helio y Neutrón) ---
# Helio (He4): 2 protones, 2 neutrones
draw_nucleus(ax, -2.5, -2.5, protons=2, neutrons=2, label="Helium\nHe$^4$", size_factor=size_factor)
# Neutrón (n): 1 neutrón
draw_single_particle(ax, 2.5, -2.5, particle_type='neutron', label="Neutron\nn", color='blue', size_factor=size_factor)

plt.title("Diagrama Simplificado de Reacción de Fusión (Deuterio-Tritio)", fontsize=16, pad=20)
plt.show()
```

Es decir, al fusionarse un núcleo de deuterio y uno de tritio se forma un núcleo de helio-4 (también llamado partícula alfa) y un neutrón libre de alta energía. El helio queda confinado en el plasma y contribuye a mantener la temperatura del sistema, mientras que el neutrón, al no tener carga eléctrica, escapa y transporta la mayor parte de la energía liberada[17]. En total, cada reacción entrega alrededor de 17.6 MeV, lo cual es unas tres o cuatro veces más energía por unidad de combustible que la fisión nuclear del uranio[17].

La razón de que esta reacción libere tanta energía se encuentra en la física de la energía de enlace nuclear. Cada nucleón (protón o neutrón) en un núcleo atómico está sujeto a una fuerza llamada interacción nuclear fuerte, que tiende a mantenerlos unidos [16]. En los núcleos ligeros, como los del hidrógeno, la energía de enlace por nucleón es baja, mientras que en el helio-4 es mucho mayor. Por eso, cuando los núcleos ligeros se fusionan y forman helio, los nucleones pasan a un estado más estable, liberando el exceso de energía en forma de partículas rápidas y radiación. Dicho de otro modo, la fusión ocurre porque el helio es un “punto de equilibrio” más estable que los núcleos iniciales, y la diferencia en energía de enlace es lo que se transforma en energía útil [13].


Las siguientes ecuaciones representan otras de las principales reacciones de fusión que ocurren con el deuterio (D) y el tritio (T) [15]:

$$D + D \to {^3He} + n + 3.2\ MeV$$

$$D + D \to T + p + 4.0\ MeV$$

$$D + T \to {^4He} + n + 17.6\ MeV$$

$$D + {^3He} \to {^4He} + p + 18.5\ MeV$$

Aunque existen otras posibles reacciones de fusión, como las deuterio-deuterio (D+D), que liberan entre 3 y 4 MeV, o la deuterio-helio-3 (D+³He), que libera 18,5 MeV, todas ellas requieren temperaturas mucho más altas o presentan menores probabilidades de ocurrir [17]. Por esta razón, la reacción deuterio-tritio sigue siendo la más estudiada y viable en la actualidad. Además, en los experimentos se utiliza una proporción cercana a 50:50 entre deuterio y tritio, de modo que se maximice la frecuencia de colisiones D–T y se reduzcan las interacciones menos eficientes (como D–D), garantizando así la mayor producción de energía posible en el plasma [15].

**Características del combustible nuclear de fusión D–T**

| Propiedad | Valor representativo |
| :--- | :--- |
| Combustible principal | Deuterio ($^2$H) y Tritio ($^3$H) |
| Proporción isotópica | 50 % Deuterio – 50 % Tritio |
| Densidad típica del plasma | $\sim 10^{20}\ \mathrm{partículas/m^3}$ |
| Temperatura de operación | $\sim 10^8\ \mathrm{K}$ (≈ 10 keV) |
| Reacción principal | $D + T \rightarrow \ ^4He\ (3.5\ \mathrm{MeV}) + n\ (14.1\ \mathrm{MeV})$ |

```python
import numpy as np

# === Parámetros físicos de fusión D-T ===
Q_MeV = 17.6                             # Energía por reacción D-T [MeV]
J_per_MeV = 1.602176634e-13
Q_J = Q_MeV * J_per_MeV                  # J/reacción

# Densidades de combustible
n_D = 1e20                                # deuterio [partículas/m³]
n_T = 1e20                                # tritio [partículas/m³]

# Masa molar
M_D = 2.014                               # g/mol
M_T = 3.016                               # g/mol
NA = 6.02214076e23

m_D_kg = M_D / 1000 / NA                   # kg/partícula
m_T_kg = M_T / 1000 / NA                   # kg/partícula

# === Parámetros del reactor ===
MWt = 3000                                 # Potencia térmica [MWt]
Pth_W = MWt * 1e6                          # Potencia térmica [W]

# === Tasas de reacción ===
# Número de reacciones por segundo necesario para generar Pth_W
R_reac = Pth_W / Q_J                       # reacciones/s

# Consumo de combustible
burn_D_kg_s = R_reac * m_D_kg
burn_T_kg_s = R_reac * m_T_kg

burn_D_kg_h = burn_D_kg_s * 3600
burn_T_kg_h = burn_T_kg_s * 3600

burn_D_kg_d = burn_D_kg_s * 86400
burn_T_kg_d = burn_T_kg_s * 86400

# === Resultados ===
print("=== Magnitudes derivadas (fusión D-T, n=1) ===")
print(f"Potencia térmica:        {MWt:,.0f} MWt")
print(f"Energía por reacción:    {Q_J:.3e} J")
print(f"Reacciones por segundo:  {R_reac:,.3e} 1/s")
print(f"Consumo Deuterio:        {burn_D_kg_h:,.3f} kg/h  (~{burn_D_kg_d:,.3f} kg/día)")
print(f"Consumo Tritio:          {burn_T_kg_h:,.3f} kg/h  (~{burn_T_kg_d:,.3f} kg/día)")
```

### 4.2 Cinética y energía de la combustión

Al igual que en el caso de la fisión nuclear, en procesos de fusión nuclear no se tienen ecuaciones químicas para la reacción sino ecuaciones nucleares que representan este proceso.

En terminos generales, la ecuación representativa de este proceso se puede escribir como:

$$D + T → α + n + 17.6 MeV$$ [15]

* Definiendo α como nucleos de helio (${}^{4}\mathrm{He}$)

Para enteder como se reparte la producción de energía en este proceso se hace un balance de energía y balance de momento para cada elemento generando un sistema de 2 ecuaciones 2 incognicas que se resuelve de la siguiente forma [15]:

$$\frac{1}{2} m_1 v_1^2 + \frac{1}{2} m_2 v_2^2 = E
$$

$$
m_1 v_1 + m_2 v_2 = 0
$$

Resolviendo para las energías cinéticas individuales de cada producto [15]:

$$
\frac{1}{2} m_1 v_1^2 = \frac{m_2}{m_1 + m_2} E
$$

$$
\frac{1}{2} m_2 v_2^2 = \frac{m_1}4{m_1 + m_2} E
$$

Reemplazando para el caso de reacción D-T se tiene la siguiente relación de masa [15]:

$$\frac{m_α}{m_n} = 4$$

$$m_α = 4m_n$$

$$E = 17.6MeV$$

$$\frac{E}{5} = 3.52 MeV$$

Así obtenemos la siguiente ecuación donde se muestra el aporte energético de cada elemento:

$$
D + T \;\;\rightarrow\;\; \alpha(3.5\ \text{MeV}) + n(14.1\ \text{MeV})
$$

De esto podemos concluir que la partícula con menor masa, en este caso el neutron, producirá más energía que la partícula α.

Esta reacción D-T es la que domina el campo de investigación de la energía nuclear de fusión.

**A continuación se muestra la conversión de unidades de la energía total generada en la reacción de fusión para que quede expresada en unidades de energía termica:**


$$
\begin{aligned}
E_f = 17.6\,\mathrm{MeV}\times 1.602\times 10^{-19}\,\frac{\mathrm{J}}{\mathrm{MeV}}
= 2.819\times 10^{-12}\ \mathrm{J}
\end{aligned}
$$

**Relación de energía producida con temperatura generada:**

Para hacer esta relación se utiliza la siguiente formula:

$$ E = kT$$
Donde:
- E = energía en Julios (J)
- T = Temperatura en Kelvin (K)
- k = constante de Boltzmann ( $1.380649\times10^{-23} \frac{J}{K}$)

Se despeja la temperatura y se realizan los calculos:

$$
T = \frac{E}{k}\\
T = \frac{2.819\times10^{-12}J}{1.380649\times10^{-23}\frac{J}{K}}\\
T ≈ 2.042\times10^{11} K
$$



Esta temperatura tan elevada es coherente con un proceso de fusión nuclear donde se libera mucha energía durante el proceso, similar a los procesos de fusión que se encuentran en el núcleo del sol.


**Calcúlo de la masa inicial:**

$$
Masa Deuterio (m_d)= 2.014 u \\
Masa Tritio (m_t)= 3.016 u \\
Masa_T = m_t + m_d = 5.030 u \\
1 \ \text{u} = 1.6605 \times 10^{-27} \ \text{kg}\\
m = 5.030 \times 1.6605 \times 10^{-27} \\
m = 8.35 \times 10^{-27} \ \text{kg}
$$

Energía expresada en $\frac{J}{Kg}$

$$
E_{kg} = \frac{2.819 \times 10^{-12} \ \text{J}}{8.35 \times 10^{-27} \ \text{kg}}
= 3.38 \times 10^{14} \ \text{J/kg}
E_{kg} ≈ 340 TJ
$$


### Balance de masa

A partir de la ecuación general de la fusión se puede realizar el balance de masa de este proceso.

$$^2_1H + ^3_1H \; \rightarrow \; ^4_2He+ n$$

* Balance de número de neutrones:
    $A: 2+3 = 4 +1 = 5$
* Balance de número atómico:
    $Z: 1+1 = 2+0 = 2$

### Tasa de la reacción

Para dos especies \(i\) y \(j\), la tasa de reacciones por unidad de volumen por segundo es [15]:

$$
R_{ij} = n_i n_j \langle \sigma v \rangle_{ij},     (\frac{m^3}{s})
$$

donde:
- $(n_i, n_j)$ = densidades de partículas (número de núcleos por unidad de volumen).  
- $(\sigma)$ = sección eficaz de fusión (probabilidad de que ocurra la reacción).  
- $(v)$ = velocidad relativa entre los núcleos.  
- $(\langle \sigma v \rangle)$ = **promedio Maxwelliano** de $(\sigma v)$.


---

#### Promedio Maxwelliano

En un plasma, cuando se tienen multiples iones, no se tiene una sola velocidad sino que cada ión tiene su propia velocidad por lo que se debe hacer el cálculo con una distribución de velocidades. Esta distribución se llama la **distribución de Maxwell-Boltzmann** [15].  
Por lo tanto, el valor relevante no es $(\sigma v)$ a una velocidad fija, sino su promedio:

$$
\langle \sigma v \rangle = \int_0^{\infty} \sigma(v) \, v \, f(v) \, dv
$$


En el caso de la fusión D-T la tasa de reacción se expresa de la siguiente forma:

$$
R_{DT} = n_D n_T \langle \sigma v \rangle_{DT}
$$

---
### Potencia generada
Para encontrar la potencia de fusión por unidad de volumen se utliza la ecuación de la tasa de reacción y se multiplica por la energía liberada por ecuación.

Cada reacción D–T libera $E_f = 17.6 \, \text{MeV}$ y la potencia de fusión por unidad de volumen es:

$$
P_{DT} = n_D n_T \langle \sigma v \rangle_{DT} E_f
$$

$$
P_{DT} = R_{DT}E_r
$$

Asumiendo que el plasma D-T tiene una proporción **50/50** en densidad iónica $(n_D = n_T = \tfrac{n}{2})$ la ecuación de potencia se define de la siguiente manera [15]:

$$
P_{DT} = \frac{n^2}{4} \langle \sigma v \rangle_{DT} \, E_f
$$

A partir de esta ecuación se mide cuanta energía útil se puede obtener a partir del plasma generado en el proceso de fusión nuclear D-T.

---
### Reactividad térmica ⟨σv⟩

La reactividad térmica se puede enteder como la probablidad de que los núcleos dentro del plasma colisionen terminando en una reacción de fusión nuclear [15]. Esta reactividad se expresa como el promedio Maxwelliano de la fusión.

$$
\langle\sigma v\rangle = \frac{2}{\sqrt{\pi m_r}\,(kT)^{3/2}}
\int_0^\infty \sigma(E_f)\, E_f\, e^{-E_f/(kT)}\, dE_f
$$

- El factor $ e^{-E_f/(kT)}$ indica la cantidad de partículas en torno a la energía térmica $KT$.
- $\sigma(E_f)$ es la secció eficaz de la reacción.
- La masa reducida $m_r$ relacióna la energia y la velocidad relativa.

Usualmente esta integral no se resuelve de manera analítica, se usan ajustes experimentales como el de Bosch-Hale para encontar este valor.

---

### Sección eficaz y factor Gamow

Es una medida de la probabilidad de que ocurra una interacción o reacción entre dos núcleos o partículas en el plasma, funcionando como un área efectiva para que ocurra un evento de fusión [15]. Se define como la siguiente ecuación:

$$
\sigma(E) = \frac{S(E_f)}{E_f}\, e^{-2\pi\eta(E_f)},
$$

- La sección eficaz se separa en una parte suave $\frac{S(E_f)}{E_f}$ y un factor de tunelamiento cuántico.
- $\frac{S(E_f)}{E_f}$ es la sección suave la cual no evidencia mucho cambio en función de la energía de reacción E_f.
- El término $(e^{-2\pi\eta})$ es el factor de tunelamiento cuántico, este expresa la probabilidad de que las partículas cargadas puedan atravesar la barrera de Coulomb.

El parámetro de Sommerfeld $(η(E_f))$ depende la la carga de los núcleos y su velociad. Esta definido con la siguiente ecuación:

$$\quad
\eta(E) = Z_1 Z_2 \frac{e^2}{\hbar}\sqrt{\frac{m_r}{2E}}$$

La barrera de Coulomb es una barrera energética causada por la repulsión electrostática entre dos partículas con carga eléctrica positiva, los núcleos deben superar esta barrera para acercarse lo suficiente y experimentar una reacción nuclear de fusión

---

### Pico de Gamow

Este factor representa la energía más probable de fusión, es decir la relación entre dos efectos:

- A bajas energias hay muchas partículas que no pueden penetrar la barrera de Coulomb
- A altas energías hay pocas partículas pero la probabilidad de dusión es muy alta.

$$
E_0 \approx \left[\left(\frac{\pi Z_1 Z_2 e^2}{\hbar}\right)^2 \frac{m_r (kT)^2}{2}\right]^{1/3}
$$

- Representa el balance entre muchas partículas de baja energía y pocas de alta energía.

Esto define la **ventana de Gamow**, un rango estrecho de niveles de energías donde occurren las reacción de fusión.

---

### Ecuaciones cinéticas de combustible

Estas ecuaciones describen como se comsumen el deuterio y tritio en función del tiempo [15].

$$
\frac{dn_D}{dt} = -n_D n_T \langle\sigma v\rangle, \quad
\frac{dn_T}{dt} = -n_D n_T \langle\sigma v\rangle
$$

Este modelo es simplificado ya que las plantas nucleares de fusión aun se encuentarn en una etapa experimental pero en un reactor nuclear se deben tener encuenta factores como la iyección de más combustible y perdidas por fugas de transporte o recombinación de las moléculas.


---

### Fracción de quemado

Este factor mide que fracción del combustible inicial realmente llega a fusionarse en un tiempo de comfinamiento $τ$.

$$
f \approx \frac{n_0 \langle\sigma v\rangle \tau}{1 + n_0 \langle\sigma v\rangle \tau}
$$

- Donde $n_0$ es la cantidad incial de combustible.

Este factor se puede entender como la eficiencia real de la reacción de fusión [15].

- Si $n_0 \langle\sigma v\rangle \tau << 1$, la fracción del combustible que llega a fusionarse es muy pequeña, por ende la eficiencia será muy baja.
- Si $n_0 \langle\sigma v\rangle \tau >> 1$, la fracción de combustible que llega a fusionarse tiene a 1, por ende la eficiencia sería muy alta.

Con las ecuaciones mostradas anteriormente podemos entender que para lograr un proceso de fusión eficiente se debe tener una densidad atómica alta, una temperatura ($T$) suficiente y un buen tiempo de confinamiento ($\tau$).

---

### Criterio de Lawson

Este criterio establece el umbral mínimo de condiciones que debe cumplir el plasma para que este llegue a fusionarse y que esta energía de fusión compense todas las pérdidas que se puedan generar en el sistema.

$$
n T \tau_E \gtrsim \text{constante}
$$

Para la fusión D–T, este criterio tipicamente es:

$$
n T \tau_E \gtrsim 10^{21}\ \mathrm{keV \cdot s \cdot m^{-3}}
$$

---
### Pérdidas radiativas

Además de las pérdidas por transporte (difusión o fugas magnéticas) el plasma también pierde energía por **radiación de bremsstrahlung** [15]. Tiene un efecto similar a el calentamiento de partículas $α$ que también genera pérdidas.


La radiación de bremsstrahlung es una radiación electromagnética (generalmente rayos X) que se produce cuando una partícula cargada (como un electrón) es desacelerada rápidamente al interactuar con el campo eléctrico de otro cuerpo cargado, como un núcleo atómicos, en este caso los electrones acelerados por los iones emiten fotones generando una pérdida de energía cinética. Estas pérdidas se definen con la siguiente ecuación:

$$
P_{\rm brem} \sim C\, n_e n_i Z_{\rm eff} \sqrt{T_e}
$$

Donde:
- C = constante de proporcionalidad, para plsma D-T usualmente $C = 1.69 \times 10^{-38} W * m^3*keV^{-1/2}.$
- $n_e$ = densidad electrónica del plasma.
- $Z_{eff}$ = número atómico effectivo, en el caso de fusión D-T $Z_{eff} = 1$.
- $T_e$ = temperatura electrónica (keV)

### Energía liberada

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
Q = 17.6 * 1.602e-13  # Joules por reacción
T_keV = 15
sigma_v = 1.1e-24 * T_keV**2 * np.exp(-19.94 / T_keV)

# Densidades iniciales
n_D0 = 1e20
n_T0 = 1e20

# Tiempo (s)
t = np.linspace(0, 5000, 500)

# Inicializamos vectores
n_D = np.zeros_like(t)
n_T = np.zeros_like(t)
E_acumulada = np.zeros_like(t)

# Condiciones iniciales
n_D[0] = n_D0
n_T[0] = n_T0
E_acumulada[0] = 0

# Paso de integración
dt = t[1] - t[0]

# Simulación por método de Euler
for i in range(1, len(t)):
    # Cambio de densidad por reacción
    dn = n_D[i-1]*n_T[i-1]*sigma_v*dt
    n_D[i] = n_D[i-1] - dn
    n_T[i] = n_T[i-1] - dn
    # Energía acumulada
    E_acumulada[i] = E_acumulada[i-1] + Q*dn

plt.figure(figsize=(8,5))
plt.plot(t, E_acumulada, color='purple', linewidth=2)
plt.title("Energía liberada acumulada vs Tiempo (realista, D-T)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Energía acumulada [J/m³]")
plt.grid(True)
plt.show()
```

```python
# Gráficas interactivas con curva dinámica centrada en T
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Función para calcular sigma_v
def sigma_v_DT(T_keV):
    return 1.1e-24 * T_keV**2 * np.exp(-19.94 / T_keV)

# Función interactiva
def energia_acumulada(n_D0, n_T0, T_keV):
    Q = 17.6 * 1.602e-13  # Joules por reacción
    sigma_v = sigma_v_DT(T_keV)

    # Tiempo para energía acumulada
    t = np.linspace(0, 5000, 500)
    n_D = np.zeros_like(t)
    n_T = np.zeros_like(t)
    E_acumulada = np.zeros_like(t)

    n_D[0] = n_D0
    n_T[0] = n_T0
    dt = t[1] - t[0]

    # Simulación de consumo de combustible
    for i in range(1, len(t)):
        dn = n_D[i-1]*n_T[i-1]*sigma_v*dt
        n_D[i] = n_D[i-1] - dn
        n_T[i] = n_T[i-1] - dn
        E_acumulada[i] = E_acumulada[i-1] + Q*dn

    # Crear figura
    plt.figure(figsize=(18,5))

    # 1. Tasa de reacción centrada en T_keV
    plt.subplot(1,3,1)
    # Rango dinámico centrado en T_keV
    T_min = max(1, T_keV - 5)
    T_max = min(30, T_keV + 5)
    T_plot = np.linspace(T_min, T_max, 200)
    plt.plot(T_plot, sigma_v_DT(T_plot), color='red', lw=2)
    plt.scatter([T_keV], [sigma_v], color='orange', zorder=5, label=f'T_sel={T_keV} keV')
    plt.title("<σv> vs Temperatura (centrada en T seleccionada)")
    plt.xlabel("T [keV]")
    plt.ylabel("<σv> [m³/s]")
    plt.grid(True)
    plt.legend()

    # 2. Potencia por volumen vs temperatura (curva continua)
    plt.subplot(1,3,2)
    T_curve = np.linspace(1, 30, 300)
    P_f_curve = n_D0 * n_T0 * sigma_v_DT(T_curve) * Q
    plt.plot(T_curve, P_f_curve, color='blue', lw=2)
    plt.scatter(T_keV, n_D0*n_T0*sigma_v*Q, color='orange', zorder=5, label=f'T_sel={T_keV} keV')
    plt.title("Potencia por volumen vs Temperatura")
    plt.xlabel("T [keV]")
    plt.ylabel("P_f [W/m³]")
    plt.yscale('log')
    plt.grid(True, which='both', ls='--')
    plt.legend()

    # 3. Energía acumulada vs tiempo
    plt.subplot(1,3,3)
    plt.plot(t, E_acumulada, color='purple', lw=2)
    plt.title("Energía acumulada vs Tiempo")
    plt.xlabel("Tiempo [s]")
    plt.ylabel("E [J/m³]")
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Sliders interactivos
interact(energia_acumulada,
         n_D0=(1e19, 5e20, 1e19),
         n_T0=(1e19, 5e20, 1e19),
         T_keV=(1, 30, 1));
```

## 6.1 Estadistica mecánica del sistema

El estudio de la reacción de fusión entre deuterio (D) y tritio (T) mediante un enfoque de mecánica estadística permite comprender cómo las propiedades microscópicas del plasma determinan la potencia térmica generada y la eficiencia del proceso de fusión. En este contexto, la tasa de reacciones, la reactividad nuclear y la probabilidad de fusión dependen directamente de la distribución energética de las partículas, descrita por la estadística de Maxwell-Boltzmann.

La potencia térmica generada en el plasma se obtiene al multiplicar la tasa de reacciones por la energía liberada en cada evento de fusión. Esta potencia depende de la densidad de los elementos usados y del promedio Maxwelliano de la sección eficaz y la velocidad relativa, denotado como ⟨σv⟩. Este término representa el promedio ponderado de las velocidades de colisión, teniendo en cuenta que no todas las partículas del plasma tienen la misma energía cinética.

El cálculo de ⟨σv⟩ requiere integrar la sección eficaz σ(E), la velocidad relativa v(E) y la distribución de energías f(E) del plasma. La sección eficaz expresa la probabilidad de que una colisión entre dos núcleos resulte en una reacción de fusión, y varía fuertemente con la energía de colisión. A bajas energías, la repulsión coulombiana entre los núcleos reduce la probabilidad de fusión, mientras que a energías demasiado altas, la sección eficaz disminuye debido a efectos cuánticos y de dispersión nuclear.

Un elemento clave para comprender este comportamiento es el pico de Gamow, el cual representa el rango de energías donde la probabilidad de fusión es máxima. Este pico surge de la competencia entre dos factores opuestos:

La disminución exponencial de la probabilidad de penetración de la barrera coulombiana (factor de Gamow, $e^{-\sqrt{E_G/E}}$), y

El incremento exponencial de la fracción de partículas con suficiente energía térmica, dada por la distribución de Maxwell-Boltzmann ($e^{-E/kT}$).

El punto donde ambos efectos se equilibran define el rango óptimo de energías de fusión, el cual se traduce en la temperatura óptima del plasma (alrededor de 10–20 keV para la reacción D–T). En este rango, la reactividad alcanza su máximo valor, lo que maximiza la potencia térmica producida.

En conjunto, estos factores muestran cómo la fusión controlada depende de una combinación precisa entre temperatura, densidad y confinamiento del plasma. La mecánica estadística permite vincular el comportamiento probabilístico de los núcleos individuales con magnitudes macroscópicas como la potencia y la eficiencia del reactor. El análisis de la sección eficaz, el factor de Gamow y el promedio Maxwelliano revela que existe una ventana energética estrecha donde las condiciones del plasma permiten una fusión eficiente, base fundamental del diseño de reactores como ITER o DEMO.


## 6.2 Potencia de fusión y reactividad nuclear

A continuación se calcula la potencia térmica generada en la reacción y la reactividad nuclear de las particulas de fusión en función de la temperatura del plasma. Este valor, conocido como promedio Maxwelliano de la sección eficaz ($\sigma$) y la velocidad relativa ($v$), es fundamental para determinar la tasa de reacciones de fusión y, por ende, la potencia generada en un reactor de fusión.

En un plasma de fusión, las partículas no tienen una velocidad única, sino que siguen una distribución de velocidades. A temperaturas relevantes para la fusión, esta distribución se aproxima a la de Maxwell-Boltzmann. La probabilidad de que ocurra una reacción de fusión depende de la energía y por lo tanto, de la velocidad de colisión entre los núcleos. Sin embargo, no todas las colisiones resultan en fusión, incluso si la energía es alta. La sección eficaz ($\sigma$) representa la "probabilidad" de fusión para una energía de colisión dada.

El promedio Maxwelliano ($\langle \sigma v \rangle$) integra la sección eficaz sobre la distribución de velocidades de Maxwell-Boltzmann. Esto significa que nos da la tasa de reacción promedio esperada para un plasma a una temperatura dada, teniendo en cuenta que no todas las partículas se mueven a la misma velocidad. Es, en esencia, una medida de cuán "reactivo" es el plasma a una temperatura específica.

Variables para el cálculo:

1.  **Constantes físicas:** Se definen constantes fundamentales como la constante de Boltzmann ($k_B$), la carga elemental ($e$), la constante de Planck reducida ($\hbar$), y la unidad de masa atómica ($m_u$).
2.  **Masas y masa reducida:** Se calculan las masas del deuterio y el tritio, y a partir de ellas, la masa reducida del sistema D-T. La masa reducida es importante en la física de colisiones.
3.  **Constante de Gamow:** Se calcula una constante ($b_{gamow}$) relacionada con la barrera de Coulomb que los núcleos deben superar para fusionarse.
4.  **Sección eficaz ($\sigma$):** aproxima la sección eficaz de la reacción D-T en función de la energía de colisión. Aunque no es una función exacta (como la de Bosch-Hale), captura la forma general del pico de sección eficaz.
5.  **Promedio Maxwelliano integrado:** La integral promedia el producto de $\sigma(E)$, la velocidad relativa $v$ (relacionada con $\sqrt{E}$), y la distribución de Maxwell-Boltzmann en energía $f(E)$, sobre todas las posibles energías de colisión. La forma de la integral es:

    $$
    \langle \sigma v \rangle = \int_0^{\infty} \sigma(E) \, v(E) \, f(E) \, dE
    $$

    donde $f(E)$ es la distribución de Maxwell-Boltzmann y $v(E)$ es la velocidad relativa a energía $E$. El factor de normalización y la forma del integrando se derivan de la estadística de Maxwell-Boltzmann y la cinemática de colisiones.


```python
!pip install ipywidgets plotly --quiet
from google.colab import output
output.enable_custom_widget_manager()

# ------------------------------
# Interactivo D–T con 3 sliders
# Funciona en Google Colab (usa FigureWidget para actualización instantánea)
# ------------------------------

import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import ipywidgets as widgets
from IPython.display import display

# ------------------------------
# Constantes
# ------------------------------
keV_to_J = 1.602176634e-16
E_f_keV = 17.6e3      # keV por reacción
E_f_J = E_f_keV * keV_to_J

# ------------------------------
# Reactividad D–T (Bosch & Hale simplificada)
# ------------------------------
def sigma_v_DT(T):
    """
    Retorna <σv> en m^3/s para T en keV.
    Fórmula empírica (Bosch & Hale style, simplificada).
    """
    a = [6.927e4, 7.454e8, 2.050e6, 5.200e4, 0.0, 0.0]
    T = np.array(T)
    S = a[0] + T*(a[1] + T*(a[2] + T*(a[3] + T*(a[4] + T*a[5]))))
    E_G = 986.0  # keV
    return 1e-24 * (S / (T**(2/3))) * np.exp(-3.0 * (E_G / (4.0 * T))**(1/3))

# ------------------------------
# Datos base
# ------------------------------
T_range = np.linspace(1, 30, 400)
reactividad_base = sigma_v_DT(T_range)

# Valores iniciales
T_init = 15.0
nD_init = 1e20
nT_init = 1e20

# ------------------------------
# Crear FigureWidget con 2 paneles:
#  - izquierda: ⟨σv⟩ vs T (log)
#  - derecha: Potencia volumétrica P(T) = nD*nT*⟨σv⟩*E_f [W/m³]
# ------------------------------
fig = make_subplots(rows=1, cols=2, subplot_titles=("⟨σv⟩ vs T (D–T)", "Potencia volumétrica P vs T"))

# Curva reactividad (izq)
trace_reac = go.Scatter(x=T_range, y=reactividad_base, mode='lines', name='⟨σv⟩ (m³/s)')
# Punto seleccionado (izq)
ov_init = sigma_v_DT(T_init)
trace_point = go.Scatter(x=[T_init], y=[ov_init], mode='markers', marker=dict(size=10, color='red'), name='T seleccionada')

# Curva potencia (der)
P_base = nD_init * nT_init * reactividad_base * E_f_J
trace_P = go.Scatter(x=T_range, y=P_base, mode='lines', name='P [W/m³]')

fig.add_trace(trace_reac, row=1, col=1)
fig.add_trace(trace_point, row=1, col=1)
fig.add_trace(trace_P, row=1, col=2)

# Ejes y layout
fig.update_yaxes(type='log', row=1, col=1, title_text='⟨σv⟩ [m³/s]')
fig.update_xaxes(title_text='T [keV]', row=1, col=1)
fig.update_yaxes(title_text='P [W/m³]', row=1, col=2)
fig.update_xaxes(title_text='T [keV]', row=1, col=2)
fig.update_layout(height=520, width=1000, template='plotly_white', title_text="Interacción: Temperatura, nD y nT")

# Convertir en FigureWidget para actualizaciones rápidas
figw = go.FigureWidget(fig)

# ------------------------------
# Widgets (sliders)
# ------------------------------
slider_T = widgets.FloatSlider(value=T_init, min=1.0, max=30.0, step=0.1, description='T [keV]', continuous_update=True, readout_format='.1f')
slider_nD = widgets.FloatLogSlider(value=nD_init, base=10, min=18, max=22, step=0.1, description='nD [m⁻³]')
slider_nT = widgets.FloatLogSlider(value=nT_init, base=10, min=18, max=22, step=0.1, description='nT [m⁻³]')

# Info numérica
out_text = widgets.HTML(value="")  # aquí mostraremos R y P actuales

# ------------------------------
# Función de actualización
# ------------------------------
def update_plot(change=None):
    T = slider_T.value
    nD = slider_nD.value
    nT = slider_nT.value

    # Reactividad y punto
    ov = sigma_v_DT(T)
    with figw.batch_update():
        # actualizar punto en la gráfica de la izquierda (traza index 1 en esa subplot)
        figw.data[1].x = [T]
        figw.data[1].y = [ov]

        # recalcular P(T) con las densidades actuales y actualizar la curva derecha (traza index 2)
        P_new = nD * nT * reactividad_base * E_f_J
        figw.data[2].y = P_new

        # actualizar título con valores numéricos
        figw.layout.title.text = f"Interacción · T={T:.2f} keV · nD={nD:.2e} m⁻³ · nT={nT:.2e} m⁻³"

    # resultados en HTML
    R = nD * nT * ov
    P_point = R * E_f_J
    out_text.value = (
        f"<b>Condición seleccionada:</b><br>"
        f"T = {T:.2f} keV — ⟨σv⟩ = {ov:.3e} m³/s<br>"
        f"R = nD·nT·⟨σv⟩ = {R:.3e} reacciones·m³/s<br>"
        f"P = R·E_f = {P_point:.3e} W/m³"
    )

# Conectar callbacks (se ejecuta mientras mueves)
slider_T.observe(update_plot, names='value')
slider_nD.observe(update_plot, names='value')
slider_nT.observe(update_plot, names='value')

# Llamada inicial para poblar los valores
update_plot()

# ------------------------------
# Layout final: sliders + figura + texto
# ------------------------------
controls = widgets.HBox([slider_T, widgets.VBox([slider_nD, slider_nT])])
display(controls)
display(figw)
display(out_text)
```

## 6.3 Sección eficaz y pico de gamow

En esta sección se analiza el comportamiento de la sección eficaz de fusión ($\sigma$) y del factor de Gamow, elementos fundamentales para comprender la probabilidad de que ocurra una reacción de fusión entre núcleos de deuterio y tritio. El objetivo del código es visualizar cómo la barrera de Coulomb, la energía de colisión y la temperatura del plasma determinan la región donde las reacciones de fusión son más probables, conocida como ventana o pico de Gamow.

La sección eficaz ($\sigma$) representa la probabilidad de que dos núcleos en colisión se fusionen a una determinada energía de colisión. Su valor depende fuertemente de la energía relativa entre los núcleos: a bajas energías, la repulsión electrostática impide la fusión, mientras que a energías demasiado altas, el tiempo de interacción entre las partículas es demasiado corto para que la reacción ocurra con eficiencia. Por esta razón, la sección eficaz presenta un máximo o pico característico, situado en un rango de energías donde las colisiones son suficientemente energéticas para superar parcialmente la barrera de Coulomb, pero aún lentas como para permitir la interacción nuclear.

En el caso del sistema deuterio–tritio (D–T), la sección eficaz alcanza su máximo alrededor de E ≈ 64 keV, donde la probabilidad de fusión es más alta. Este comportamiento se reproduce en el código mediante una función aproximada de la forma:

$$σ(E) = \frac{A}{E}e^{\frac{-B}{\sqrt{E}}}$$


Esto refleja la dependencia exponencial debida a la penetración cuántica a través de la barrera de Coulomb y la atenuación por energía. Los factores $A$ y $B$ se ajustan empíricamente para reproducir la forma general de los datos experimentales de fusión D–T.

La constante de Gamow ($b_{Gamow}$) y el correspondiente factor de Gamow ($e^{-\sqrt{E_G / E}}$) describen la probabilidad cuántica de penetración a través de la barrera de Coulomb. Este factor decrece exponencialmente con la raíz inversa de la energía, lo que implica que solo una pequeña fracción de las partículas con suficiente energía puede atravesar la barrera y fusionarse. Al mismo tiempo, la distribución de velocidades en el plasma (descrita por la estadística de Maxwell–Boltzmann) decrece exponencialmente para energías altas.

La combinación de ambos efectos se genera una región de solapamiento donde las colisiones son más eficaces: el pico de Gamow. Este pico define el rango de energías más probable para la fusión a una temperatura dada y es esencial para calcular el promedio Maxwelliano ⟨σv⟩, que cuantifica la tasa de reacción promedio en el plasma.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.constants import elementary_charge, physical_constants, pi
from ipywidgets import interact, FloatSlider, VBox, HBox, Label, IntSlider
from IPython.display import display

# Constantes físicas
k_B_J_per_K = physical_constants['Boltzmann constant'][0] # J/K
e_sq = elementary_charge**2 # C^2
hbar_J_s = physical_constants['Planck constant over 2 pi'][0] # J*s
m_u_kg = physical_constants['atomic mass constant'][0] # kg

# Masas de Deuterio y Tritio en kg
m_D_kg = 2.014 * m_u_kg
m_T_kg = 3.016 * m_u_kg

# Masa reducida D-T en kg
m_r_kg = (m_D_kg * m_T_kg) / (m_D_kg + m_T_kg)

# Cargas atómicas (para D y T, Z=1)
Z1 = 1
Z2 = 1

# Constante de Gamow (b)
C_gamow = pi * Z1 * Z2 * e_sq * np.sqrt(m_r_kg / (2 * hbar_J_s**2))

# S-factor empírico (aproximado para D-T) en J*m^2
S_factor_empirical_keV_barn = 1.09e-21 # keV * barns
S_factor_J_m2 = S_factor_empirical_keV_barn * (1.60218e-16) # 1 keV = 1.60218e-16 J, 1 barn = 1e-28 m^2

# Rango de energías de colisión (en Joules)
E_keV_min, E_keV_max = 0.1, 500 # Rango en keV
E_J_range = np.linspace(E_keV_min * 1000 * elementary_charge, E_keV_max * 1000 * elementary_charge, 200)

# Rango de temperaturas (en Kelvin)
T_keV_min, T_keV_max = 1, 30 # Rango en keV
T_K_range = np.linspace(T_keV_min * 1000 * elementary_charge / k_B_J_per_K, T_keV_max * 1000 * elementary_charge / k_B_J_per_K, 100)

# Crear grillas para E_J y T_K
E_J_grid, T_K_grid = np.meshgrid(E_J_range, T_K_range)

# Calcular el factor de Gamow (solo depende de E_J)
E_G_J = (pi * Z1 * Z2 * e_sq * np.sqrt(m_r_kg/2) / hbar_J_s)**2
Gamow_factor_grid = np.exp(-np.sqrt(E_G_J / E_J_grid))

# Aproximación funcional para sigma(E) en m^2 (similar a la usada en el código anterior)
E_keV_grid = E_J_grid / (1000 * elementary_charge)
E_keV_grid = np.maximum(E_keV_grid, 0.1)
sigma_m2_grid = (4.8e-24 / E_keV_grid) * np.exp(-44.2 / np.sqrt(E_keV_grid)) * 1e-4 # cm^2 a m^2 y ajuste de escala heurístico
sigma_m2_grid[E_J_grid <= 0] = 0

# Calculate the integrand: sigma(E) * E * exp(-E/kT)
integrand_grid = sigma_m2_grid * E_J_grid * np.exp(-E_J_grid / (k_B_J_per_K * T_K_grid))

# Convert E_J_grid to keV for plotting axis
E_keV_grid_plot = E_J_grid / (1000 * elementary_charge)

# Convert T_K_grid to keV for plotting axis (using kT in keV)
T_keV_grid_plot = T_K_grid * k_B_J_per_K / (1000 * elementary_charge)

# --- 3D Plotting Function ---
def plot_3d_integrand(elevation, azimuth):
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    surf = ax.plot_surface(E_keV_grid_plot, T_keV_grid_plot, integrand_grid, cmap='plasma', alpha=0.9)

    ax.set_xlabel('Energy $E$ [keV]')
    ax.set_ylabel('Temperature $T$ [keV]')
    ax.set_zlabel('Integrand $\sigma(E) E e^{-E/kT}$ [units]')
    ax.set_title('Integrand of Maxwellian Average $\sigma(E) E e^{-E/kT}$')

    fig.colorbar(surf, label='Integrand Value')

    ax.view_init(elev=elevation, azim=azimuth) # Interactive view angle

    plt.tight_layout()
    plt.show()

# --- Interactive Controls for 3D Plot ---
print("--- Interactive 3D Plot (Adjust View Angle) ---")
interact(plot_3d_integrand,
         elevation=IntSlider(min=0, max=90, step=5, value=30, description='Elevation'),
         azimuth=IntSlider(min=0, max=360, step=5, value=45, description='Azimuth'));


# --- Interactive Slice of the Integrand ---
def plot_integrand_slice(T_selected_keV):
    T_selected_K = T_selected_keV * 1000 * elementary_charge / k_B_J_per_K
    kT_selected_J = k_B_J_per_K * T_selected_K

    integrand_slice = sigma_m2_grid[0, :] * E_J_range * np.exp(-E_J_range / kT_selected_J)

    plt.figure(figsize=(8, 5))
    plt.plot(E_keV_grid_plot[0, :], integrand_slice, color='purple', linewidth=2)
    plt.title(f'Integrand at T = {T_selected_keV:.1f} keV')
    plt.xlabel('Energy E [keV]')
    plt.ylabel('$\sigma(E) E e^{-E/kT}$ [units]')
    plt.grid(True)
    plt.show()

print("\n--- Interactive Slice of the Integrand ---")
interact(plot_integrand_slice,
         T_selected_keV=FloatSlider(min=1.0, max=30.0, step=0.5, value=15.0, description='Select T (keV)'));

# Let's also make an interactive plot of just the Gamow Factor vs Energy
def plot_gamow_factor(E_max_keV):
    E_plot_keV = np.linspace(0.1, E_max_keV, 200)
    E_plot_J = E_plot_keV * 1000 * elementary_charge
    Gamow_factor_plot = np.exp(-np.sqrt(E_G_J / E_plot_J))

    plt.figure(figsize=(8, 5))
    plt.plot(E_plot_keV, Gamow_factor_plot, color='green', linewidth=2)
    plt.title('Gamow Factor vs Collision Energy')
    plt.xlabel('Energy E [keV]')
    plt.ylabel('Gamow Factor $e^{-b/\sqrt{E}}$')
    plt.grid(True)
    plt.show()

print("\n--- Interactive Gamow Factor Plot ---")
interact(plot_gamow_factor,
         E_max_keV=FloatSlider(min=10.0, max=500.0, step=10.0, value=100.0, description='Max Energy (keV)'));


# Finally, an interactive plot of just the Cross Section vs Energy
def plot_cross_section(E_max_keV):
    E_plot_keV = np.linspace(0.1, E_max_keV, 200)
    E_plot_J = E_plot_keV * 1000 * elementary_charge

    # Recalculate sigma_m2 for the plot range
    # Avoid division by zero or very small energy values
    E_plot_keV_safe = np.maximum(E_plot_keV, 0.1)
    sigma_m2_plot = (4.8e-24 / E_plot_keV_safe) * np.exp(-44.2 / np.sqrt(E_plot_keV_safe)) * 1e-4 # cm^2 a m^2 y ajuste de escala heurístico
    sigma_barns_plot = sigma_m2_plot * 1e28 # Convert to barns

    plt.figure(figsize=(8, 5))
    plt.plot(E_plot_keV, sigma_barns_plot, color='red', linewidth=2)
    plt.title('Fusion Cross Section $\sigma(E)$ vs Collision Energy')
    plt.xlabel('Energy E [keV]')
    plt.ylabel('Cross Section $\sigma$ [barns]')
    plt.yscale('log') # Often plotted on a log scale
    plt.grid(True, which='both', ls='--')
    plt.show()

print("\n--- Interactive Cross Section Plot ---")
interact(plot_cross_section,
         E_max_keV=FloatSlider(min=10.0, max=500.0, step=10.0, value=100.0, description='Max Energy (keV)'));
```

El código visualiza esta relación mediante tres gráficos interactivos principales:

Gráfica 3D del integrando $\sigma(E)E e^{-E/kT}$:
Muestra cómo varía el producto de la sección eficaz, la energía y el factor térmico con la temperatura. La superficie revela una “cresta” que se desplaza hacia energías mayores a medida que aumenta la temperatura, lo que corresponde al desplazamiento del pico de Gamow hacia valores más altos de energía.

Cortes del integrando para diferentes temperaturas:
Permiten observar cómo cambia la forma del pico de Gamow con la temperatura. A bajas temperaturas, el pico es estrecho y se sitúa a bajas energías; conforme la temperatura aumenta, el pico se ensancha y se traslada hacia energías mayores, indicando que un rango más amplio de partículas contribuye significativamente a la fusión.

Factor de Gamow y sección eficaz vs energía:
Los gráficos individuales del factor de Gamow muestran su rápida caída exponencial a bajas energías, mientras que la curva de la sección eficaz exhibe un máximo en la región de interés para fusión D–T. Estas curvas complementan la interpretación del pico de Gamow como el punto donde ambos efectos de distribución térmica y penetración cuántica se superponen de manera óptima.

De esta manera se logra comprender que el pico de Gamow representa el equilibrio entre las colisiones suficientemente energéticas y las probabilidades cuánticas de penetración. Su posición y amplitud determinan la eficiencia con que un plasma puede sostener reacciones de fusión. Para la fusión D–T, este pico se sitúa en torno a 10–20 keV de temperatura de plasma, lo que coincide con el rango óptimo para lograr una alta reactividad ⟨σv⟩ y, por ende, una mayor potencia de fusión.

# 9 Ciclo Rankine de una planta nuclear

image.png





El ciclo Rankine es el modelo termodinámico que describe cómo la planta convierte el calor generado en el reactor nuclear en **energía eléctrica**. En esta central, el ciclo Rankine se implementa en el **circuito secundario de vapor**, mientras que el reactor y el lazo primario se encargan únicamente de generar y transferir calor de forma segura al generador de vapor.

A grandes rasgos, el ciclo se compone de cuatro procesos ideales:

1. **Bombeo del líquido a alta presión**  
2. **Aporte de calor a presión casi constante (generador de vapor)**  
3. **Expansión del vapor en la turbina**  
4. **Condensación del vapor a baja presión (condensador)**  

En esta planta, estos procesos se asocian a equipos reales como bombas centrífugas, generadores de vapor, turbina de vapor y el condensador principal, junto con los sistemas de agua de circulación y agua de alimentación.

---

### 9.1 Esquema general del ciclo Rankine en una planta nuclear

Aunque en una central térmica convencional el calor se genera en una caldera a partir de combustibles fósiles, en esta planta **el origen del calor es el reactor nuclear**. Sin embargo, el ciclo Rankine se mantiene prácticamente igual en el lazo secundario:

- El **circuito primario** (con las bombas RCP) extrae el calor del núcleo del reactor y lo lleva al **generador de vapor**.  
- En el **generador de vapor**, ese calor se transfiere al **circuito secundario**, donde el agua de alimentación se evapora y se convierte en vapor de alta presión y temperatura.  
- Ese vapor acciona la **turbina de vapor**, que a su vez mueve el **generador eléctrico**.  
- Tras la expansión, el vapor se condensa en el **condensador** y el agua resultante se bombea de nuevo como **agua de alimentación**, cerrando el ciclo.

De esta forma, el reactor nuclear nunca entra en contacto directo con la turbina ni con el condensador: el ciclo Rankine se desarrolla completamente en el **lado secundario**, usando agua desmineralizada como fluido de trabajo.

---

### 9.2 Etapas termodinámicas del ciclo Rankine

Podemos describir el ciclo Rankine de la planta en cuatro etapas principales, asociando cada una con los equipos reales:

#### (1) Condensador → (2) Bomba de condensado / agua de alimentación  
**Proceso 1–2: Bombeo (compresión líquida)**  

- El vapor que sale de la turbina entra al **condensador**, donde se enfría y condensa hasta convertirse en agua líquida a baja presión.  
- Esta agua condensada es aspirada por las **bombas de condensado** y, posteriormente, por las **bombas de agua de alimentación principales (MFWP)**.  
- En esta etapa se incrementa la **presión del fluido**, con un costo de trabajo relativamente pequeño comparado con la potencia generada en la turbina, debido a que el agua es prácticamente incompresible.  

En la planta del señor Vernns, las **MFWP** son las responsables de elevar la presión y asegurar el caudal adecuado hacia el generador de vapor.

---

#### (2) Bomba → (3) Generador de vapor  
**Proceso 2–3: Aporte de calor a presión casi constante**

- El agua de alimentación a alta presión entra al **generador de vapor**.  
- En el lado primario, el refrigerante calentado en el reactor cede calor al lado secundario a través de los tubos del generador.  
- En el lado secundario, el agua se calienta, hierve y se convierte en **vapor saturado o ligeramente sobrecalentado**, manteniendo prácticamente constante la presión.  

En esta etapa se produce la **entrada principal de calor al ciclo Rankine**, y es el punto donde el acoplamiento con el reactor nuclear se hace efectivo (aunque sea a través de un intercambiador de calor).

---

#### (3) Generador de vapor → (4) Turbina de vapor  
**Proceso 3–4: Expansión adiabática en la turbina**

- El vapor de alta presión y temperatura se dirige a la **turbina de vapor**.  
- En la turbina, el vapor se expande, disminuyen su **presión y temperatura**, y parte de su energía interna se transforma en **trabajo mecánico en el eje de la turbina**.  
- Este trabajo mecánico se transmite al **generador eléctrico**, donde se convierte en energía eléctrica que se entrega a la red.  

En esta etapa es donde se genera la mayor parte del trabajo útil del ciclo.

---

#### (4) Turbina → (1) Condensador  
**Proceso 4–1: Rechazo de calor en el condensador**

- El vapor que sale de la turbina entra al **condensador**, donde se enfría gracias al **agua de circulación**.  
- La **condensación** del vapor se realiza a presión baja, favoreciendo un mayor salto de entalpía en la turbina y, por tanto, mayor producción de trabajo.  
- El fluido de trabajo vuelve a ser agua líquida, cerrando así el ciclo.  

En este punto son fundamentales las **bombas de circulación (CP)**, que mueven grandes caudales de agua a través del condensador para poder extraer el calor rechazado del ciclo.

---

### 9.3 Rol de las bombas centrífugas dentro del ciclo Rankine de la planta

En el contexto del ciclo Rankine de la planta, las bombas centrífugas se integran de la siguiente manera:

- **Bombas RCP (Reactor Coolant Pump)**  
  - Están en el **circuito primario**, no forman parte directa del ciclo Rankine, pero son esenciales para **transportar el calor desde el reactor hasta el generador de vapor**.  

- **Bombas MFWP (Main Feedwater Pump)**  
  - Forman parte directa del **ciclo Rankine**, en la etapa de **bombeo del agua de alimentación**.  
  - Elevan la presión del agua líquida desde el condensador/desaireador hasta los valores requeridos en la entrada del generador de vapor.  

- **Bombas CP (Circulating Pumps)**  
  - Son indispensables para el **rechazo de calor en el condensador**, ya que mantienen el caudal de agua de enfriamiento que permite condensar el vapor de escape de la turbina.

---

### 9.4 Comentarios sobre la eficiencia del ciclo

La **eficiencia térmica** del ciclo Rankine en una planta nuclear está limitada principalmente por:

- La **temperatura máxima** del vapor a la salida del generador de vapor, restringida por materiales y seguridad del reactor.  
- La **temperatura mínima** en el condensador, que depende de la temperatura del agua de enfriamiento disponible.

Aun así, combinando el reactor nuclear (como fuente de calor) con un ciclo Rankine bien diseñado incluyendo bombas, turbina, condensador y sistemas auxiliares, se logra convertir una fracción significativa del calor de fisión en energía eléctrica de forma continua y controlada.

### 8.5 Tipo de ciclo Rankine utilizado en la planta

En la literatura se suelen distinguir varios tipos de ciclos Rankine:

- **Ciclo Rankine directo (o simple):**  
  El calor entra al ciclo en la caldera o generador de vapor y la máquina principal (turbina) entrega trabajo hacia un generador eléctrico. Es el ciclo típico de **plantas de potencia**.

- **Ciclo Rankine inverso:**  
  Es el Rankine “al revés” y se usa como **ciclo de refrigeración o bomba de calor**. En este caso el objetivo principal no es producir trabajo eléctrico, sino **extraer calor de una fuente fría** (como un evaporador) y cederlo a una fuente caliente. No es el caso de una central nuclear de generación eléctrica.

- **Ciclos Rankine modificados:**  
  Incluyen mejoras como:
  - **Recalentamiento (reheat):** el vapor se expande parcialmente en la turbina, se vuelve a calentar y se sigue expandiendo.
  - **Regeneración:** parte del vapor se extrae de la turbina para calentar el agua de alimentación en calentadores regenerativos.
  - Versiones supercríticas, multicuerpo, etc.

---

#### Tipo específico en la planta

En este proyecto, la planta modelada corresponde a una **central nuclear tipo PWR**, en la que:

- El **reactor nuclear** entrega calor al **generador de vapor** en el circuito primario.
- El **ciclo termodinámico que se desarrolla en el circuito secundario** es un ciclo de vapor que:
  - Recibe calor en el generador de vapor.
  - Expande el vapor en la **turbina de vapor**.
  - Condensa el vapor en el **condensador**.
  - Bombea el condensado de vuelta como **agua de alimentación**.

En los cálculos realizados en este cuaderno:

- No se modelaron explícitamente **calentadores regenerativos** ni **etapas de recalentamiento del vapor**.
- La turbina se consideró como un solo cuerpo con una expansión desde una presión alta hasta la presión del condensador.
- Las bombas (RCP, MFWP y CP) se incorporan para representar el movimiento del refrigerante en el primario, del agua de alimentación y del agua de circulación.

Por lo tanto, desde el punto de vista termodinámico, el ciclo que se está utilizando y representando en el diagrama T–s es:

# 10 Condensador de superficie

Un condensador de superficie es un intercambiador de calor diseñado espcialmente para condesar vapor a presiones muy bajas. En este tipo de intercambiadores no hay contacto entre los dos fludios y usualmente se ubica para que este trabaje con el vapor de salida de la turbina en un ciclo Rankine.

En el caso de una planta nuclear con reactor de agua en ebullición (BWR) este tipo de condensadores es escencial para el correcto funcionamiento de la planta ya que este dispositivo permite enfríar y, valga la reundancia, condensar el vapor de agua que sale de la turbina para poder ser reingresado a el núcleo del reactor y continuar con el ciclo.

## 10.2 Balance de masa
La ecuación general del balance de masa en un sistema cerrado  viene dada por
$$
    \frac{d M }{d t} = \sum_i \dot{m}_i \tag{1}
$$
donde $M$ es la masa total del sistema y $\dot{m}_i$ es el i-ésimo flujo (positivo entrante, negativo saliente) del sistema.
Este flujo másico se puede calcular sabiendo la velocidad promedio del flujo $\overline{V}$, la densidad del fluido $\rho$ y el área de sección transversal del canal de flujo $A$, de la forma $\dot{m}=\rho\overline{V}{A}$.

En el caso de un líquido fluyendo en el interior de una tubería, donde no hay acumulación ni compresibilidad del fluido, se tiene que el flujo másico a la entrada es igual a flujo másico de la salida $\dot{m}_1=\dot{m}_2$.
Por el contrario, en el caso del vapor, siendo un gas compresible, se debe evaluar el cambio de masa en el sistema debido a procesos de compresión o expansión. En este caso se asume que la totalidad del vapor de entrada al dispositibo se condensa por completo generando una disminución significativa en el volúmen de salida.


### **10.2.1 Hidrodinámica del agua refrigerante**

En el modelo propuesto, la caracterización hidráulica del agua dentro de los tubos es un paso fundamental, pues la determinación de coeficientes convectivos y gradientes de temperatura depende directamente del régimen de flujo y de la distribución dentro del haz tubular.
Por esta razón, se analizan de manera secuencial las siguientes zonas hidráulicas:

- Ingreso del agua refrigerante desde la torre de enfriamiento,

- Distribución del flujo en el cabezal del banco de tubos, y

- Hidrodinámica interna en cada tubo del condensador.

#### Condiciones de ingreso del agua al condensador

El flujo de agua que ingresa al condensador proviene del circuito de enfriamiento, impulsado por las bombas de circulación. Esta condición de ingreso se considera un valor medido y conocido, dado que las plantas nucleares de 300 MW (SMR) disponen de sensores instalados en la línea de suministro hacia el condensador.

A partir de la lectura del caudal volumétrico $\dot{m}w_1$ entregado por el sistema de bombeo, puede determinarse el flujo másico de ingreso como:

$$\dot{m}w_1 = ρw_1\dot{q}w_1 \tag{2}$$

Donde:
- $ρw_1$  es la densidad del agua a la temperatura de entrada.
- $\dot{q}w_1$ es el caudal volumétrico entregado al condensador.

Asímismo, este flujo másico constituye la condición inicial para el balance de masa del lado frío y determina la velocidad promedio en la tubería de entrada:

$$ \bar{V}_1 = \frac{ \dot{q}w_1}{A_1} \tag{3}$$

#### Distribución del flujo en el cabezal del condensador

Para determinar la velocidad promedio del flujo en el cabezal del condensador se toma la ecuación de balance de masa y, entendiendo que no hay acumulación o perdidas de masa en la tubería se tiene
$\dot{m}_{W1}=\dot{m}_{W2}$ siendo $\dot{m}_{W2}$ el flujo másico en la tapa distribuidora, asi obtenemos:

$$\dot{m}w_2 = ρw_2\dot{q}w_2 \tag{4}$$

$$ \bar{V}_2 = \frac{ \dot{q}w_2}{A_2} \tag{5}$$

#### Flujo en cada tubo al interior del condensador

El flujo másico de la tapa se divide en cada uno de los tubos por lo que el balance de masa queda de la siguiente manera:

$$\dot{m}_{W2}= \rho_{W1}\sum\overline{V}_{Wi}{A_i}  \tag{6} $$

Donde $\overline{V}_{Wi}{A_i}$ es el caudal de un iésimo tubo. Dependiendo de la cantidad $N$ de tubos que contenga el intercambiador la velocidad promedio del flujo en cada tubo se puede calcular de la sieguiente manera:

$$
\overline{V}_{W3}=\frac{\dot{m}_{W2}}{N\rho_{W1}{A_3}}. \tag{7}
$$

Una vez obtenida la velocidad de flujo en cada tubo se puede calcular el número de Reynolds para así indicar el régimen en el que se encuentra el fluido y posteriormente calcular el coeficiente de transferencia de calor por convección en el interior de los tubos. El número de Reynolds es definido por la siguiente relación:

$$
\text{Re}=\frac{\rho{D}\overline{V}}{\mu}, \tag{8}
$$

Donde $D$ es asociado con el diametro de la tuberia y $\mu$ es la viscosidad del fluido. Especificamente para cada tubo, en número de Reynolds se puede definir como: $\text{Re}=\rho_{W1}{D}_{1}\overline{V}_{W3}/\mu_{W1}$ donde ${D}_{1}$ es el diámetro interno de los tubos BWG y $\mu_{W1}$ es la viscosidad del agua líquida.

Para determinar las condiciones de salida del condensador se hace un analisis similar al que se desarrolla con la entrada ya que la geometría es similar.


### **10.2.2 Hidrodinámica del vapor de agua**

 En un condensador de superficie el vapor se condensa al contacto con la superficie fría de los tubos, por lo que su caudal másico disminuye a lo largo del eje del haz de tubos. A continuación se presentan las ecuaciones y definiciones necesarias para caracterizar la hidrodinámica del vapor en presencia de condensación.

Para evaluar este comportamiento de condensación del vapor dentro del dispositivo se toma una aproximación discretizando la carcasa en $n$ secciones axiales, filas de tubos, a partir de esto se obtiene la siguiente ecuación:

$$ \dot{m}_{v,i+1}(t) = \dot{m}_{v,i}(t) - Δ\dot{m}_{cond,i}(t) \tag{9}$$

Donde $\dot{m}_{v,i} = \dot{m}s_1$ y:

$$Δ\dot{m}_{cond,i}(t) = \frac{q_i(t)}{h_{fg}(T_{sat})} \tag{10}$$

Siendo $q_i(t)$ con la potencia extraída en la sección $i(W)$ y $h_{fg}$ el calor latente a la temperatura de saturación correspondiente. Esta expresión relaciona directamente la transferencia de calor local con la tasa de condensación local.

Para encontrar la velocidad promedio del vapor se emplea la siguiente formula:

$$\overline{V}_{S2}=\frac{\dot{m}_{v,i}(t)}{\rho_{v}(T_{sat}) A_{2}} \tag{11}$$

Para determinar la velocidad máxima del fluido a través del banco de tubos se emplea la siguiente ecuación:

$$
V_{max}=\frac{ST}{ST-D_{2}}{\overline{V}_{S2,i}(t)} \tag{12}
$$

De igual manera que con el flujo frío, se debe determinar el número de Reynolds para estimar la transferencia de calor convectiva. la velocidad máxima es la que se usa para hahcer este calculo.




## 10.3 Balance de energía

La ecuación general del balance de energía en un sistema cerrado viene dada por:
$$
 \rho c_p\frac{d T }{d t} \mathcal{V} = \sum_i \dot{m}_i c_p T_i + \sum_j \dot{Q}_j \tag{13}
 $$
donde $T$ es la temperatura del sistema, $\mathcal{V}$ es el volúmen del sistema, $\dot{m}_i c_p T_i$ es la energía transportada por el i-ésimo flujo (positivo entrante, negativo saliente) sobre el sistema y $\dot{Q}_j$ es el j-ésimo flujo de calor sobre el sistema.
La aplicación particular de esta ecuación sobre el agua refrigerante y el vapor se describe a continuación.

### 10.3.1 Agua refrigerante

El balande energético para el agua refrigerante es el resultado de un analisis de la cuantificación de los cambios de temperatura del agua al interior de los tubos. El balance energético del agua resulta en:

\begin{align}
\rho \mathcal{V}c_{pW}\frac{dT_{W}}{dt}=30\ \dot{m}_{W3}c_{pW}T_{W1}-\dot{m}_{W5}c_{pW}T_{W5}+\dot{Q}_{1} \tag{14}
\end{align}

Siendo $T_W(t)$ la temperatura del agua al interior de los tubos, $T_{W1}$ la temperatura (conocida) de entrada del agua y $T_{W5}$ la temperatura (desconocida) del agua a la salida, $c_{pW}$ el calor específico a presión constante del agua líquida, $\dot{Q}_{1}$ el flujo de calor transferido del vapor al agua y $\mathcal{V}$ el volumen total interior de los tubos. Para fines prácticos se asume que la temperatura de salida del fluido es la misma que la temperatura del fluido dentro del tubo.

#### Números adimensionales

Para determinar el el coeficiente de transferencia de calor por convección, valor que influye en el cálculo de $\dot{Q}_{1}$ se necesita calcular los números adimensionales.  Para calcular el número de Prandtl, realción entre la viscosidad y la difusividad térmica, se tiene la siguiente ecuación:

\begin{align}
\text{Pr}=\frac{\nu}{\alpha}=\frac{c_p\mu}{K} \tag{15}
\end{align}

Donde $\nu$ es la viscosidad cinemática, $\alpha$ es la difusividad térmica y $K$ es la conductividad térmica del fluido.

El número de Nusselt es empleado para poder determinar el coeficiente de transferencia de calor por convección ($h$) y se define de la siguiente manera:

\begin{align}
\text{Nu}=\frac{h}{DK} \tag{16}
\end{align}

Donde D es el diametro de la tubería. Las relaciones de transferencia de calor por convección generalmente se expresan escribiendo el número de Nusselt en función del número de Reynolds y del número de Prandtl.
Para el cálculo del número de Nusselt de un fluido que fluye en el interior de un tubo, la literatura recomienda la ecuación de Chilton-Colburn:
\begin{align}
\text{Nu}=0.023\text{Re}^{0.8}\text{Pr}^{0.4} \tag{17}
\end{align}

Al tener tanto $Re$ como $Nu$ como $Pr$ se puede calcular el h del fluido y así calcular el calor transferido por convección ($\dot{Q}_{1}$). Esta transferencia se puede calcular usando la siguiente ecuación:

\begin{align}
\dot{Q}_1= h_{W}(T_{C2}-T_{W})A_{int} \tag{18}
\end{align}

donde $T_{C2}$ es la temperatura de la superficie interior del tubo, $h_{W}$ es el coeficiente de transferencia de calor por convección en el agua interior y $A_{int}$ es el área de transferencia de calor del interior del tubo.



### 10.3.2 Vapor de agua en la carcasa (Condensación)

Para estimar la tasa de transferencia de calor asociada al proceso de condensación del vapor en el condensador, se realiza un balance energético sobre un volumen de control que contiene el vapor y la película condensada sobre el haz tubular.  
Se adopta el supuesto simplificado de **condensación completa**, es decir, todo el vapor que ingresa se transforma en líquido saturado a la presión del condensador.

Los supuestos principales del modelo son:

- El vapor ingresa como vapor saturado o ligeramente sobrecalentado a presión \(P_{sat}\).
- La presión permanece constante a lo largo del proceso.
- El 100% del vapor que entra se condensa:  
  \[
  \dot{m}_{S2} = \dot{m}_{cond}
  \]
- El condensado sale como agua saturada a temperatura \(T_{sat}\).


#### Balance energético del vapor en proceso de condensación

Planteando un balance de energía sobre el volumen ocupado por el vapor:

$$
\rho_S \mathcal{V}_S c_{pS}\frac{d T_{S}(t)}{dt}
=
\dot{m}_{S2} c_{pS} T_{S2}
-
\dot{m}_{S3} c_{pS} T_{S}
-
\dot{Q}_{1}
-
\dot{Q}_{2}
\tag{19}
$$

donde:

- $\dot{Q}_{1} $ es el calor transferido del vapor hacia los tubos (condensación).
- $\dot{Q}_{2} $ es el calor perdido hacia el ambiente.
- Se toma $T_S = T_{S3}$ (temperatura uniforme del vapor en la carcasa).


#### Relación energética de la condensación

Si el vapor se condensa completamente, el calor liberado por unidad de tiempo es:

$$
\dot{Q}_i = \dot{m}_{v,i}\, h_{fg}(T_{sat})
\tag{20}
$$

Para relacionarlo con la convección en la superficie externa de los tubos:

$$
\dot{Q}_i = h_{S,i}\, A_{i}\, (T_{sat} - T_{w,i})
\tag{21}
$$

Igualando (18) y (19):

$$
h_{S,i} =
\frac{\dot{m}_{v,i}\, h_{fg}}
{A_i (T_{sat} - T_{w,i})}
\tag{22}
$$

Esta expresión surge del balance energético directo, pero para conectar con parámetros geométricos e hidrodinámicos se usa una correlación basada en números adimensionales.



#### Cálculo del coeficiente convectivo mediante números adimensionales

La convección del vapor en flujo cruzado sobre el haz tubular se estima a partir de los números de *Reynolds*, *Prandtl* y *Nusselt*.  
El número de Prandtl se obtiene mediante la expresión:

$$
\mathrm{Pr} = \frac{c_{pS} \mu_S}{k_S}
\tag{23}
$$

La correlación de Chilton–Colburn para flujo cruzado en bancos de tubos alineados es:

$$
\mathrm{Nu} =
C\, \mathrm{Re}^{m}\, \mathrm{Pr}^{n}\left(\frac{\mathrm{Pr}}{\mathrm{Pr_s}}\right)^{0.25}
\tag{24}
$$

donde:

- $\mathrm{Pr_s}$ es el número de Prandtl evaluado a la temperatura de película,
- $(C, m, n)$ son coeficientes determinados experimentalmente.

Para un banco alineado, los valores son:

$$
\mathrm{Nu}(\mathrm{Re}, \mathrm{Pr}, \mathrm{Pr_s}) =
\begin{cases}
0.9\, \mathrm{Re}^{0.4}\mathrm{Pr}^{0.36}(\mathrm{Pr}/\mathrm{Pr_s})^{0.25}, & \mathrm{Re}\le 100 \\[6pt]
0.52\, \mathrm{Re}^{0.5}\mathrm{Pr}^{0.36}(\mathrm{Pr}/\mathrm{Pr_s})^{0.25}, & 100<\mathrm{Re}\le 1000\\[6pt]
0.27\, \mathrm{Re}^{0.63}\mathrm{Pr}^{0.36}(\mathrm{Pr}/\mathrm{Pr_s})^{0.25}, & 1000<\mathrm{Re}\le 2\times10^5\\[6pt]
0.033\, \mathrm{Re}^{0.8}\mathrm{Pr}^{0.8}(\mathrm{Pr}/\mathrm{Pr_s})^{0.25}, & 2\times10^5<\mathrm{Re}\le 2\times10^6
\end{cases}
\tag{25}
$$

Debido a que el haz posee menos de 16 filas, se aplica un factor de corrección:

$$
\mathrm{Nu}_{corr} = F \cdot \mathrm{Nu},
\qquad F = 0.99
$$

#### Convección natural (carcasa – ambiente)

En esta sección se estima la pérdida de calor desde la superficie externa de la carcasa del condensador hacia el aire ambiente mediante convección natural.  
Esta pérdida corresponde a $\dot{Q}_2$, uno de los términos del balance energético general del vapor.


El flujo de calor total disipado desde la carcasa hacia el ambiente está dado por:

$$
\dot{Q}_2 = h_E\, A_E \,(T_S - T_\infty)
\tag{26}
$$

donde  

- $h_E$ es el coeficiente de convección natural del aire,  
- $A_E = \pi D L$ es el área externa de la carcasa,  
- $T_S$ es la temperatura de la superficie de la carcasa (igualada a la del vapor saturado interno),  
- $T_\infty$ es la temperatura ambiente.


Para obtener $h_E$, primero se evalúan los números de **Grashof**, **Rayleigh** y luego el **Nusselt**.

#### Número de Grashof

$$
\text{Gr} =
\frac{g\,\beta \,(T_S - T_\infty)\, L_C^3}{\nu^2}
\tag{27}
$$

donde:

- $g$: gravedad  
- $\beta$: coeficiente de expansión térmica del aire  
- $L_C = D$: diámetro de la carcasa (longitud característica)  
- $\nu\$: viscosidad cinemática del aire  
- Propiedades evaluadas a la temperatura de película  
$$
T_f = \frac{T_S + T_\infty}{2}
$$

#### Número de Rayleigh

$
\text{Ra} = \text{Gr}\,\text{Pr}
\tag{28}
$

Donde $\text{Pr}$ es el número de Prandtl del aire a $T_f$.


#### Número de Nusselt para convección natural

El número de Nusselt para superficies externas se obtiene mediante la correlación empírica:

$$
\text{Nu}=
\left[
0.6+
\frac{
0.387\,\text{Ra}^{1/6}
}{
\left(1 + (0.559/\text{Pr})^{9/16}\right)^{8/27}
}
\right]^2
\tag{29}
$$

Con este número se calcula el coeficiente de convección natural:

$$
h_E = \frac{k}{L_C}\,\text{Nu}
$$

donde $k$ es la conductividad térmica del aire evaluada a $T_f$.


#### Expresión final para el calor perdido

Finalmente, el flujo de calor por convección natural queda como:

$$
\dot{Q}_{2}=h_E A_E (T_S - T_\infty)
\tag{30}
$$

con  
$
A_E=\pi D L
$

#### Transferencia de energía a través de la pared del tubo

En esta sección se analiza la transferencia de calor por **conducción** a través de la pared metálica del tubo, que conecta los mecanismos convectivos del vapor (carcasa) y del agua refrigerante (interior del tubo).



#### Ecuación de conducción en la pared del tubo

El flujo de calor que atraviesa la pared del tubo se expresa mediante:

$
\dot{Q}_{1}=K\,A\, (T_{C1} - T_{C2})
\tag{31}
$

donde:

- $K$ es la conductividad térmica del material del tubo,  
- $T_{C1}$ es la temperatura en la superficie externa del tubo (lado del vapor),  
- $T_{C2}$ es la temperatura en la superficie interna del tubo (lado del agua refrigerante),  
- $A$ es el área de transferencia de calor correspondiente a la superficie externa de los tubos.

Este flujo de calor $\dot{Q}_1$ es **idéntico** al que se transfiere:

- por convección desde el vapor a la superficie externa del tubo,  
- y por convección desde la superficie interna del tubo al agua refrigerante.



#### Área efectiva de conducción

El área de transferencia de calor se considera como:

$$
A = A_{\text{ext}} = N_{\text{tubos}} \,\pi D_2 L
$$

donde:

- $N_{\text{tubos}}$: número total de tubos,  
- $D_2$: diámetro exterior del tubo,  
- $L$: longitud de los tubos.

Esta área es equivalente al área externa total del banco de tubos por la cual se transmite el calor.

### 10.4.2 Simulación

```python
import math
import pandas as pd

# --- Attempt to use CoolProp for higher fidelity (optional) ---
USE_COOLPROP = False
try:
    from CoolProp.CoolProp import PropsSI
    USE_COOLPROP = True
    # print("CoolProp available: using for properties.")
except Exception:
    # print("CoolProp NOT available: using default properties.")
    USE_COOLPROP = False

# ---------- Constants ----------
g = 9.81  # m/s2

# ---------- Default Properties (fallback) ----------
def default_props_water(T=293.15, p=101325.0):
    # Values for liquid water at ~20C and 1 atm
    return {"rho": 998.2, "cp": 4182.0, "k": 0.597, "mu": 1.002e-3, "nu": 1.004e-6}

def default_props_air(T=293.15, p=101325.0):
    # Values for air at ~20C and 1 atm
    return {"rho": 1.204, "cp": 1007.0, "k": 0.0255, "mu": 1.81e-5, "nu": 1.5e-5, "beta": 1.0/T, "Pr": 0.71}

def default_props_steam_sat(T_sat=373.15, p=101325.0):
    # Values for saturated steam at ~100C and 1 atm
    return {"rho_v": 0.597, "h_fg": 2257000.0, "cp_v": 2010.0, "mu_v": 1.2e-5, "k_v": 0.024}

# ---------- Functions for properties (uses CoolProp if available) ----------
def props_water(T=293.15, p=101325.0):
    # Temperature in K
    if USE_COOLPROP:
        try:
            rho = PropsSI("D", "T", T, "P", p, "Water")
            cp = PropsSI("C", "T", T, "P", p, "Water")
            k = PropsSI("L", "T", T, "P", p, "Water")
            mu = PropsSI("V", "T", T, "P", p, "Water")  # dynamic Pa.s
            nu = mu / rho
            return {"rho": rho, "cp": cp, "k": k, "mu": mu, "nu": nu}
        except Exception:
            return default_props_water(T,p)
    else:
        return default_props_water(T,p)

def props_air(T=293.15, p=101325.0):
    if USE_COOLPROP:
        try:
            rho = PropsSI("D", "T", T, "P", p, "Air")
            cp = PropsSI("C", "T", T, "P", p, "Air")
            k = PropsSI("L", "T", T, "P", p, "Air")
            mu = PropsSI("V", "T", T, "P", p, "Air")
            nu = mu / rho
            beta = 1.0/T # Ideal gas assumption
            Pr = cp * mu / k
            return {"rho": rho, "cp": cp, "k": k, "mu": mu, "nu": nu, "beta": beta, "Pr": Pr}
        except Exception:
            return default_props_air(T,p)
    else:
        return default_props_air(T,p)

def props_steam_sat(T_sat=373.15, p=101325.0):
    if USE_COOLPROP:
        try:
            # returns properties at saturation (vapor)
            rho_v = PropsSI("D", "T", T_sat, "Q", 1, "Water")
            h_v = PropsSI("H", "T", T_sat, "Q", 1, "Water")  # J/kg
            h_f = PropsSI("H", "T", T_sat, "Q", 0, "Water")
            h_fg = h_v - h_f
            cp_v = PropsSI("C", "T", T_sat, "Q", 1, "Water")
            mu_v = PropsSI("V", "T", T_sat, "Q", 1, "Water")
            k_v = PropsSI("L", "T", T_sat, "Q", 1, "Water")
            return {"rho_v": rho_v, "h_fg": h_fg, "cp_v": cp_v, "mu_v": mu_v, "k_v": k_v}
        except Exception:
            return default_props_steam_sat(T_sat, p)
    else:
        return default_props_steam_sat(T_sat, p)

# ---------- Correlaciones ----------
def nu_natural_churchill_chu(Gr, Pr):
    # Churchill & Chu for natural convection on horizontal cylinders
    Ra = Gr * Pr
    if Ra <= 0:
        return 0.0
    term = (0.387 * Ra**(1.0/6.0)) / (1.0 + (0.559/Pr)**(9.0/16.0))**(8.0/27.0)
    Nu = (0.6 + term)**2
    return Nu

def h_natural_convection_air(T_surface, T_inf, L_c, air_props=None):
    # L_c: characteristic length (diameter for horizontal cylinder)
    T_f = (T_surface + T_inf) / 2.0
    if air_props is None:
        air_props = props_air(T_f)

    beta = air_props.get("beta", 1.0 / max(T_f, 1.0))
    nu = air_props["nu"]
    Pr = air_props["Pr"]
    k_air = air_props["k"]

    Gr = g * beta * abs(T_surface - T_inf) * (L_c**3) / (nu**2)
    Nu = nu_natural_churchill_chu(Gr, Pr)
    h = Nu * k_air / L_c
    return max(h, 1e-4) # ensure h is not zero

def overall_UA(h_steam, A_ext, h_water, A_int, k_wall, t_wall, A_wall_mean, R_fouling):
    # All areas are total areas for the section being considered
    R_steam = 1.0 / (h_steam * A_ext) if h_steam > 0 and A_ext > 0 else 1e12
    R_wall = t_wall / (k_wall * A_wall_mean) if k_wall > 0 and t_wall > 0 and A_wall_mean > 0 else 0.0
    R_water = 1.0 / (h_water * A_int) if h_water > 0 and A_int > 0 else 1e12
    R_fouling_total = R_fouling # Assuming R_fouling is given per unit area and already scaled by area

    Rtot = R_steam + R_wall + R_water + R_fouling_total
    U_overall = 1.0 / Rtot if Rtot > 0 else 0.0
    return U_overall, Rtot


default_condenser_params = {
    # Geometrical Parameters
    "D1": 0.02,          # m, diámetro interior del tubo
    "D2": 0.025,         # m, diámetro exterior del tubo
    "ST": 0.035,         # m, straight pitch (pitch entre centros, en dirección transversal)
    "D_shell": 2.0,      # m, diámetro interno de la carcasa
    "L": 5.0,            # m, longitud efectiva del haz de tubos y carcasa (longitud de intercambio)
    "N_tubes": 500,      # Nº de tubos en el haz
    "N_sections": 20,    # Nº de secciones axiales para discretización
    "A1": 0.5,           # m^2, tubería de ingreso del agua al water box (example value)
    "A2": 0.8,           # m^2, tapa de ingreso de vapor (example value)

    # Fluid Properties (initial/reference values, CoolProp will update dynamically)
    "T_ref_water": 293.15, # K, Reference temp for water properties
    "P_ref": 101325.0,     # Pa, Reference pressure

    # Operating Conditions
    "m_dot_v_in": 0.5,   # kg/s, Flujo másico de vapor entrante (turbina → condensador)
    "pc": 8000.0,        # Pa, Presión dentro de la carcasa (condenser pressure) (8 kPa = ~39.4 C)
    "Tc_in": 288.15,     # K, Temperatura de entrada del agua (tube-side in) (15 C)
    "T_ambient": 293.15, # K, Temperatura ambiente (20 C)
    "mdot_c": 10.0,      # kg/s, Flujo másico de agua de enfriamiento (tube-side)

    # Material Properties
    "k_tube": 50.0,      # W/mK, conductividad térmica del material del tubo (ej. acero)
    "wall_rho": 7800.0,  # kg/m^3, densidad del material de la pared
    "wall_cp": 500.0,    # J/kgK, calor específico del material de la pared
    "R_fouling": 0.0001, # m^2K/W, Resistencia por fouling (in/out) (conservative value for nuclear)

    # Correction Factors/Others
    "phi_bank": 0.9      # Factor empírico de corrección por banco de tubos
}

# Derived geometrical parameters (for initial definition or later use)
def calculate_derived_geometrical_params(params):
    D1 = params["D1"]
    D2 = params["D2"]
    L = params["L"]
    N_tubes = params["N_tubes"]
    D_shell = params["D_shell"]

    params["t_wall"] = (D2 - D1) / 2.0  # m, espesor de la pared del tubo
    params["A3"] = math.pi * (D1/2.0)**2  # m^2, sección de un tubo (for water flow area)
    params["A_int_total"] = N_tubes * math.pi * D1 * L  # m^2, área interna total de todos los tubos
    params["A_ext_total"] = N_tubes * math.pi * D2 * L  # m^2, área externa total de los tubos
    params["A_E_total"] = math.pi * D_shell * L  # m^2, área externa de la carcasa expuesta al ambiente
    params["V_vap_total"] = (math.pi * (D_shell/2.0)**2 * L) - (N_tubes * math.pi * (D2/2.0)**2 * L) # m^3, Volume available for vapor in shell
    return params

default_condenser_params = calculate_derived_geometrical_params(default_condenser_params.copy())

print("Default condenser parameters defined:")
for key, value in default_condenser_params.items():
    print(f"{key}: {value}")
```

```python
import math
import pandas as pd

# --- Attempt to use CoolProp for higher fidelity (opcional) ---
USE_COOLPROP = False
try:
    from CoolProp.CoolProp import PropsSI
    USE_COOLPROP = True
    # print("CoolProp available: using for properties.")
except Exception:
    # print("CoolProp NOT available: using default properties.")
    USE_COOLPROP = False

# ---------- Constants ----------
g = 9.81  # m/s2

# ---------- Default Properties (fallback) ----------
def default_props_water(T=293.15, p=101325.0):
    # Values for liquid water at ~20C and 1 atm
    return {"rho": 998.2, "cp": 4182.0, "k": 0.597, "mu": 1.002e-3, "nu": 1.004e-6}

def default_props_air(T=293.15, p=101325.0):
    # Values for air at ~20C and 1 atm
    return {"rho": 1.204, "cp": 1007.0, "k": 0.0255, "mu": 1.81e-5, "nu": 1.5e-5, "beta": 1.0/T, "Pr": 0.71}

def default_props_steam_sat(T_sat=373.15, p=101325.0):
    # Values for saturated steam at ~100C and 1 atm
    # This function needs to return T_sat as well
    return {"T_sat": T_sat, "rho_v": 0.597, "h_fg": 2257000.0, "cp_v": 2010.0, "mu_v": 1.2e-5, "k_v": 0.024}

# ---------- Functions for properties (uses CoolProp if available) ----------
def props_water(T=293.15, p=101325.0):
    # Temperature in K
    if USE_COOLPROP:
        try:
            rho = PropsSI("D", "T", T, "P", p, "Water")
            cp = PropsSI("C", "T", T, "P", p, "Water")
            k = PropsSI("L", "T", T, "P", p, "Water")
            mu = PropsSI("V", "T", T, "P", p, "Water")  # dynamic Pa.s
            nu = mu / rho
            return {"rho": rho, "cp": cp, "k": k, "mu": mu, "nu": nu}
        except Exception:
            return default_props_water(T,p)
    else:
        return default_props_water(T,p)

def props_air(T=293.15, p=101325.0):
    if USE_COOLPROP:
        try:
            rho = PropsSI("D", "T", T, "P", p, "Air")
            cp = PropsSI("C", "T", T, "P", p, "Air")
            k = PropsSI("L", "T", T, "P", p, "Air")
            mu = PropsSI("V", "T", T, "P", p, "Air")
            nu = mu / rho
            beta = 1.0/T # Ideal gas assumption
            Pr = cp * mu / k
            return {"rho": rho, "cp": cp, "k": k, "mu": mu, "nu": nu, "beta": beta, "Pr": Pr}
        except Exception:
            return default_props_air(T,p)
    else:
        return default_props_air(T,p)

def props_steam_sat(T_sat=373.15, p=101325.0):
    if USE_COOLPROP:
        try:
            # returns properties at saturation (vapor)
            # If T_sat is not provided, calculate from p
            if T_sat is None and p is not None:
                T_sat = PropsSI("T", "P", p, "Q", 1, "Water")

            rho_v = PropsSI("D", "T", T_sat, "Q", 1, "Water")
            h_v = PropsSI("H", "T", T_sat, "Q", 1, "Water") # J/kg
            h_f = PropsSI("H", "T", T_sat, "Q", 0, "Water")
            h_fg = h_v - h_f
            cp_v = PropsSI("C", "T", T_sat, "Q", 1, "Water")
            mu_v = PropsSI("V", "T", T_sat, "Q", 1, "Water")
            k_v = PropsSI("L", "T", T_sat, "Q", 1, "Water")
            return {"T_sat":T_sat, "rho_v": rho_v, "h_fg": h_fg, "cp_v": cp_v, "mu_v": mu_v, "k_v": k_v}
        except Exception:
            return default_props_steam_sat(T_sat, p)
    else:
        return default_props_steam_sat(T_sat, p)

# ---------- Correlaciones ----------
def nu_natural_churchill_chu(Gr, Pr):
    # Churchill & Chu for natural convection on horizontal cylinders
    Ra = Gr * Pr
    if Ra <= 0:
        return 0.0
    term = (0.387 * Ra**(1.0/6.0)) / (1.0 + (0.559/Pr)**(9.0/16.0))**(8.0/27.0)
    Nu = (0.6 + term)**2
    return Nu

def h_natural_convection_air(T_surface, T_inf, L_c, air_props=None):
    # L_c: characteristic length (diameter for horizontal cylinder)
    T_f = (T_surface + T_inf) / 2.0
    if air_props is None:
        air_props = props_air(T_f)

    beta = air_props.get("beta", 1.0 / max(T_f, 1.0))
    nu = air_props["nu"]
    Pr = air_props["Pr"]
    k_air = air_props["k"]

    Gr = g * beta * abs(T_surface - T_inf) * (L_c**3) / (nu**2)
    Nu = nu_natural_churchill_chu(Gr, Pr)
    h = Nu * k_air / L_c
    return max(h, 1e-4) # ensure h is not zero

def overall_UA(h_steam, A_ext, h_water, A_int, k_wall, t_wall, A_wall_mean, R_fouling):
    # All areas are total areas for the section being considered
    R_steam = 1.0 / (h_steam * A_ext) if h_steam > 0 and A_ext > 0 else 1e12
    R_wall = t_wall / (k_wall * A_wall_mean) if k_wall > 0 and t_wall > 0 and A_wall_mean > 0 else 0.0
    R_water = 1.0 / (h_water * A_int) if h_water > 0 and A_int > 0 else 1e12
    R_fouling_total = R_fouling # Assuming R_fouling is given per unit area and already scaled by area

    Rtot = R_steam + R_wall + R_water + R_fouling_total
    U_overall = 1.0 / Rtot if Rtot > 0 else 0.0
    return U_overall, Rtot

def h_convection_water(Re, Pr, D_hydraulic, k_water):
    # Dittus-Boelter correlation for turbulent flow in tubes (equation 15)
    # Nu = 0.023 * Re^0.8 * Pr^0.4 (for heating fluid)
    if Re < 2300: # Laminar flow
        Nu = 3.66 # Fully developed laminar flow in circular tube
    elif Re > 10000: # Turbulent flow, typical range for Dittus-Boelter
        Nu = 0.023 * (Re**0.8) * (Pr**0.4)
    else: # Transition flow, linear interpolation or other correlations can be used. For simplicity, we use the turbulent correlation.
        Nu = 0.023 * (Re**0.8) * (Pr**0.4)

    h = Nu * k_water / D_hydraulic
    return max(h, 1e-4) # ensure h is not zero

def h_film_condensation_horizontal_tube(T_sat, T_wall, D_outer, g=9.81):
    # Film condensation correlation for a single horizontal tube (Nusselt's theory)
    # Properties of the liquid condensate film are evaluated at the film temperature
    T_film = (T_sat + T_wall) / 2.0

    liquid_props = props_water(T_film) # Liquid properties (condensate)
    steam_props = props_steam_sat(T_sat) # Steam properties (vapor)

    rho_l = liquid_props["rho"]
    mu_l = liquid_props["mu"]
    k_l = liquid_props["k"]

    rho_v = steam_props["rho_v"]
    h_fg = steam_props["h_fg"]

    delta_T = T_sat - T_wall
    if delta_T <= 0:
        return 1e-4 # Avoid division by zero or negative delta_T

    # Nusselt's correlation for horizontal tube
    h_cond = 0.729 * ((g * rho_l * (rho_l - rho_v) * k_l**3 * h_fg) / (mu_l * delta_T * D_outer))**0.25

    return max(h_cond, 1e-4)

def discretized_condenser_adapted(params):
    # Extract parameters from the dictionary
    D1 = params["D1"]          # m, diámetro interior del tubo
    D2 = params["D2"]          # m, diámetro exterior del tubo
    L = params["L"]            # m, longitud efectiva del haz de tubos y carcasa
    N_tubes = params["N_tubes"]  # Nº de tubos en el haz
    N_sections = params["N_sections"] # Nº de secciones axiales para discretización
    mdot_c = params["mdot_c"]    # kg/s, Flujo másico de agua de enfriamiento
    Tc_in = params["Tc_in"]      # K, Temperatura de entrada del agua
    m_dot_v_in = params["m_dot_v_in"] # kg/s, Flujo másico de vapor entrante
    pc = params["pc"]          # Pa, Presión dentro de la carcasa
    T_ambient = params["T_ambient"] # K, Temperatura ambiente
    k_tube = params["k_tube"]    # W/mK, conductividad térmica del material del tubo
    t_wall = params["t_wall"]    # m, espesor de la pared del tubo
    R_fouling = params["R_fouling"] # m^2K/W, Resistencia por fouling (in/out)
    phi_bank = params["phi_bank"] # Factor empírico de corrección por banco de tubos
    D_shell = params["D_shell"]

    # Derived parameters for sections
    A_int_section = (N_tubes * math.pi * D1 * L) / N_sections # A_int_total / N_sections
    A_ext_section = (N_tubes * math.pi * D2 * L) / N_sections # A_ext_total / N_sections
    A_wall_mean_section = (N_tubes * math.pi * ((D1 + D2)/2) * L) / N_sections
    A_E_total = params["A_E_total"]
    A_E_section = A_E_total / N_sections

    # Calculate saturated steam properties at condenser pressure
    steam_sat_props = props_steam_sat(p=pc)
    T_sat = steam_sat_props["T_sat"] # Get T_sat from the returned dictionary
    h_fg = steam_sat_props["h_fg"]
    rho_v_sat = steam_sat_props["rho_v"]

    # Initial conditions for the loop
    T_water_current = Tc_in
    m_dot_v_current = m_dot_v_in
    results = []

    # Mass balance for cooling water (Eq. 6)
    rho_w_in = props_water(Tc_in, params["P_ref"])[ "rho"]
    V_water_tube = (mdot_c / N_tubes) / (rho_w_in * math.pi * (D1/2.0)**2)  # Average velocity in a single tube

    for i in range(N_sections):
        # --- Cooling Water Side (Tube-side) ---
        water_props = props_water(T_water_current, params["P_ref"])
        rho_w = water_props["rho"]
        mu_w = water_props["mu"]
        k_w = water_props["k"]
        cp_w = water_props["cp"]
        Pr_w = water_props["cp"] * water_props["mu"] / water_props["k"]

        # Reynolds number for water (Eq. 7)
        Re_water = (rho_w * D1 * V_water_tube) / mu_w # Corrected: Denominator is mu_w

        # Nusselt number for water and h_W (Eq. 15 and 14)
        Nu_water = h_convection_water(Re_water, Pr_w, D1, k_w) # D1 is hydraulic diameter for internal flow
        h_W = Nu_water * k_w / D1

        # --- Iterative Loop for Condensation and Wall Temperature ---
        T_wall_prev = T_sat - (T_sat - T_water_current) * 0.2 # Initial guess for T_wall

        tolerance = 1e-3 # K
        max_iterations = 100

        # Initialize variables before loop
        h_S_i = 0.0
        U_section = 0.0 # Initialize U_section

        for iter_count in range(max_iterations):
            # Condensation heat transfer coefficient (h_S_i)
            h_S_i = h_film_condensation_horizontal_tube(T_sat, T_wall_prev, D2, g) * phi_bank

            # Resistances (inverse of conductance)
            R_steam_conv = 1.0 / (h_S_i * A_ext_section) if h_S_i > 0 and A_ext_section > 0 else 1e12
            R_wall_cond = t_wall / (k_tube * A_wall_mean_section) if k_tube > 0 and A_wall_mean_section > 0 else 1e12
            R_water_conv = 1.0 / (h_W * A_int_section) if h_W > 0 and A_int_section > 0 else 1e12
            R_fouling_val = R_fouling # Assuming R_fouling is per unit area already scaled

            R_total_from_wall_to_water = R_wall_cond + R_water_conv + R_fouling_val

            # Solve for T_wall_new based on energy balance across resistances
            if (R_steam_conv + R_total_from_wall_to_water) > 0:
                T_wall_new = (T_sat * R_total_from_wall_to_water + T_water_current * R_steam_conv) / \
                             (R_steam_conv + R_total_from_wall_to_water)
            else:
                T_wall_new = T_wall_prev # Avoid division by zero, keep previous guess

            if abs(T_wall_new - T_wall_prev) < tolerance:
                T_wall_current = T_wall_new
                break
            T_wall_prev = T_wall_new
        else:
            T_wall_current = T_wall_new # If loop finishes without convergence

        # --- Heat Loss to Ambient (Q2) ---
        h_E = h_natural_convection_air(T_sat, T_ambient, D_shell, props_air((T_sat+T_ambient)/2.0))
        Q2_section = h_E * A_E_section * (T_sat - T_ambient)

        # Refine Q1_section_val Calculation: Use heat transfer from steam to wall after convergence
        # This Q1_section_val represents the heat released by condensation
        if R_steam_conv > 0:
            Q1_section_val = (T_sat - T_wall_current) / R_steam_conv
        else:
            Q1_section_val = 0.0 # No steam side heat transfer

        # Recalculate U_section with converged values
        U_section, _ = overall_UA(h_S_i, A_ext_section, h_W, A_int_section, k_tube, t_wall, A_wall_mean_section, R_fouling)

        # --- Energy Balance for Water ---
        T_water_out = T_water_current + (Q1_section_val - Q2_section) / (mdot_c * cp_w)

        # --- Mass Balance for Steam (Eq. 9) ---
        delta_m_cond_i = Q1_section_val / h_fg
        m_dot_v_current -= delta_m_cond_i

        results.append({
            "section": i + 1,
            "T_water_in": T_water_current,
            "T_water_out": T_water_out,
            "m_dot_steam_in": m_dot_v_in if i == 0 else results[i-1]['m_dot_steam_out'],
            "m_dot_steam_out": m_dot_v_current,
            "delta_m_cond": delta_m_cond_i,
            "Q1_section": Q1_section_val,
            "Q2_section": Q2_section,
            "h_W": h_W,
            "h_S_i": h_S_i,
            "h_E": h_E,
            "U_overall": U_section,
            "T_wall": T_wall_current
        })
        T_water_current = T_water_out # Update water temperature for next section

    return results

# Example usage with default parameters
condenser_results = discretized_condenser_adapted(default_condenser_params)

# Convert results to a pandas DataFrame for better visualization
df_results = pd.DataFrame(condenser_results)
print(df_results.head())
```

#### **Aplicaciones en un SMR**

#### 1. Diseño optimizado del condensador
Permite ajustar longitud, número y distribución de tubos para maximizar la eficiencia térmica en espacios compactos típicos de los SMR.

#### 2. Sistemas de enfriamiento eficientes
La predicción de la temperatura de salida del agua es crucial para diseñar torres de enfriamiento con bajo consumo de agua.

#### 3. Mitigación de ensuciamiento
El parámetro *R_fouling* evidencia la importancia de materiales que minimicen depósitos y mantengan alta eficiencia térmica.

#### 4. Integración y control del ciclo
El modelo puede conectarse al ciclo Rankine completo para desarrollar estrategias de control dinámico en un SMR.

#### 5. Monitoreo de desempeño
Los valores de h_W, h_S_i y U_overall sirven como línea base para evaluar degradación o ensuciamiento del condensador durante operación.
