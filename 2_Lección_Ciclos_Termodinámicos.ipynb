# Lección sobre Ciclos Termodinámicos
La termodinámica es una rama de la física que describe los equilibrios a nivel macroscópico. Los ciclos termodinámicos son una serie de procesos termodinámicos tales que el estado inicial y final son el mismo. Los motores de calor, las bombas de calor y los refrigeradores son ejemplos de dispositivos que operan en ciclos termodinámicos.

Cada ciclo se puede representar en varios tipos de diagramas, incluyendo diagramas P-V (Presión-Volumen), P-T (Presión-Temperatura) y S-T (Entropía-Temperatura). Estos diagramas nos ayudan a visualizar los procesos que ocurren en cada ciclo.

Además, discutiremos los coeficientes termodinámicos y las eficiencias de cada ciclo, así como las diferencias entre los ciclos ideales y los reales.

## Tipos de Ciclos Termodinámicos
Existen varios tipos de ciclos termodinámicos que son fundamentales en la ingeniería y la física. Algunos de los más importantes son:
1. **Ciclo de Carnot:** Es un ciclo idealizado compuesto por cuatro etapas reversibles: dos isotérmicas alternando con dos adiabáticas. Este ciclo es muy importante porque establece el límite de eficiencia que puede tener cualquier motor térmico.
2. **Ciclo de Rankine:** Este ciclo describe el funcionamiento de las plantas de potencia de vapor típicas, como las centrales eléctricas de carbón, nucleares o de biomasa.
3. **Ciclo de Brayton:** Este ciclo describe el funcionamiento de las turbinas de gas, que son la base de las plantas de energía de gas natural y los motores de los aviones a reacción.
4. **Ciclo de Otto:** Este ciclo describe el funcionamiento de los motores de combustión interna de encendido por chispa, que son el tipo de motor más comúnmente utilizado en los automóviles.
5. **Ciclo de Diesel:** Este ciclo describe el funcionamiento de los motores de combustión interna de encendido por compresión, que se utilizan en muchos vehículos pesados, como camiones y autobuses.

## Eficiencia de los Ciclos Termodinámicos
La eficiencia de un ciclo termodinámico se define como la relación entre la energía útil obtenida del ciclo (trabajo) y la energía que se ha tenido que aportar (calor). En términos matemáticos, se puede expresar de la siguiente manera:

![Eficiencia](https://latex.codecogs.com/svg.latex?%5Ceta%20%3D%20%5Cfrac%7BW%7D%7BQ%7D)

donde:
- ![Eta](https://latex.codecogs.com/svg.latex?%5Ceta) es la eficiencia del ciclo,
- ![W](https://latex.codecogs.com/svg.latex?W) es el trabajo obtenido del ciclo, y
- ![Q](https://latex.codecogs.com/svg.latex?Q) es el calor aportado al ciclo.

Es importante notar que la eficiencia siempre es un número entre 0 y 1, o expresado en porcentaje, entre 0% y 100%. Un ciclo con una eficiencia del 100% sería un ciclo que convierte toda la energía aportada en trabajo, sin pérdidas. Sin embargo, en la práctica, esto no es posible debido a las pérdidas inherentes en cualquier proceso real, como la fricción y la disipación de calor.

### Ciclo de Carnot
El ciclo de Carnot consta de cuatro etapas:
1. **Expansión isotérmica**: El gas se expande, realizando trabajo sobre el entorno. La temperatura del gas permanece constante debido al contacto térmico con la fuente de calor de alta temperatura.
2. **Expansión adiabática**: El gas continúa expandiéndose, realizando trabajo sobre el entorno. No hay intercambio de calor con el entorno, por lo que la temperatura del gas disminuye.
3. **Compresión isotérmica**: El gas se comprime, con el trabajo realizado sobre el gas por el entorno. La temperatura del gas permanece constante debido al contacto térmico con la fuente de calor de baja temperatura.
4. **Compresión adiabática**: El gas continúa comprimiéndose, con el trabajo realizado sobre el gas por el entorno. No hay intercambio de calor con el entorno, por lo que la temperatura del gas aumenta.
Al final de este ciclo, el gas ha vuelto a su estado inicial.

![Ciclo de Carnot](https://upload.wikimedia.org/wikipedia/commons/thumb/0/06/Carnot_cycle_p-V_diagram.svg/320px-Carnot_cycle_p-V_diagram.svg.png)

La eficiencia del ciclo de Carnot (η) se calcula como:

η = 1 - Tc/Th

donde:

- Tc es la temperatura de la fuente fría (en kelvin)
- Th es la temperatura de la fuente caliente (en kelvin)

```python
def carnot_efficiency(T_hot, T_cold):
    """
    Calcula la eficiencia de un ciclo de Carnot.

    Parámetros:
    T_hot : float
        Temperatura de la fuente de calor en kelvin.
    T_cold : float
        Temperatura de la fuente fría en kelvin.

    Devuelve:
    float
        Eficiencia del ciclo de Carnot.
    """
    return 1 - T_cold / T_hot

# Ejemplo de uso:
T_hot = 800  # Temperatura de la fuente de calor en K
T_cold = 300  # Temperatura de la fuente fría en K

efficiency = carnot_efficiency(T_hot, T_cold)
efficiency
```

### Ciclo de Otto
El ciclo de Otto es el ciclo idealizado que se utiliza para modelar los motores de encendido por chispa (como los motores de gasolina). Este ciclo consta de cuatro procesos principales:
1. **Admisión (Proceso 0-1):** Entra la mezcla de aire y combustible al cilindro.
2. **Compresión (Proceso 1-2):** La mezcla de aire y combustible se comprime adiabáticamente, lo que aumenta su temperatura y presión.
3. **Combustión (Proceso 2-3):** La chispa de la bujía provoca la combustión de la mezcla, lo que aumenta aún más la temperatura y la presión. Este proceso se modela como una adición de calor a volumen constante.
4. **Expansión (Proceso 3-4):** Los gases calientes empujan el pistón, realizando trabajo sobre él. Este proceso se modela como una expansión adiabática.
5. **Escape (Proceso 4-0):** Los gases de escape son expulsados del cilindro, y el ciclo se repite.

La eficiencia del ciclo de Otto se puede expresar en términos del ratio de compresión $r = V_1/V_2$ y la relación de calores específicos $\gamma = c_p/c_v$:

$$\eta = 1 - \frac{1}{r^{\gamma-1}}$$


![Ciclo de Otto](https://upload.wikimedia.org/wikipedia/commons/9/9f/P-V_Otto_cycle.svg)

La eficiencia del ciclo de Otto (η) se calcula como:

η = 1 - (1 / r^((γ - 1) / γ))

donde:

- r es la relación de compresión (volumen antes de la compresión / volumen después de la compresión)
- γ es el índice de adiabaticidad del aire

```python
from CoolProp.CoolProp import PropsSI
def otto_efficiency(r):
    """
    Calcula la eficiencia de un ciclo de Otto.

    Parámetros:
    r : float
        Relación de compresión (volumen antes de la compresión / volumen después de la compresión).

    Devuelve:
    float
        Eficiencia del ciclo de Otto.
    """
    # Calcular el índice de adiabaticidad (gamma) del aire
    gamma = PropsSI('Cpmass', 'T', 300, 'P', 101325, 'Air') / PropsSI('Cvmass', 'T', 300, 'P', 101325, 'Air')
    print(gamma)
    return 1 - (1 / r**(gamma - 1))

# Ejemplo de uso:
r = 4  # Relación de compresión

efficiency = otto_efficiency(r)
efficiency
```

```python
# Código en Python para calcular la eficiencia del ciclo Otto
def eficiencia_otto(T_h, T_c, r, gamma):
    """
    Calcula la eficiencia del ciclo Otto utilizando las temperaturas y la relación de compresión.

    Parámetros:
    - T_h: Temperatura de la fuente caliente (K)
    - T_c: Temperatura de la fuente fría (K)
    - r: Relación de compresión
    - gamma: Índice adiabático del gas

    Retorna:
    - eta: Eficiencia del ciclo Otto
    """
    # Calcular el calor rechazado a volumen constante
    Q_rechazado = T_h / (r**(gamma - 1)) - T_c

    # Calcular el calor añadido a volumen constante
    Q_anadido = T_h - T_c * (r**(gamma - 1))

    # Eficiencia del ciclo Otto
    eta = 1 - Q_rechazado / Q_anadido
    return eta

# Parámetros de ejemplo:
T_h = 800.0   # Temperatura caliente en Kelvin
T_c = 300.0   # Temperatura fría en Kelvin
r = 9.0       # Relación de compresión
gamma = 1.4   # Índice adiabático (para aire, por ejemplo)

# Cálculo de la eficiencia
eficiencia = eficiencia_otto(T_h, T_c, r, gamma)
print(f"Eficiencia del ciclo Otto: {eficiencia:.4f}")
```

### Ciclo Diesel

El ciclo Diesel consta de cuatro procesos principales:

1. **Admisión (Proceso 0-1):** Entra aire al cilindro.
2. **Compresión (Proceso 1-2):** El aire se comprime adiabáticamente, lo que aumenta su temperatura y presión.
3. **Combustión (Proceso 2-3):** Se inyecta combustible en el aire caliente, lo que provoca su combustión y aumenta aún más la temperatura y la presión. Este proceso se modela como una adición de calor a presión constante.
4. **Expansión (Proceso 3-4):** Los gases calientes empujan el pistón, realizando trabajo sobre él. Este proceso se modela como una expansión adiabática.
5. **Escape (Proceso 4-0):** Los gases de escape son expulsados del cilindro, y el ciclo se repite.

La eficiencia del ciclo Diesel se puede expresar en términos del ratio de compresión $r = V_1/V_2$, la relación de calores específicos $\gamma = c_p/c_v$ y el ratio de corte $\rho = V_3/V_2$:

$$\eta = 1 - \frac{1}{r^{\gamma-1}} \cdot \frac{\rho^\gamma - 1}{\gamma (\rho - 1)}$$

```python
# Definir el ratio de compresión, la relación de calores específicos y el ratio de corte
r = 16.0  # Ratio de compresión
gamma = 1.4  # Relación de calores específicos para el aire
rho = 2.0  # Ratio de corte

# Calcular la eficiencia del ciclo Diesel
efficiency = 1 - 1 / r**(gamma - 1) * ((rho**gamma - 1) / (gamma * (rho - 1)))

print(f'La eficiencia del ciclo Diesel es {efficiency:.2f}')
```

```python
def eficiencia_diesel(r, rho, gamma):
    """
    Calcula la eficiencia del ciclo Diesel ideal utilizando la fórmula estándar:

      η = 1 - (1 / r^(γ-1)) * ((ρ^γ - 1) / (γ*(ρ - 1)))

    Parámetros:
      - r: relación de compresión (V1/V2)
      - rho: relación de corte (cut-off ratio), definida como V3/V2
      - gamma: índice adiabático del gas

    Retorna:
      - eta: eficiencia del ciclo Diesel (valor adimensional)

    Importante:
      En el ciclo Diesel ideal se tienen las siguientes relaciones:
         T_2 = T_1 * r^(γ-1)
         T_3 = T_2 * rho = T_h  (temperatura máxima)
         T_4 = T_3 * (V_3/V_4)^(γ-1) = T_h * (ρ/r)^(γ-1)
         T_1 = T_h / (rho * r^(γ-1))
      Por ello, T_3 (o T_h) y T_4 están ligados a los parámetros r, ρ y γ.
    """
    eta = 1 - (1 / (r**(gamma - 1))) * ((rho**gamma - 1) / (gamma * (rho - 1)))
    return eta

# Ejemplo de uso:
r = 16.0        # Relación de compresión
rho = 2.0       # Relación de corte
gamma = 1.4     # Índice adiabático (por ejemplo, para aire)

eta_diesel = eficiencia_diesel(r, rho, gamma)
print(f"Eficiencia del ciclo Diesel ideal: {eta_diesel:.4f}")

# Para referencia, la eficiencia del ciclo Otto (en función de r y γ) es:
eta_otto = 1 - 1/(r**(gamma-1))
print(f"Eficiencia del ciclo Carnot ideal: {eta_otto:.4f}")

# También se puede calcular las temperaturas (en función de T_h) que quedan ligadas al ciclo:
T_h = 800.0  # Temperatura máxima, en Kelvin (estado 3)
T_4 = T_h * (rho / r)**(gamma - 1)
T_1 = T_h / (rho * r**(gamma - 1))
print(f"En el ciclo Diesel ideal, con T_h = {T_h:.1f} K:")
print(f"  T_4 (después de la expansión): {T_4:.1f} K")
print(f"  T_1 (inicio de compresión): {T_1:.1f} K")
```

### Ciclo Brayton
El ciclo de Brayton, también conocido como ciclo de Joule, es el ciclo fundamental de las turbinas de gas. Este ciclo consta de cuatro procesos principales:
1. **Admisión y Compresión (Proceso 1-2):** El aire es aspirado y comprimido adiabáticamente en el compresor, lo que aumenta su temperatura y presión.
2. **Combustión (Proceso 2-3):** Se inyecta combustible en el aire caliente, lo que provoca su combustión y aumenta aún más la temperatura y la presión. Este proceso se modela como una adición de calor a presión constante.
3. **Expansión (Proceso 3-4):** Los gases calientes se expanden adiabáticamente en la turbina, realizando trabajo sobre ella y sobre el compresor.
4. **Escape (Proceso 4-1):** Los gases de escape son expulsados, y el ciclo se repite.

La eficiencia del ciclo Brayton se puede expresar en términos del ratio de presión $r_p = P_2/P_1$ y la relación de calores específicos $\gamma = c_p/c_v$:

$$\eta = 1 - \frac{1}{r_p^{\frac{\gamma-1}{\gamma}}}$$

![Ciclo de Brayton](https://upload.wikimedia.org/wikipedia/commons/3/3c/Brayton_cycle.svg)

La eficiencia del ciclo de Brayton (η) se calcula como:

η = 1 - (1 / r_p^((γ - 1) / γ))

donde:

- r_p es la relación de presiones (presión después de la compresión / presión antes de la compresión)
- γ es el índice de adiabaticidad del aire

```python
def carnot_like_efficiency(T1, T3):
    """
    Calcula la eficiencia 'tipo Carnot' usando la temperatura
    más baja (T1) y la más alta (T3) de un ciclo Brayton idealizado.

    Parámetros:
    T1 : float
        Temperatura baja del ciclo (reservorio frío) en K.
    T3 : float
        Temperatura alta del ciclo (reservorio caliente) en K.

    Retorna:
    float
        Eficiencia tipo Carnot para esas dos temperaturas.
    """
    return 1 - (T1 / T3)

# Ejemplo de uso:
T1 = 300   # K, temperatura de entrada al compresor
T3 = 800  # K, temperatura máxima en la cámara de combustión

eff = carnot_like_efficiency(T1, T3)
print(f"Eficiencia 'tipo Carnot': {eff:.4f}")
```

```python
def eficiencia_brayton(T_h, T_c, r_c, r_e, gamma):
    """
    Calcula la eficiencia del ciclo Brayton ideal.

    Parámetros:
      - T_h: Temperatura máxima (estado 3) en Kelvin.
      - T_c: Temperatura mínima (estado 1) en Kelvin.
      - r_c: Relación de compresión (en el compresor).
      - r_e: Relación de expansión (en la turbina).
      - gamma: Índice adiabático del gas.

    La secuencia del ciclo es:
      1. Compresión isentrópica: T1 = T_c, T2 = T_c * r_c^((γ-1)/γ)
      2. Calentamiento a presión constante: T3 = T_h
      3. Expansión isentrópica: T4 = T_h / r_e^((γ-1)/γ)
      4. Rechazo de calor a presión constante: de T4 a T1 = T_c.

    Retorna:
      - eta: eficiencia del ciclo Brayton.

    Nota:
      La eficiencia se calcula mediante:
        η = 1 - [ (T_h / r_e^((γ-1)/γ) - T_c) / (T_h - T_c * r_c^((γ-1)/γ)) ].
    """
    # Temperaturas intermedias
    T2 = T_c * (r_c**((gamma - 1)/gamma))
    T4 = T_h / (r_e**((gamma - 1)/gamma))

    # Calor añadido (a presión constante)
    Q_in = T_h - T2
    # Calor rechazado (a presión constante)
    Q_out = T4 - T_c

    eta = 1 - (Q_out / Q_in)
    return eta

# Parámetros de ejemplo:
T_h = 800.0      # Temperatura máxima en Kelvin (estado 3)
T_c = 300.0       # Temperatura mínima en Kelvin (estado 1)
r_c = 10.0        # Relación de compresión del compresor
r_e = 10.0        # Relación de expansión de la turbina (idealmente igual a r_c en ciclo cerrado)
gamma = 1.4       # Índice adiabático (para aire, por ejemplo)

eta_brayton = eficiencia_brayton(T_h, T_c, r_c, r_e, gamma)
print(f"Eficiencia del ciclo Brayton ideal: {eta_brayton:.4f}")

# Para comparar, la eficiencia del ciclo de Carnot (que es el máximo teórico) es:
eta_carnot = 1 - T_c/T_h
print(f"Eficiencia del ciclo Carnot (entre {T_h} K y {T_c} K): {eta_carnot:.4f}")
```

### Ciclo Rankine
El ciclo de Rankine es el ciclo fundamental de las centrales eléctricas de vapor. Este ciclo consta de cuatro procesos principales:
1. **Admisión y Compresión (Proceso 1-2):** El agua es bombeada desde la condensadora a la caldera a presión constante.
2. **Calentamiento y Vaporización (Proceso 2-3):** El agua se calienta y se vaporiza en la caldera a presión constante.
3. **Expansión (Proceso 3-4):** El vapor se expande en la turbina, realizando trabajo sobre ella.
4. **Condensación (Proceso 4-1):** El vapor se condensa en el condensador a presión constante, y el ciclo se repite.

La eficiencia del ciclo Rankine se puede expresar en términos de las entalpías en los puntos clave del ciclo:

$$\eta = \frac{h_3 - h_4}{h_3 - h_2}$$


![Ciclo de Rankine](https://upload.wikimedia.org/wikipedia/commons/b/be/Rankine_cycle_Ts.png)

La eficiencia del ciclo de Rankine (η) se calcula como:

η = (W_turbine - W_pump) / W_turbine

donde:

- W_turbine es el trabajo realizado por la turbina
- W_pump es el trabajo realizado por la bomba

```python
def eficiencia_rankine(T_h, T_c, r_c, r_e, gamma):
    """
    Calcula la eficiencia aproximada del ciclo Rankine ideal.

    Parámetros:
      - T_h: Temperatura máxima en Kelvin (salida de la caldera, vapor supercalentado).
      - T_c: Temperatura mínima en Kelvin (entrada del condensador).
      - r_c: Relación de compresión (P_boiler / P_condensador).
      - r_e: Relación de expansión en la turbina (idealmente, r_e ≈ r_c).
      - gamma: Índice adiabático del vapor (valor aproximado, por ejemplo 1.3).

    Notas:
      - Se asume que el trabajo de la bomba es despreciable.
      - La expansión en la turbina es isentrópica, de modo que se aproxima:
            T4 = T_h / r_e^((γ-1)/γ)
      - Se asume que el proceso de calentamiento ocurre a presión constante,
        y que la entrada al condensador es casi a T_c.

    Retorna:
      - eta: eficiencia térmica del ciclo Rankine (valor adimensional).
    """
    # Temperatura a la salida de la turbina (expansión isentrópica)
    T4 = T_h / (r_e ** ((gamma - 1) / gamma))

    # Eficiencia aproximada del ciclo Rankine
    eta = 1 - (T4 - T_c) / (T_h - T_c)
    return eta

# Parámetros de ejemplo:
T_h = 800.0   # Temperatura máxima en Kelvin (vapor supercalentado en la caldera)
T_c = 300.0   # Temperatura mínima en Kelvin (temperatura del condensador)
r_c = 8.0    # Relación de compresión (P_boiler / P_condensador)
r_e = 8.0    # Relación de expansión en la turbina (idealmente igual a r_c)
gamma = 1.3   # Índice adiabático aproximado para el vapor (valor típico aproximado)

eta_rankine = eficiencia_rankine(T_h, T_c, r_c, r_e, gamma)
print(f"Eficiencia del ciclo Rankine ideal (aproximada): {eta_rankine:.4f}")

# Para referencia, la eficiencia del ciclo de Carnot operando entre T_h y T_c es:
eta_carnot = 1 - T_c/T_h
print(f"Eficiencia del ciclo Carnot (entre {T_h:.1f} K y {T_c:.1f} K): {eta_carnot:.4f}")
```

## Comparativa general de eficiencias

En esta sección se fijan las condiciones de operación comunes a todos los ciclos.

```python
# Condiciones de operación (puedes ajustar estos valores o usar ipywidgets más adelante)
T_caliente = 800.0 # K (≈527 °C)
T_fria = 300.0 # K (≈27 °C)
# Relación de presiones útil para Brayton, Rankine, etc.
p_caliente = 15e6 # Pa (ejemplo para vapor saturado a 350 °C)
p_fria = 100e3 # Pa (≈1 atm)
```

```python
import math

def eta_carnot(T_h, T_c):
    """Eficiencia de un ciclo de Carnot."""
    return 1.0 - T_c / T_h

def eta_otto(T_h, T_c, gamma=1.4):
    """
    Eficiencia de Otto ideal si fijamos T1=T_c y T3=T_h.
    Se obtiene la relación de compresión óptima r_c = (T_h/T_c)**(1/(gamma-1)).
    """
    r_c = (T_h / T_c)**(1/(gamma - 1))
    return 1.0 - 1.0 / (r_c**(gamma - 1))

def eta_diesel(T_h, T_c, gamma=1.4, rho=2.0):
    """
    Eficiencia Diesel ideal con cut-off rho=T3/T2.
    r_c se fija como en Otto; rho se deja como parámetro (valor típico 2.0).
    """
    r_c = (T_h / T_c)**(1/(gamma - 1))
    return 1.0 - (1/r_c**(gamma-1)) * (rho**gamma - 1)/(gamma*(rho-1))

def eta_brayton(T_h, T_c, gamma=1.4):
    """
    Eficiencia Brayton ideal cuando T3=T_h y T1=T_c.
    La r_p óptima es (T_h/T_c)**(gamma/(2*(gamma-1))).
    """
    r_p_opt = (T_h / T_c)**(gamma/(2*(gamma - 1)))
    return 1.0 - 1.0 / (r_p_opt**((gamma-1)/gamma))

def eta_stirling(T_h, T_c):
    """Eficiencia ideal de Stirling (igual a Carnot)."""
    return eta_carnot(T_h, T_c)

# Si trabajas con Rankine a vapor usa CoolProp/IAPWS-97:
import CoolProp.CoolProp as CP
def eta_rankine(T_h, T_c, p_h=p_caliente, p_c=p_fria):
    """
    Rankine ideal con saturación:
    – caldera a T_h, condensador a T_c
    – ignoramos recalentamiento y sobrecalentamiento de bomba
    """
    h1 = CP.PropsSI('H','T',T_c,'Q',0,'Water')      # punto 1: salida condensador
    s1 = CP.PropsSI('S','T',T_c,'Q',0,'Water')
    h2 = CP.PropsSI('H','P',p_h,'S',s1,'Water')     # salida de bomba (isentrópica)
    h3 = CP.PropsSI('H','T',T_h,'P',p_h,'Water')    # vapor saturado/seco
    s3 = CP.PropsSI('S','T',T_h,'P',p_h,'Water')
    h4 = CP.PropsSI('H','P',p_c,'S',s3,'Water')     # after turbine (isentropic)
    q_in  = h3 - h2
    w_out = (h3 - h4) - (h2 - h1)
    return w_out / q_in

# Diccionario práctico para la comparación
ETA_FUNCS = {
    "Carnot"  : eta_carnot,
    "Otto"    : eta_otto,
    "Diesel"  : eta_diesel,
    "Brayton" : eta_brayton,
    "Rankine" : eta_rankine,
    "Stirling": eta_stirling,
}
```
